在量子计算和量子信息理论中，克里福德门（Clifford gates）的严格数学定义是：能够将泡利群（Pauli group）映射到其自身的幺正算符。换言之，克里福德群是泡利群在幺正群中的规范化子（Normalizer）。

**严格的数学表述**
设 $\mathcal{P}_n$ 为 $n$ 量子比特的泡利群。该群由 $n$ 体系的泡利矩阵张量积 $\{I, X, Y, Z\}^{\otimes n}$ 乘上整体相位因子 $\{\pm 1, \pm i\}$ 构成。

如果一个 $n$ 量子比特的幺正算符 $C$ 满足，对于任意泡利算符 $P \in \mathcal{P}_n$，对其进行共轭变换后依然属于泡利群：
$$C P C^\dagger \in \mathcal{P}_n$$
那么 $C$ 就是一个克里福德门。所有满足该条件的算符构成了克里福德群（Clifford Group），通常记作 $\mathcal{C}_n$。

### 1. 泡利群与辛向量空间的同构

**叙述：**
任意 $n$ 比特泡利算符（忽略整体相位）可以完全由一个 $2n$ 维的二元向量 $v = (x, z) \in \mathbb{F}_2^{2n}$ 刻画：
$$P_v = i^{x \cdot z} \prod_{j=1}^n X_j^{x_j} Z_j^{z_j}$$
其中 $x, z \in \mathbb{F}_2^n$，这里的点乘表示内积。

两个泡利算符 $P_u$ 和 $P_v$ 的对易关系由它们向量的辛内积（Symplectic inner product）决定：
$$P_u P_v = (-1)^{\omega(u,v)} P_v P_u, \quad \text{其中} \quad \omega(u,v) = x_u \cdot z_v - x_v \cdot z_u \pmod 2$$
如果 $\omega(u,v) = 0$，则两者对易；如果 $\omega(u,v) = 1$，则两者反对易。

**证明：**

对于单量子比特，我们知道 $X$ 和 $Z$ 是反对易的，即 $XZ = -ZX$ 或 $XZ = (-1)ZX$。
一般地，可以写为 $X^a Z^b = (-1)^{ab} Z^b X^a$ 对于 $a,b \in \{0,1\}$。

考虑两个 $n$ 量子比特泡利算符（忽略相位）：
$P_u \propto X^{x_u} Z^{z_u} = X_1^{x_{u,1}}Z_1^{z_{u,1}} \otimes \cdots \otimes X_n^{x_{u,n}}Z_n^{z_{u,n}}$
$P_v \propto X^{x_v} Z^{z_v} = X_1^{x_{v,1}}Z_1^{z_{v,1}} \otimes \cdots \otimes X_n^{x_{v,n}}Z_n^{z_{v,n}}$

计算它们的乘积 $P_u P_v$：
$$P_u P_v \propto (X^{x_u} Z^{z_u}) (X^{x_v} Z^{z_v}) = \bigotimes_{j=1}^n (X_j^{x_{u,j}} Z_j^{z_{u,j}} X_j^{x_{v,j}} Z_j^{z_{v,j}})$$

在第 $j$ 个子系统中，交换 $Z_j^{z_{u,j}}$ 和 $X_j^{x_{v,j}}$ 会产生一个相位因子 $(-1)^{z_{u,j} x_{v,j}}$：
$$X_j^{x_{u,j}} Z_j^{z_{u,j}} X_j^{x_{v,j}} Z_j^{z_{v,j}} = (-1)^{z_{u,j} x_{v,j}} X_j^{x_{u,j}} X_j^{x_{v,j}} Z_j^{z_{u,j}} Z_j^{z_{v,j}}$$

同理，计算 $P_v P_u$：
$$P_v P_u \propto \bigotimes_{j=1}^n (X_j^{x_{v,j}} Z_j^{z_{v,j}} X_j^{x_{u,j}} Z_j^{z_{u,j}}) = \bigotimes_{j=1}^n (-1)^{z_{v,j} x_{u,j}} X_j^{x_{v,j}} X_j^{x_{u,j}} Z_j^{z_{v,j}} Z_j^{z_{u,j}}$$

比较两者，由于 $X_j^{x_{u,j}} X_j^{x_{v,j}} = X_j^{x_{u,j} \oplus x_{v,j}}$，它们仅相差一个总体符号：
$$P_u P_v = \left[ \prod_{j=1}^n (-1)^{z_{u,j} x_{v,j}} (-1)^{z_{v,j} x_{u,j}} \right] P_v P_u = (-1)^{\sum_{j=1}^n (z_{u,j} x_{v,j} - z_{v,j} x_{u,j})} P_v P_u$$

指数部分即为：
$$z_u \cdot x_v - z_v \cdot x_u \pmod 2 = x_v \cdot z_u - x_u \cdot z_v \pmod 2$$

因为在 $\mathbb{F}_2$ 中加法和减法等价，上式可以写为 $\omega(u,v) = x_u \cdot z_v - x_v \cdot z_u \pmod 2$。
这正是辛内积的形式。证毕。

### 2. 克里福德门作为辛变换

因为任意克里福德门 $C \in \mathcal{C}_n$ 必须保持对易关系（若 $[P_1, P_2]=0$，则 $[C P_1 C^\dagger, C P_2 C^\dagger]=0$），所以 $C$ 在 $\mathbb{F}_2^{2n}$ 上诱导了一个必须保持辛内积 $\omega(\cdot, \cdot)$ 的线性变换。

**补充说明：什么是辛内积，以及为什么克里福德变换必须保辛内积？**
*   **什么是辛内积**：正如前文所述，在二元向量空间 $\mathbb{F}_2^{2n}$ 中，两个向量 $u = (x_u, z_u)$ 和 $v = (x_v, z_v)$ 的辛内积定义为 $\omega(u,v) = x_u \cdot z_v - x_v \cdot z_u \pmod 2$。它的物理意义是完全决定了这两个向量所对应的泡利算符 $P_u$ 和 $P_v$ 是对易（内积为0）还是反对易（内积为1）。
*   **为什么保辛内积**：因为 $C$ 是一个幺正算符，幺正变换保持算符的代数关系。具体来说，考虑对易关系 $P_u P_v = \pm P_v P_u$。如果我们对系统施加克里福德门 $C$，新的算符变为 $P'_u = C P_u C^\dagger$ 和 $P'_v = C P_v C^\dagger$。由于 $C(P_u P_v)C^\dagger = (C P_u C^\dagger)(C P_v C^\dagger) = P'_u P'_v$，显然有：若 $P_u P_v = \pm P_v P_u$，则必定有 $P'_u P'_v = \pm P'_v P'_u$。这意味着变换前的对易性（符号为+，对应辛内积为0）或反对易性（符号为-，对应辛内积为1）被完全保留。因此，诱导出的在 $\mathbb{F}_2^{2n}$ 上的线性变换，在作用前后对应向量的辛内积必须相等。

这就构成了一个同态（此处尚未证明其为满射）：
$$\phi: \mathcal{C}_n \to \text{Sp}(2n, \mathbb{F}_2)$$

我们如果能证明 $\phi$ 的核是 $\mathcal{P}_n \times U(1)$，且能证明 $\phi$ 是满同态，那么我们就可以写出短正合列：
$$1 \to \mathcal{P}_n \times U(1) \to \mathcal{C}_n \xrightarrow{\phi} \text{Sp}(2n, \mathbb{F}_2) \to 1$$

所以，要证明完备性（即这几个门能生成整个克里福德群），只要能证明以下两点：
1. $H, S, \text{CNOT}$ 的映射能生成整个 $\text{Sp}(2n, \mathbb{F}_2)$（这顺便证明了满射性）。
2. 它们能生成核 $\mathcal{P}_n \times U(1)$。因此，在生成元集合中必须额外补充上产生任意整体相位的操作 $e^{i\theta}I$。

### 3. 同态 $\phi$ 的核为 $\mathcal{P}_n \times U(1)$ 的证明

**恒等映射的代数刻画**
设同态 $\phi: \mathcal{C}_n \to \mathrm{Sp}(2n, \mathbb{F}_2)$ 将任意克里福德算符 $C$ 映射为对泡利群 $\mathcal{P}_n$ 诱导的辛变换。如果 $C \in \ker(\phi)$，意味着 $\phi(C)$ 在 $\mathbb{F}_2^{2n}$ 上是恒等映射 $I$。在算符共轭层面，这等价于对于所有的泡利算符 $P \in \mathcal{P}_n$，$C P C^\dagger$ 在商去全局相位后等于 $P$ 本身。也就是说，必须存在一个依赖于 $P$ 的标量相位 $\lambda_P$，使得：
$$C P C^\dagger = \lambda_P P$$

**锁定相位 $\lambda_P \in \{+1, -1\}$**
对于生成元 $X_j$ 和 $Z_j$，它们满足厄密性（$P^\dagger = P$）以及对合性（$P^2 = I$）。幺正共轭操作 $C (\cdot) C^\dagger$ 不改变算符的这些代数性质。因此，$(C P C^\dagger)^2 = I$ 必然成立：
$$(\lambda_P P)^2 = \lambda_P^2 I = I$$
这要求 $\lambda_P^2 = 1$，从而得出相位只能是离散的正负号：$\lambda_P \in \{+1, -1\}$。设对于单比特生成元，共轭作用产生的正负号分别为：
$$C X_j C^\dagger = (-1)^{s_j} X_j$$
$$C Z_j C^\dagger = (-1)^{r_j} Z_j$$
其中 $s_j, r_j \in \{0, 1\}$，且 $j = 1, 2, \dots, n$。

**构造泡利匹配算符**
泡利算符的反对易性质恰好能完美生成这些正负号模式。考虑以下特定的泡利群元素：
$$P_{\text{corr}} = \prod_{j=1}^n Z_j^{s_j} X_j^{r_j}$$
我们来考察 $P_{\text{corr}}$ 对生成元 $X_k$ 和 $Z_k$ 的伴随作用。由于 $Z_k X_k Z_k = -X_k$ 且 $X_k Z_k X_k = -Z_k$，当 $P_{\text{corr}}$ 与 $X_k$（或 $Z_k$）交换位置时，只有当 $s_k = 1$（或 $r_k = 1$）时才会产生一个负号。因此：
$$P_{\text{corr}} X_k P_{\text{corr}}^\dagger = (-1)^{s_k} X_k$$
$$P_{\text{corr}} Z_k P_{\text{corr}}^\dagger = (-1)^{r_k} Z_k$$
这说明，泡利算符 $P_{\text{corr}}$ 对所有生成元产生的共轭符号变化，与算符 $C$ 完全一致。

**通过矩阵元计算证明算符相差全局相位**
前面的步骤不变，我们已经构造出了幺正矩阵 $U' = P_{\text{corr}}^\dagger C$，并且由定义已知它与所有的生成元对易：
$$U' X_k = X_k U', \quad U' Z_k = Z_k U', \quad \forall k \in \{1, \dots, n\}$$
我们现在要纯粹通过矩阵元素的计算，证明 $U'$ 必定是标量矩阵 $\exp[i\theta] I$。

**利用与 $Z_k$ 对易证明 $U'$ 是对角矩阵**
在这个 $2^n$ 维的复希尔伯特空间中，我们选取计算基底（Computational basis），记为 $\vert{}x\rangle$，其中 $x = x_1 x_2 \dots x_n \in \{0,1\}^n$ 是一个长度为 $n$ 的经典比特串。根据 $Z_k$ 的定义，它作用在计算基底上只产生正负号：
$$Z_k \vert{}x\rangle = (-1)^{x_k} \vert{}x\rangle$$
因为 $U'$ 与 $Z_k$ 对易，即 $U' Z_k - Z_k U' = 0$。我们将其两端分别夹在任意两个基底向量 $\langle x\vert{}$ 和 $\vert{}y\rangle$ 之间：
$$\langle x\vert{} (U' Z_k - Z_k U') \vert{}y\rangle = 0$$
展开并利用 $Z_k$ 作用在基底上的本征值：
$$\langle x\vert{} U' ((-1)^{y_k} \vert{}y\rangle) - (\langle x\vert{} Z_k) U' \vert{}y\rangle = 0$$
$$(-1)^{y_k} \langle x\vert{} U' \vert{}y\rangle - (-1)^{x_k} \langle x\vert{} U' \vert{}y\rangle = 0$$
提取公共的矩阵元 $U'_{xy} \equiv \langle x\vert{} U' \vert{}y\rangle$：
$$U'_{xy} \left[ (-1)^{y_k} - (-1)^{x_k} \right] = 0$$
上式对任意的比特位置 $k \in \{1, \dots, n\}$ 都必须成立。假设 $x \neq y$，那么必然存在至少一个位置 $k$，使得 $x_k \neq y_k$（即一个是 0，一个是 1）。对于这个位置 $k$，方括号内的值为 $(-1)^1 - (-1)^0 = -2$ 或 $2$，显然不等于 $0$。因此，要使等式成立，必须有：
$$U'_{xy} = 0 \quad (\text{当 } x \neq y \text{ 时})$$
这从第一性原理证明了：$U'$ 只能是一个对角矩阵。我们现在可以把 $U'$ 写成：
$$U' = \sum_{x} c_x \vert{}x\rangle \langle x\vert{}$$
其中 $c_x$ 是对角线上的元素。

**利用与 $X_k$ 对易证明对角元全部相等**
根据 $X_k$ 的定义，它是一个翻转门，作用在基底上会将第 $k$ 位的比特反转。我们记翻转后的比特串为 $x \oplus e_k$ （其中 $e_k$ 是第 $k$ 位为 1，其余为 0 的串）。
$$X_k \vert{}x\rangle = \vert{}x \oplus e_k\rangle$$
已知 $X_k U' = U' X_k$。因为 $X_k$ 是厄密的且 $X_k^2 = I$，我们可以将等式写为：
$$X_k U' X_k = U'$$
将我们刚才得出的对角阵形式代入等式左边：
$$X_k \left( \sum_{x} c_x \vert{}x\rangle \langle x\vert{} \right) X_k = \sum_{x} c_x (X_k \vert{}x\rangle) (\langle x\vert{} X_k)$$
$$= \sum_{x} c_x \vert{}x \oplus e_k\rangle \langle x \oplus e_k\vert{}$$
由于求和号是遍历所有可能的 $x$，我们不妨做个变量代换，令 $y = x \oplus e_k$（那么 $x = y \oplus e_k$）。上面的式子重写为：
$$= \sum_{y} c_{y \oplus e_k} \vert{}y\rangle \langle y\vert{}$$
现在我们将它与等式右边原封不动的 $U' = \sum_{y} c_y \vert{}y\rangle \langle y\vert{}$ 进行对比。因为计算基底是正交的，两个对角矩阵相等意味着对应位置的矩阵元必须逐个相等：
$$c_{y \oplus e_k} = c_y$$
这个结论的物理图景非常明确：对于任意一个比特串 $y$，如果我们翻转它的第 $k$ 位，对应的对角元数值不变。由于这个结论对任意的 $k \in \{1, \dots, n\}$ 都成立，我们可以通过逐位翻转的方式，从全 0 串 $00\dots0$ 变到任意的比特串 $x$。这直接证明了：
$$c_x = c_{00\dots0} \equiv c, \quad \forall x \in \{0, 1\}^n$$

**得出最终结论**
既然所有的对角元都等于同一个常数 $c$，矩阵 $U'$ 就被严格写成了：
$$U' = \sum_{x} c \vert{}x\rangle \langle x\vert{} = c I$$
由于 $U' = P_{\text{corr}}^\dagger C$ 是两个幺正算符的乘积，所以 $U'$ 本身必须是幺正的：
$$(c I)(c I)^\dagger = \vert{}c\vert{}^2 I = I \implies \vert{}c\vert{} = 1$$
因此 $c$ 是一个纯相位 $c = \exp[i\theta]$。我们不需要调用任何外部的高级定理，仅仅通过写出矩阵元素并比较，就严格得到了：
$$P_{\text{corr}}^\dagger C = \exp[i\theta] I \implies C = \exp[i\theta] P_{\text{corr}}$$
证毕。由此 $\ker(\phi) = \mathcal{P}_n \times U(1)$。

### 4. 证明基本门集生成整个辛群 $\text{Sp}(2n, \mathbb{F}_2)$

这里我们将显式地证明：$\{H, S, \text{CNOT}\}$ 诱导的线性映射能够生成整个辛群 $\text{Sp}(2n, \mathbb{F}_2)$。整个证明可通过以下清晰的代数逻辑展开。

**将门映射为 $\mathbb{F}_2^{2n}$ 上的线性变换**

我们定义 $2n$ 维向量空间 $V = \mathbb{F}_2^{2n}$。取一组标准辛基底：$e_1, \dots, e_n$（对应单比特的 $X$ 算符）和 $f_1, \dots, f_n$（对应单比特的 $Z$ 算符）。辛内积定义为：
$$\omega(u, v) = u^T \Omega v \pmod 2$$
其中基础关系为 $\omega(e_i, e_j) = 0$、$\omega(f_i, f_j) = 0$、$\omega(e_i, f_j) = \delta_{ij}$。

群 $\text{Sp}(2n, \mathbb{F}_2)$ 就是所有满足 $\omega(Mu, Mv) = \omega(u, v)$ 的可逆线性映射 $M$ 的集合。根据这三个门对泡利生成元的共轭作用，它们在 $V$ 上诱导的矩阵变换为：

*   **$H_i$（阿达马）**：交换 $e_i$ 和 $f_i$。
$$H_i(e_i) = f_i, \quad H_i(f_i) = e_i$$

*   **$S_i$（相位）**：
$$S_i(e_i) = e_i + f_i, \quad S_i(f_i) = f_i$$

*   **$\text{CNOT}_{c \to t}$**：
$$\text{CNOT}(e_c) = e_c + e_t, \quad \text{CNOT}(e_t) = e_t$$
$$\text{CNOT}(f_t) = f_t + f_c, \quad \text{CNOT}(f_c) = f_c$$

所有门对未标出的其他基底向量均表现为恒等映射。

**证明基本门集能够在 $V \setminus \{0\}$ 上传递（Transitivity）**

我们需要证明：利用群 $G = \langle H, S, \text{CNOT} \rangle$，我们可以把任意一个非零向量 $v \in V$ 映射到特定的基底向量 $f_1$。设任意非零向量 $v = \sum_{k=1}^n (x_k e_k + z_k f_k)$。

*块内归化*
考察第 $k$ 个子空间的坐标 $(x_k, z_k)$。我们用 $H_k$ 和 $S_k$ 操作它：
*   若 $(x_k, z_k) = (1, 0)$，应用 $H_k$，变为 $(0, 1)$，即局部变为了 $f_k$。
*   若 $(x_k, z_k) = (1, 1)$，其对应向量 $e_k + f_k$。我们依次应用 $H_k S_k$：
$$S_k(e_k + f_k) = (e_k + f_k) + f_k = e_k$$
$$H_k(e_k) = f_k$$
所以 $H_k S_k$ 将 $(1, 1)$ 映射为 $(0, 1)$。

通过局部的 $H$ 和 $S$ 操作，我们可以把原向量转化为 $v' = \sum_{k \in I} f_k$，其中 $I$ 是非空下标集。

*跨块坍缩*
现在我们需要把 $\sum_{k \in I} f_k$ 坍缩到单一位上。利用 $\text{CNOT}_{j \to i}$ 对 $f$ 基底的作用 $f_i \mapsto f_i + f_j$：对于向量 $f_i + f_j$，我们对其作用 $\text{CNOT}_{j \to i}$，结果是 $(f_i + f_j) + f_j = f_i$。通过反复选择 $I$ 中的两个下标，应用 CNOT，我们可以将求和项逐个消掉，直到只剩下一个 $f_m$。

*对齐到 $f_1$*
如果 $m \neq 1$，我们需要交换第 $1$ 块和第 $m$ 块。SWAP 门可由三个 CNOT 精确构造：$\text{CNOT}_{1\to m}\text{CNOT}_{m\to 1}\text{CNOT}_{1\to m}$。它将 $f_m$ 完美映射为 $f_1$。

**推论**：群 $G$ 作用在 $V \setminus \{0\}$ 上是传递的。任意 $v \neq 0$，存在 $M_v \in G$ 使得 $v = M_v f_1$。

**定义并构造全体辛平移（Symplectic Transvections）**

在线性代数中，沿着向量 $u$ 的辛平移定义为线性映射：
$$T_u(w) = w + \omega(w, u)u$$
可以极容易验证 $T_u$ 保持辛内积，即 $T_u \in \text{Sp}(2n, \mathbb{F}_2)$。辛平移有一个极为重要的代数共轭性质：对于任意辛矩阵 $M$，有 $M T_u M^{-1} = T_{Mu}$。

回看我们的相位门 $S_1$，按定义它作用为 $e_1 \mapsto e_1 + f_1$ 且 $f_1 \mapsto f_1$。检查 $T_{f_1}$ 的作用：
$$T_{f_1}(e_1) = e_1 + \omega(e_1, f_1)f_1 = e_1 + 1\cdot f_1 = e_1 + f_1$$
$$T_{f_1}(f_1) = f_1 + \omega(f_1, f_1)f_1 = f_1 + 0 = f_1$$
这说明：$S_1$ 在代数上严格等价于辛平移 $T_{f_1}$。结合前述的传递性，对于任意非零向量 $v$，我们有 $v = M_v f_1$（其中 $M_v \in G$）。因此：
$$T_v = T_{M_v f_1} = M_v T_{f_1} M_v^{-1} = M_v S_1 M_v^{-1}$$
因为 $M_v, S_1 \in G$，所以 $T_v \in G$。

**结论**：群 $G = \langle H, S, \text{CNOT} \rangle$ 包含了 $\text{Sp}(2n, \mathbb{F}_2)$ 中的所有辛平移。

**证明辛平移生成整个 $\text{Sp}(2n, \mathbb{F}_2)$（核心闭环）**

现在我们只需要证明，任意一个辛矩阵 $M \in \text{Sp}(2n, \mathbb{F}_2)$ 都可以被拆解为辛平移的乘积。
设 $M$ 将标准基映射为新的辛基底 $E_i = M e_i$ 和 $F_i = M f_i$。我们要构造一系列辛平移，把 $(E_1, F_1)$ 映射回 $(e_1, f_1)$。

*将 $E_1$ 映射到 $e_1$*

*   **情况 A**：若 $\omega(E_1, e_1) = 1$。直接施加 $T_{E_1+e_1}$：
$$T_{E_1+e_1}(E_1) = E_1 + \omega(E_1, E_1+e_1)(E_1+e_1) = E_1 + 1\cdot(E_1+e_1) = e_1$$
一步完成。

*   **情况 B**：若 $\omega(E_1, e_1) = 0$。我们需要找一个“跳板”向量 $u$，满足 $\omega(E_1, u) = 1$ 且 $\omega(u, e_1) = 1$。设 $E_1 = \sum (x_i e_i + z_i f_i)$。因内积为 0 知 $z_1 = 0$。由于 $E_1 \neq 0$，必存在另一非零坐标：
    *   若某 $z_k = 1$ $(k>1)$，令 $u = f_1 + e_k$。验证可知 $\omega(E_1, u) = x_1 + 1 = 1$，且 $\omega(u, e_1) = 1 + 0 = 1$。
    *   若所有 $z_k = 0$，则某 $x_k = 1$ $(k>1)$。令 $u = f_1 + f_k$。验证可知 $\omega(E_1, u) = x_1 + 1 = 1$，且 $\omega(u, e_1) = 1 + 0 = 1$。

找到 $u$ 后，连续施加两个辛平移 $T_{e_1+u} T_{E_1+u}$：先作用 $T_{E_1+u}(E_1) = u$，再作用 $T_{e_1+u}(u) = e_1$。映射完成。

*将 $F_1$ 映射到 $f_1$ 且不移动 $e_1$*

经过上述操作后，基底变成了 $\{e_1, F_1', \dots\}$。因辛结构保持不变，必有 $\omega(e_1, F_1') = 1$。令 $w = F_1' + f_1$。注意 $\omega(e_1, w) = \omega(e_1, F_1') + \omega(e_1, f_1) = 1 + 1 = 0$。因为内积为 0，$T_w$ 作用在 $e_1$ 上等于自身，绝对不会破坏刚才对齐的 $e_1$。

*   **情况 A**：若 $\omega(F_1', f_1) = 1$。
$$T_w(F_1') = F_1' + \omega(F_1', w)w = F_1' + \omega(F_1', f_1)(F_1'+f_1) = F_1' + (F_1'+f_1) = f_1$$
一步完成。

*   **情况 B**：若 $\omega(F_1', f_1) = 0$。我们先施加 $T_{e_1}$。显然 $T_{e_1}(e_1) = e_1$。看它对 $F_1'$ 的作用：$F_1'' = T_{e_1}(F_1') = F_1' + \omega(F_1', e_1)e_1 = F_1' + e_1$。此时考察新向量 $F_1''$ 与 $f_1$ 的内积：$\omega(F_1'+e_1, f_1) = 0 + 1 = 1$。这完美转化为了情况 A！随后再施加 $T_{F_1''+f_1}$ 即可将其送至 $f_1$。

**归纳总结**：我们纯粹用辛平移构成的算符 $U_1$，完成了 $(E_1, F_1) \mapsto (e_1, f_1)$ 的映射。剩余的基底 $\{E_k, F_k\}_{k=2}^n$ 现在被限制在了与 $(e_1, f_1)$ 严格正交的 $2n-2$ 维子空间中。我们在该子空间内递归执行上述过程，构造出的后续辛平移其平移向量仅存在于子空间内，它们与 $e_1, f_1$ 内积恒为 0，因此绝对不会扰动已经排好的基底。直到所有基底对齐完毕，我们显式证明了任意 $M \in \text{Sp}(2n, \mathbb{F}_2)$ 均可写成有限个辛平移的乘积。

### 5. 生成泡利群核

这非常简单，纯靠 $S$ 和 $H$ 即可生成单比特的泡利群：
$$S^2 = Z$$
$$H Z H = X$$
$$Y = i X Z$$
结合张量积，我们已经得到了整个 $\mathcal{P}_n$。

### 6. 哥特斯曼-基尔定理（Gottesman-Knill Theorem）

哥特斯曼-基尔定理（Gottesman-Knill Theorem）是量子计算理论中最具反直觉深度的定理之一。它证明了：一个仅仅由初始计算基底、克里福德门和泡利基底测量构成的量子线路，无论产生多么复杂的“多体纠缠”，都可以被经典计算机在多项式时间内完美模拟。为了严格证明这一定理，我们需要彻底抛弃维度高达 $2^n$ 的薛定谔态矢图像（State Vector），转而使用海森堡图像（Heisenberg Picture）下的稳定子体系（Stabilizer Formalism）。证明的核心在于构造一个算法，证明量子态的初始化、演化和测量这三个基本过程，在稳定子表述下均只需多项式时间（$\mathcal{O}(n^2)$ 或 $\mathcal{O}(n^3)$）的经典计算开销。

**态的经典表示（初始化，空间复杂度 $\mathcal{O}(n^2)$）**

对于一个 $n$ 量子比特的纯态 $\vert{}\psi\rangle$，如果存在一个由 $n$ 个相互对易、且彼此独立的泡利算符组成的生成元集合 $\langle g_1, g_2, \dots, g_n \rangle$，满足：
$$g_i \vert{}\psi\rangle = (+1) \vert{}\psi\rangle, \quad \forall i \in \{1, \dots, n\}$$
我们就称 $\vert{}\psi\rangle$ 是一个稳定子态。由于这 $n$ 个算符在 $2^n$ 维空间中唯一确定了一个非退化的共同本征态（即 $\vert{}\psi\rangle$），我们只需要追踪这 $n$ 个生成元，就能等价地描述这个复杂的量子态。

在量子线路的初始时刻，系统处于 $\vert{}00\dots 0\rangle$。这个态显然是 $Z_1, Z_2, \dots, Z_n$ 的共同 $+1$ 本征态。因此，初始状态的生成元即为：
$$g_i = Z_i \quad (i = 1, \dots, n)$$
如我们前文所述，每个泡利算符 $g_i$ 可以用一个 $2n$ 维的布尔向量外加一个相位位（$\pm 1$）来精确表示。追踪 $n$ 个生成元，只需要维护一个 $n \times (2n+1)$ 的布尔矩阵（称为稳定子表，Stabilizer Tableau）。
**空间开销**：$\mathcal{O}(n^2)$，属于多项式级别。

**酉演化的经典跟踪（时间复杂度 $\mathcal{O}(n^2)$）**

当系统经过一个克里福德门 $C$ 演化时，态变为 $\vert{}\psi'\rangle = C \vert{}\psi\rangle$。我们考察原稳定子生成元的共轭：
$$(C g_i C^\dagger) \vert{}\psi'\rangle = C g_i C^\dagger C \vert{}\psi\rangle = C g_i \vert{}\psi\rangle = C (+1 \vert{}\psi\rangle) = (+1) \vert{}\psi'\rangle$$
这意味着，演化后的新态 $\vert{}\psi'\rangle$ 的稳定子生成元，恰好就是 $g_i' = C g_i C^\dagger$。由于 $C$ 是克里福德门，根据定义，$g_i'$ 依然是泡利算符。我们只需根据 $H, S, \text{CNOT}$ 对单比特泡利算符的映射规则（即前文论述的伴随作用），更新这 $n$ 个生成元对应的布尔向量即可。更新 $n$ 个向量中的局部比特，每一次门操作需要的时间与系统的比特数成正比。
**时间开销**：每个克里福德门导致 $\mathcal{O}(n)$ 或 $\mathcal{O}(n^2)$ 的经典计算时间，属于多项式级别。

**测量的经典模拟（核心难点，时间复杂度 $\mathcal{O}(n^3)$）**

这是证明中最精彩的部分：当我们在系统上演化了一半，决定对某个泡利算符 $M$（例如测量第一比特的 $Z_1$）进行测量时，量子力学告诉我们要以特定概率发生波函数坍缩。我们要证明，这个概率和坍缩后的态都能被经典快速计算。

假设当前态由生成元 $G = \langle g_1, \dots, g_n \rangle$ 稳定。测量算符为 $M$（必定是一个泡利算符）。我们逐一检查 $M$ 与当前所有生成元 $g_i$ 的对易关系（通过计算前文提到的辛内积，这只需 $\mathcal{O}(n^2)$ 时间）。存在两种截然不同的物理情况：

*   **测量结果是完全随机的（存在反对易）**
若 $M$ 至少与一个生成元（设为 $g_p$）反对易，即 $\{M, g_p\} = 0$。这意味着当前态 $\vert{}\psi\rangle$ 不可能是 $M$ 的本征态，测量 $M$ 将以 **精确的 $50\% : 50\%$ 概率** 得到结果 $+1$ 或 $-1$。
在经典模拟器中，我们只需调用一个随机数生成器产生一个结果 $c \in \{+1, -1\}$ 即可。
坍缩后的态如何更新？若还有其他生成元 $g_q$ 也与 $M$ 反对易，我们令 $g_q' = g_q g_p$。因为两个与 $M$ 反对易的算符相乘，结果必定与 $M$ 对易。通过这一步，我们可以构造出一组新的等价生成元，使得有且仅有 $g_p$ 依然与 $M$ 反对易。
测量完成后，系统坍缩到 $M$ 的本征态，本征值为 $c$。因此 $c M$ 成为新的稳定子。原本的 $g_p$ 因为与 $M$ 反对易，其稳定子性质被测量破坏，必须被踢出生成元集合。我们用 $c M$ 替换掉 $g_p$。新的生成元集合变为 $\langle g_1', \dots, g_{p-1}', c M, g_{p+1}', \dots, g_n' \rangle$。
**时间开销**：计算对易并进行行代换（高斯消元），复杂度为 $\mathcal{O}(n^2)$。

*   **测量结果是决定性的（全对易）**
若 $M$ 与所有的生成元 $g_i$ 都能对易，即 $[M, g_i] = 0, \forall i$。在最大对易集构成的希尔伯特空间中，这意味着 $M$ 本身必定已经存在于由 $G$ 生成的群中（确切地说，要么 $M \in G$，要么 $-M \in G$）。此时，测量将以 $100\%$ 的概率得到一个确定性的结果，且测量不改变当前的量子态（不发生坍缩更新）。
如何算出这个决定性的结果？因为 $M$ 是生成元 $g_1, \dots, g_n$ 的乘积（可能带正负号），我们只需要在 $\mathbb{F}_2$ 向量空间中，解一个线性方程组：
$$M = (-1)^s \prod_{i=1}^n g_i^{x_i}$$
找出系数 $x_i \in \{0, 1\}$。由于所有算符都可以用长度为 $2n$ 的二元向量表示，这等价于解一个有 $n$ 个未知数、$2n$ 个方程的经典线性方程组。我们使用二元高斯消元法即可求解。解出 $\{x_i\}$ 后，将对应的 $g_i$ 矩阵相乘，追踪前面产生的相位累积。如果最终结果是 $M$，则测量输出 $+1$；如果最终结果是 $-M$，则测量输出 $-1$。
**时间开销**：二元高斯消元求解线性方程组，复杂度为 $\mathcal{O}(n^3)$。

**证明终结（复杂度总结）**

综合以上三步：
*   长度为 $n$ 的初始态可以在 $\mathcal{O}(n^2)$ 空间和时间内被经典构建。
*   线路中包含的每一个克里福德门演化，需要消耗 $\mathcal{O}(n^2)$ 时间模拟。
*   线路中的每一次测量，需要消耗至多 $\mathcal{O}(n^3)$ 的时间模拟，并给出完全等效的物理概率分布。

假设一个量子线路包含 $n$ 个量子比特，以及总共 $m$ 个克里福德门和测量操作。那么，利用上述稳定子跟踪算法，经典计算机模拟输出这个量子线路完整结果的总时间复杂度为：
$$\mathcal{O}(m \cdot n^3)$$
因为 $m$ 和 $n$ 的多项式时间（Polynomial Time）属于复杂性理论中的 $\text{P}$ 类，我们从第一性原理和线性代数严格证明了：**纯克里福德量子线路不能解决任何经典计算机无法在多项式时间内解决的问题。** Q.E.D.

### 7. 密度算符表述与退相干模拟

经典计算机在模拟量子计算时，通常不可能去显式地追踪整个指数级庞大的概率分布向量（或密度矩阵）。然而，在纯克里福德线路下，系统的密度算符被极其强烈的代数对称性锁定在了一个极度狭窄的流形上。经典计算机不需要记录整个 $2^n \times 2^n$ 的密度矩阵，它只需要记录这个密度矩阵的“生成元”，就能在多项式时间内完美、全局地追踪整个密度算符的演化。以下使用密度算符（Density Operator）的语言来重构这一证明。

**一、 稳定子密度算符的代数压缩**

对于一个 $n$ 量子比特的纯态，如果它是稳定子群 $\mathcal{S} = \langle g_1, \dots, g_n \rangle$ 的共同 $+1$ 本征态，那么它的密度算符 $\rho$ 有一个极其优雅的严格解析式：
$$\rho = \frac{1}{2^n} \sum_{P \in \mathcal{S}} P$$
由于群 $\mathcal{S}$ 完全由 $n$ 个独立的生成元决定，上述密度算符可以等价地写成完全因式分解的形式：
$$\rho = \frac{1}{2^n} \prod_{i=1}^n (I + g_i)$$

在纯克里福德线路中，系统无论怎么纠缠，其密度矩阵永远保持上述因式分解的形态。经典计算机并不需要去计算或存储这 $2^n$ 个群元素的系数，它只需要在内存中存下这 $n$ 个生成元 $\{g_1, \dots, g_n\}$（即 $n$ 个长度为 $2n$ 的布尔向量）。这就把存储一个 $2^n \times 2^n$ 矩阵的指数级空间开销，暴力压缩成了 $\mathcal{O}(n^2)$ 的多项式空间。

**二、 幺正演化：海森堡图像与薛定谔图像的等价**

当系统经过一个克里福德门 $C$ 演化时，在密度算符层面（薛定谔图像），新的状态为：
$$\rho' = C \rho C^\dagger$$
将刚才的因式分解形式代入：
$$\rho' = C \left( \frac{1}{2^n} \prod_{i=1}^n (I + g_i) \right) C^\dagger = \frac{1}{2^n} \prod_{i=1}^n C (I + g_i) C^\dagger = \frac{1}{2^n} \prod_{i=1}^n (I + C g_i C^\dagger)$$
因为 $C$ 是克里福德门，根据定义，$g_i' = C g_i C^\dagger$ 依然是一个合法的泡利算符。因此，新的密度算符 $\rho'$ 完美地保持了稳定子形态。经典计算机只需用 $\mathcal{O}(n^2)$ 的时间去更新那 $n$ 个生成元。我们全局地更新了整个密度矩阵，而没有丢失任何信息。

**三、 测量操作与密度算符坍缩**

假设我们对某个泡利算符 $M$ 进行测量。测量的投影算符为 $\Pi_{\pm} = \frac{1}{2}(I \pm M)$。根据量子力学的公理，测量得到 $\pm 1$ 的概率分别为 $p_{\pm} = \text{Tr}(\rho \Pi_{\pm})$。坍缩后的密度算符为 $\rho_{\pm} = \frac{1}{p_{\pm}} \Pi_{\pm} \rho \Pi_{\pm}$。我们将密度算符的群求和形式 $\rho = \frac{1}{2^n} \sum_{P \in \mathcal{S}} P$ 代入测量过程。这里同样分为两种情况：

*   **情况 A：$M$ 与所有生成元对易 (确定性结果)**
如果 $M$ 与所有的 $P \in \mathcal{S}$ 对易，由于 $\mathcal{S}$ 是最大对易群，必然有 $M \in \mathcal{S}$ 或 $-M \in \mathcal{S}$。假设 $M \in \mathcal{S}$，那么对于所有的群元素，$\rho M = \rho$ 成立（这在物理上意味着 $\rho$ 处于 $M$ 的 $+1$ 偏振态）。此时我们计算概率：
$$p_+ = \text{Tr}\left( \rho \frac{I + M}{2} \right) = \frac{1}{2}(\text{Tr}(\rho) + \text{Tr}(\rho M)) = \frac{1}{2}(1 + 1) = 1$$
概率为 $1$，密度算符不变：$\rho_+ = \rho$。经典计算机只需解线性方程组判定 $M$ 是否在群内即可。

*   **情况 B：$M$ 与某个生成元反对易 (随机性结果)**
如果 $M$ 与至少一个生成元 $g_k$ 反对易，这意味着群 $\mathcal{S}$ 可以被平均分为两半：一半与 $M$ 对易，记为子群 $\mathcal{S}^+$；另一半与 $M$ 反对易，可以写成 $g_k \mathcal{S}^+$。于是我们把密度矩阵拆开：
$$\rho = \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} P + \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} g_k P$$
因为 $M$ 与 $\mathcal{S}^+$ 对易，与 $g_k$ 反对易，我们可以轻易算出：
$$M \rho = \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} M P + \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} M g_k P = \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} P M - \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} g_k P M = \rho M - \text{something}$$
经过严格求迹发现，反对易部分的迹恒为 $0$，最终得到 $\text{Tr}(\rho M) = 0$。这意味着概率 $p_+ = p_- = 1/2$。

最绝妙的一步在于坍缩后的密度算符计算：假设经典计算机随机抛硬币得到了 $+1$，投影算符为 $\Pi_+ = \frac{I+M}{2}$。由于 $\Pi_+$ 是投影子，$\Pi_+ \Pi_+ = \Pi_+$。
$$\rho_+ = \frac{\Pi_+ \rho \Pi_+}{1/2} = 2 \Pi_+ \rho \Pi_+ = 2 \left(\frac{I+M}{2}\right) \left( \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} P + \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} g_k P \right) \left(\frac{I+M}{2}\right)$$
由于 $\frac{I+M}{2}$ 会彻底零化掉所有与 $M$ 反对易的项（即 $g_k P$ 这一半），剩下的项与 $M$ 对易，可以直接穿透投影符。化简后极其干净：
$$\rho_+ = \frac{1}{2^n} (I+M) \sum_{P \in \mathcal{S}^+} P = \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} (P + MP)$$
观察这个新的密度矩阵，它恰好是由一个新的稳定子群生成的：这个新群就是原来的对易子群 $\mathcal{S}^+$，加上了强制塞进去的测量算符 $M$。在经典模拟中，这等价于将那个反对易的生成元 $g_k$ 踢掉，换成了 $M$。

如果在上述线路中，哪怕只加入一个非克里福德门（例如 T 门），密度矩阵就会立刻离开 $\rho = \frac{1}{2^n} \prod (I+g_i)$ 这个狭窄的代数流形，变成大量不同稳定子态的线性组合。那时，经典计算机将被迫去跑单次的蒙特卡洛轨迹，或者面临指数级爆炸的张量网络收缩。纯克里福德密度算符的“极度代数冗余性”正是 Gottesman-Knill 定理的奇迹所在。

**四、 非选择性测量（退相干信道）的代数抵消**

当执行测量却无视（丢弃）测量结果时，系统从一个纯态变成了一个统计混合的混态。在量子信息中，这等价于对系统施加了一个非选择性测量（Non-selective measurement），也就是纯粹的退相干信道（Decoherence channel）。在哥特斯曼-基尔定理的框架下，系统变成了混合稳定子态（Mixed Stabilizer State）。

根据量子力学，无视测量结果的演化，等价于对密度算符 $\rho$ 施加如下的完全正迹保（CPTP）映射：
$$\mathcal{E}(\rho) = \Pi_+ \rho \Pi_+ + \Pi_- \rho \Pi_-$$
将投影算符 $\Pi_{\pm} = \frac{I \pm M}{2}$ 代入并展开：
$$\mathcal{E}(\rho) = \left(\frac{I+M}{2}\right) \rho \left(\frac{I+M}{2}\right) + \left(\frac{I-M}{2}\right) \rho \left(\frac{I-M}{2}\right)$$
由于交叉项 $M\rho - \rho M$ 互相抵消，这个信道可以化简为一个极其干净的形式：
$$\mathcal{E}(\rho) = \frac{1}{2} (\rho + M \rho M)$$

现在，把测量时的两种情况代入这个信道：

*   **情况 A：$M$ 与所有生成元对易**
此时 $\rho$ 已经是 $M$ 的本征态（假定处于 $+1$），所以 $M \rho M = M^2 \rho = \rho$。代入信道公式：
$$\mathcal{E}(\rho) = \frac{1}{2} (\rho + \rho) = \rho$$
**物理图像**：因为系统已经处于测量基底的确定态上，再去测量它并丢弃结果，系统不会发生任何改变。生成元列表原封不动。

*   **情况 B：$M$ 与某个生成元 $g_k$ 反对易**
把系统的稳定子群拆成了两半：对易子群 $\mathcal{S}^+$，以及反对易的陪集 $g_k \mathcal{S}^+$。密度矩阵为：
$$\rho = \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} P + \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} g_k P$$
现在来计算 $M \rho M$。因为 $M$ 与 $\mathcal{S}^+$ 里的元素对易（$M P M = P$），而与 $g_k$ 反对易（$M g_k M = -g_k$），所以：
$$M \rho M = \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} P - \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} g_k P$$
将这两者相加求平均 $\mathcal{E}(\rho) = \frac{1}{2} (\rho + M \rho M)$，那一半包含 $g_k$ 的反对易项被精确地抵消得一干二净！
$$\mathcal{E}(\rho) = \frac{1}{2^n} \sum_{P \in \mathcal{S}^+} P$$

**经典模拟器的操作**：最后这个混合密度矩阵 $\mathcal{E}(\rho)$ 是由 $2^{n-1}$ 个相互对易的泡利算符组成的群（即 $\mathcal{S}^+$）完全定义的。模拟器发现要无视这个导致随机性的测量结果时，只需直接把那个反对易的生成元 $g_k$ 从列表中彻底删除，且不添加任何新算符。

此时，系统的稳定子生成元从 $n$ 个变成了 $n-1$ 个。在物理上，生成元数量少于比特数（$r < n$），意味着系统的密度算符不再是纯态（秩为 $1$），而是变成了纯度为 $2^{r-n}$ 的混态。失去一个生成元，就意味着系统在某一个自由度上变成了最大混态（完全退相干）。

**结论**

如果测量后无视结果，只需把与测量算符反对易的那个生成元直接删掉即可。这证明了 Gottesman-Knill 定理的经典模拟算法（如 CHP 算法）可以毫不费力地追踪任何克里福德网络中的退相干、噪音（只需将噪音拆分为泡利通道）以及信息丢失。系统不论是处在纯态还是混态，其密度矩阵永远被限制在用少量泡利生成元就能精确表达的代数流形上。
