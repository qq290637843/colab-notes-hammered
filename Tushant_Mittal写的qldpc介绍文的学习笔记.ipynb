### 什么是链复形 (Chain Complex)？

在同调代数和代数拓扑中，**链复形** (Chain Complex) 是一系列阿贝尔群 (或模) 以及连接这些群的同态（称为边界算子或微分）组成的序列，并且满足一个关键性质：连续两次应用边界算子的结果为零。

形式上，一个链复形 $(C_\bullet, \partial_\bullet)$ 通常写成如下形式：

$$\cdots \xrightarrow{\partial_{n+2}} C_{n+1} \xrightarrow{\partial_{n+1}} C_n \xrightarrow{\partial_n} C_{n-1} \xrightarrow{\partial_{n-1}} \cdots$$

其中：
*   $C_n$ 是阿贝尔群（或某个环上的模，比如向量空间）。
*   $\partial_n: C_n \to C_{n-1}$ 是群同态，称为**边界算子** (boundary operator) 或**微分** (differential)。

**基本性质：**

对于所有的 $n$，必须满足：
$$\partial_n \circ \partial_{n+1} = 0$$

或者简写为 $\partial^2 = 0$。这意味着：
*   $\text{Im}(\partial_{n+1}) \subseteq \text{Ker}(\partial_n)$

即，第 $n+1$ 个边界算子的像（称为**边界**，Boundaries, 记为 $B_n$）包含在第 $n$ 个边界算子的核（称为**闭链**，Cycles, 记为 $Z_n$）中。

基于此，我们可以定义**同调群** (Homology Group)：
$$H_n = Z_n / B_n = \text{Ker}(\partial_n) / \text{Im}(\partial_{n+1})$$

**注记：与范畴论抽象定义的区分**
在这个语境下，我们使用的是具体代数结构（阿贝尔群/向量空间）上的链复形，而非阿贝尔范畴中的纯粹抽象定义。在量子纠错（尤其是同调码）中，我们需要可计算的结构：
*   **对象 $C_n$**：通常是有限域（如 $\mathbb{F}_2$）上的有限维向量空间。
*   **边界算子 $\partial_n$**：通常用**奇偶校验矩阵**表示。此时 $\partial^2 = 0$ 的条件直接对应于 CSS 码成立的矩阵条件 $H_X H_Z^T = 0$。

在量子纠错码（如CSS码和同调码）中，链复形的概念被用来构造和分析稳定子码的逻辑算子和纠错性能。例如，码字可以看作是闭链，而稳定子等价的码字相差一个边界。

### 什么是对偶复形 / 上链复形 (Dual Complex / Cochain Complex)？

在定义了链复形之后，我们可以通过对其取**对偶 (Dual)** 来得到**对偶复形**（在同调代数中通常称为**上链复形 Cochain Complex**）。

如果我们有一个链复形 $(C_\bullet, \partial_\bullet)$，我们可以对每个空间 $C_n$ 取对偶空间 $C^n = \text{Hom}(C_n, F)$ （在量子码语境下，$F$ 通常是 $\mathbb{F}_2$），并对边界算子 $\partial_n$ 取伴随（或转置），得到**上边缘算子** (coboundary operator) 或余微分 $d^n: C^n \to C^{n+1}$。

形式上，上链复形 $(C^\bullet, d^\bullet)$ 的序列方向与链复形相反，指标递增：

$$\cdots \xrightarrow{d^{n-2}} C^{n-1} \xrightarrow{d^{n-1}} C^n \xrightarrow{d^n} C^{n+1} \xrightarrow{d^{n+1}} \cdots$$

其中：
*   $C^n$ 是 $C_n$ 的对偶空间。
*   $d^n: C^n \to C^{n+1}$ 是上边缘算子。

**基本性质：**

与链复形类似，连续两次应用上边缘算子必须为零：
$$d^n \circ d^{n-1} = 0$$

或者简写为 $d^2 = 0$。这意味着：
*   $\text{Im}(d^{n-1}) \subseteq \text{Ker}(d^n)$

类似地，我们可以定义**上同调群** (Cohomology Group)：
$$H^n = \text{Ker}(d^n) / \text{Im}(d^{n-1})$$

**在量子纠错码 (CSS 码) 中的物理与代数对应：**

在量子 LDPC 码和拓扑码的设计中，链复形和它的对偶复形通常是成对出现的，它们完美地描述了 CSS 码的 $X$ 型和 $Z$ 型校验子结构：
1.  **对偶空间与转置**：由于我们工作在有限域（如 $\mathbb{F}_2$）上的向量空间中，$C_n$ 的对偶空间 $C^n$ 可以自然地等同于 $C_n$ 本身。更关键的是，边界算子 $\partial_n$ 的对偶（上边缘算子 $d^{n-1}$）可以直接对应于原矩阵的**转置矩阵** $\partial_n^T$。
2.  **正交性条件**：链复形的条件 $\partial_n \partial_{n+1} = 0$ 在矩阵表示下就是 $M_1 M_2 = 0$。取转置后，上链复形的条件 $d^{n} d^{n-1} = 0$ 对应于 $M_2^T M_1^T = 0$。这正是 CSS 码要求 $X$ 型稳定子和 $Z$ 型稳定子互相正交的条件 ($H_X H_Z^T = 0$)。
3.  **同调与逻辑比特**：在很多构造（如表面码 / Surface Codes）中，链复形的同调群 $H_n$ 对应于一种类型的逻辑算子（比如逻辑 $Z$ 算子），而其对偶复形的上同调群 $H^n$ 则对应于另一种类型的逻辑算子（比如逻辑 $X$ 算子）。

简而言之，如果我们用链复形 $\partial$ 来描述 $Z$ 错误和 $X$ 校验，那么它的对偶复形 $\partial^T$ 就可以用来描述 $X$ 错误和 $Z$ 校验。

### 码距 $d_j(\mathcal{C})$ 与非平凡逻辑算子

您在书中看到的定义，完整写出来通常带有求最小值的符号（$\min$），用来定义链复形中第 $j$ 个位置的**码距 (Code Distance)**：

$$d_j(\mathcal{C}) = \min \{ |x| \mid x \in \text{ker}(\partial_j) \setminus \text{im}(\partial_{j+1}) \}$$

下面我们逐一拆解这个定义中的每个部分及其在量子纠错码中的物理意义：

#### 1. $|x|$ 是什么？
正如您所推测的，在这里 $|x|$ 指的是向量 $x$ 的**汉明重量 (Hamming Weight)**。因为我们通常在 $\mathbb{F}_2$（由 0 和 1 组成的域）上工作，$|x|$ 就是向量中 $1$ 的个数。在物理上，它代表了这个算子（或错误）**作用在多少个物理量子比特上**。

#### 2. $\text{ker}(\partial_j)$：闭链 (Cycles)
*   **代数意义**：被算子 $\partial_j$ 映射为零的向量集合。
*   **物理意义**：在 QEC 中，边界算子相当于校验矩阵（Syndrome measurement）。如果一个错误 $x$ 满足 $\partial_j(x) = 0$，意味着它**不会触发任何校验子报警**。在稳定子码中，这意味着 $x$ 与所有的校验算子对易。

#### 3. $\text{im}(\partial_{j+1})$：边界 (Boundaries)
*   **代数意义**：可以写成 $\partial_{j+1}(y)$ 形式的向量集合。由于 $\partial^2 = 0$，边界必然是闭链（$\text{im} \subseteq \text{ker}$）。
*   **物理意义**：对应于**稳定子群 (Stabilizer Group)** 本身。稳定子虽然作用在量子比特上，但对处于码空间中的量子态没有任何实质改变（等效于恒等操作）。

#### 4. 差集 $\text{ker}(\partial_j) \setminus \text{im}(\partial_{j+1})$
这是理解量子码最核心的一步：
*   我们在所有“不报警的操作”（$\text{ker}$）中，剔除掉那些“实际上什么也没做（或者说是平凡的）的操作”（$\text{im}$）。
*   剩下的集合，就是**非平凡的逻辑算子 (Non-trivial Logical Operators)**！也就是那些既不触发校验，又实实在在地改变了逻辑状态的错误/操作。

#### 总结
因此，$d_j(\mathcal{C})$ 的含义是：在所有能改变逻辑态且不触发报警的错误模式中，**寻找作用量子比特数最少（汉明重量最小）的那一个**。这个最小的作用数量，就是我们熟知的**量子码距 (Distance)**。码距越大，需要发生多少个物理错误才能“悄无声息”地破坏量子信息的门槛就越高。

### 图的谱参数 $\lambda(G)$ 与第二大特征值 $\lambda_2(A)$

在量子 LDPC 码的文献中（特别是在讨论**扩展图 / 膨胀图 Expander Graphs** 时），我们经常会用到图的“谱（Spectral）”性质。这里的“谱”指的就是矩阵的**特征值**。这部分内容属于**谱图理论 (Spectral Graph Theory)**。

让我们一步步来拆解这个定义：

#### 1. 邻接矩阵 $A$ 与它的特征值
假设我们有一个无向图 $G$，它的**邻接矩阵 (Adjacency Matrix)** $A$ 是一个方阵。如果顶点 $i$ 和顶点 $j$ 之间有边相连，那么 $A_{ij} = 1$，否则 $A_{ij} = 0$。

因为是无向图，$A$ 是一个实对称矩阵。线性代数告诉我们，实对称矩阵一定有 $n$ 个实数特征值（$n$ 是图的顶点数）。

#### 2. 什么是 $\lambda_2(A)$？
我们把邻接矩阵 $A$ 的所有 $n$ 个特征值**从大到小进行排序**：
$$\lambda_1 \ge \lambda_2 \ge \lambda_3 \ge \dots \ge \lambda_n$$

*   $\lambda_1$ 也就是**最大特征值**。如果图 $G$ 是一个 $d$-正则图（每个顶点都恰好有 $d$ 条边），那么 $\lambda_1$ 的值就等于 $d$。
*   $\lambda_2$ 就是在这个降序排列中的**第二大特征值**。您书中所写的 $\lambda_2(A)$ 就是指它。

*(注：在一些文献中，$\lambda(G)$ 也可能被定义为除了 $\lambda_1$ 之外，绝对值最大的那个特征值，即 $\lambda(G) = \max_{i \ge 2} |\lambda_i|$。具体采用哪种取决于上下文，但核心思想是一致的。)*

#### 3. 为什么 $\lambda_2$ 这么重要？它代表了什么？
在图论中，**最大特征值 $\lambda_1$ 和第二大特征值 $\lambda_2$ 之间的差值 $(\lambda_1 - \lambda_2)$ 被称为“谱隙” (Spectral Gap)**。这个谱隙的大小直接决定了图的连通性和“扩展”能力：

*   **$\lambda_2$ 越小（即谱隙越大），图的连通性越好**。这意味着信息在图上随机游走时，能非常迅速地散布到整个图中（混合速度快）。
*   如果图具有**稀疏**的特性（每个点的度数 $d$ 较小，适合做 LDPC 码，因为校验矩阵稀疏），同时又具有**极好的连通性**（$\lambda_2$ 远小于 $d$），我们就称这种图为**扩展图 (Expander Graph)**。

#### 4. 在量子 LDPC 码中的意义
为什么搞量子纠错要关心图的特征值？
因为构造高性能的量子 LDPC 码（特别是距离呈线性增长的 good QLDPC codes），其核心组件就是**扩展图**。扩展图保证了：
1.  **稀疏性**（对应 LDPC 里的 Low-Density，每个量子比特只参与少数几个校验，方便物理实现）。
2.  **高码距 / 高容错率**（因为图的连通性极强，局部的小错误无法轻易聚集成逻辑错误，对应的就是 $\lambda_2$ 足够小）。

所以，当书上写 $\lambda(G) = \lambda_2(A)$ 时，它其实是在定义一个用来**衡量这个图作为“纠错骨架”有多强壮的几何参数**。

### 矩阵的 $(\alpha, \beta)$-扩展性质 (Expanding Property)

在利用扩展图构造纠错码（尤其是经典和量子 LDPC 码）时，图的几何/拓扑扩展性会直接转化为其对应的**奇偶校验矩阵的代数扩展性**。这就是您提到的 $(\alpha, \beta)$-扩展定义的来源。

假设我们有一个线性映射（通常对应于校验矩阵或链复形中的边界算子）：
$$H: \mathbb{F}_2^m \to \mathbb{F}_2^n$$

**定义：**
如果对于所有满足 $|x| \le \alpha m$ 的向量 $x \in \mathbb{F}_2^m$，都成立不等式：
$$|Hx| \ge \beta |x|$$
那么我们就称矩阵 $H$（或其对应的二分图）是 **$(\alpha, \beta)$-扩展的**。

#### 参数与物理意义的拆解：

1.  **$x$ 与 $|x|$：**
    在这里，$x$ 通常代表一个**错误模式 (Error pattern)**。$|x|$ 依然是它的汉明重量（即实际发生错误的物理量子比特的数量）。
2.  **$|x| \le \alpha m$ (小权重假设)：**
    $\alpha$ 是一个介于 $0$ 和 $1$ 之间的小常数（通常被称为相对距离的下限）。这个条件限定了我们只关注**“规模不太大”**的错误（错误数量不超过总长度 $m$ 的 $\alpha$ 比例）。对于全图级别的巨大错误，扩展性必然会失效（因为最大界限是固定的），但在纠错的实际操作中，我们只关心能在码距范围内被纠正的小错误。
3.  **$Hx$ 与 $|Hx|$：**
    $Hx$ 就是错误 $x$ 触发的**校验子 (Syndrome)** 或计算出的边界。$|Hx|$ 就是被触发（产生报警）的校验方程的个数。
4.  **核心不等式 $|Hx| \ge \beta |x|$：**
    $\beta$ 被称为扩展系数（通常 $\beta > 0$）。这句话的直观含义是：**任何一个足够小的错误 $x$，都会触发数量与错误大小成正比（比例至少为 $\beta$）的校验子报警。**

#### 为什么这个性质对量子 LDPC 码至关重要？

*   **抗噪与高解码率 (Decodability)**：如果一个复杂的错误只触发了极少数的校验子报警，解码算法（Decoder）就会一头雾水，很容易把它误认为是一个平凡的小错误。而 $(\alpha, \beta)$-扩展性保证了：只要错误多，报警就一定多。这使得局部解码算法（例如 Small-Set Flip 或 Belief Propagation）能够非常高效地寻找并消除错误。
*   **单次发射纠错 (Single-Shot Error Correction)**：在真实的量子硬件中，**测量校验子这个动作本身也是会出错的**。如果一个码具备良好的扩展性质，即便有几个校验子的测量结果发生了翻转（被硬件噪声污染），因为真正的报警数量 $|Hx|$ “足够庞大”，解码器依然可以透过噪声看清本质，直接猜出错误 $x$。这使得我们可以放弃表面码中极其昂贵的“反复多次测量校验子”的做法，极大降低了量子计算的资源开销。

### 什么是“一个码” (What is a Code?)

在深入讨论诸如 Tanner 码这样的复杂结构之前，我们先回到最基本的问题：在纠错码理论（以及代数编码理论）中，到底什么是“一个码”（通常记为 $\mathcal{C}$ 或 $C$）？

通俗地说，**码就是一套人为设定的、带有冗余信息的合法规则**。凡是符合规则的字符串就是安全的“真信息”，不符合的就被判定为“发生了错误”。

#### 严格的代数定义（线性纠错码）

在经典理论中，信息通常被表示为 $0$ 和 $1$ 组成的字符串。所有长度为 $n$ 的字符串构成了一个巨大的向量空间，记作 $\mathbb{F}_2^n$（总共有 $2^n$ 种可能的字符串）。

一个**长度为 $n$ 的线性码 $\mathcal{C}$**，仅仅是这个巨大向量空间 $\mathbb{F}_2^n$ 中的一个**线性子空间 (Linear Subspace)**。

*   **码字 (Codewords)**：属于这个子空间 $\mathcal{C}$ 内的每一个向量，都被称为“码字”。只有码字才是被系统承认的“合法”信息状态。
*   **非合法态 (Errors)**：任何不在这个子空间里的向量，都被认为是在传输或存储过程中因为噪声发生了错误的态。

#### 怎么规定哪些向量属于这个子空间？

这就需要用到我们在前面多次提到的**奇偶校验矩阵 (Parity-Check Matrix) $H$**。

矩阵 $H$ 的每一行代表一个“校验方程”或“安检门”。一个长度为 $n$ 的向量 $x$ 想要成为合法的码字，就必须完美通过所有的校验：

$$H x = 0$$

用同调代数的语言来说，**码 $\mathcal{C}$ 就是校验矩阵 $H$ 的零空间（核，Kernel）**，即 $\mathcal{C} = \text{ker}(H)$。

#### 为 Tanner 码铺垫：从“全局方程”到“局部视角”

传统上，我们把矩阵 $H$ 看作一个整体。但在图论视角（尤其是 LDPC 码）下，我们会把 $H$ 拆解来看：
$H$ 的每一行其实就是一个**局部的约束条件**，它只“监督”少数几个特定的比特。一个字符串之所以能被称为“合法码字”，**充要条件是它满足所有的局部约束条件**。

这种**“只要所有局部都合法，全局就合法”**的思想，正是接下来我们要介绍的 Tanner 码的灵魂核心。

### 什么是 Tanner 码 $T(G, C_0)$？

在纠错码理论中，**Tanner 码**是一种非常优雅的“从局部到全局”的构造方法。当我们看到 $T(G, C_0)$ 时，它指的是利用一个**全局图 $G$** 和一个**局部基码 (Base Code) $C_0$** 共同拼凑出来的一个大型纠错码。

这种构造最早由 Sipser 和 Spielman 在研究扩展码 (Expander Codes) 时发扬光大，它也是近年来构造极高性能**量子 Tanner 码 (Quantum Tanner Codes)** 的经典基础。

#### 构造要素拆解：

1.  **全局图 $G = (V, E)$**：
    *   通常，我们要求 $G$ 是一个 **$d$-正则图 ($d$-regular graph)**，即图中的每一个顶点都恰好连接着 $d$ 条边。
    *   在 Tanner 码的经典视角中，**物理比特被放置在图的边 (Edges, $E$) 上**。因此，这个全局码的码长 $N = |E|$。
    *   图的顶点 $V$ 则扮演着“局部监督者”（校验约束）的角色。

2.  **局部码 $C_0$**：
    *   这是一个非常小的、长度为 $d$ 的经典线性纠错码（通常被称为内码或基码）。

*(注：在高级的 Tanner 码中，$C_0$ 可能本身就是一个小型的汉明码（比如 $d=15$），此时图上的每一个顶点都内嵌了一个迷你的解码器，不仅能报警还能直接纠正局部错误。)*

#### $T(G, C_0)$ 的正式定义：

因此，一个长度为 $|E|$ 的比特串（分配在图 $G$ 的所有边上），想要成为全局 Tanner 码 $T(G, C_0)$ 的一个**合法码字**，必须满足：

> 对于图 $G$ 中的**任意一个顶点 $v \in V$**，如果你只观察与它直接相连的那 $d$ 条边上的比特，这 $d$ 个比特组成的小字符串，**必须是局部码 $C_0$ 的一个合法码字**。

用更物理的话说：图上的每一个顶点都在“审视”它周围的 $d$ 个比特。只有当所有顶点都认为它周围的 $d$ 个比特符合 $C_0$ 的规则时，整个巨大的多比特态才算是一个合法的全局码字。

#### 为什么 Tanner 码如此重要？

1.  **局部性质决定全局性质 (Local-to-Global)**：
    这正是我们前面讨论“扩展图 (Expander Graph)”大显身手的地方。如果 $G$ 是一个连通性极好的扩展图（即第二大特征值 $\lambda_2$ 很小），并且局部码 $C_0$ 具有一定的最小码距，我们就可以通过极其简单的数学不等式证明：**全局码 $T(G, C_0)$ 也必然具有与之成比例的巨大码距**。这就实现了码距随码长线性增长（$d \propto N$）。
2.  **高度并行的局部解码**：
    因为所有的约束都是高度局域的（只涉及一个顶点及其邻居），所以解码算法可以非常高效。所有顶点可以同时并行地检查和纠正自己周围的错误。
3.  **向量化与量子化 (Quantum Tanner Codes)**：
    在最近的量子 LDPC 码突破中，研究人员将这种一维图上的 Tanner 码推广到了二维平方复形 (Square Complexes) 或多维张量积结构上，让 $X$ 校验和 $Z$ 校验分别遵循对应的局部规则，从而构造出了具有线性码率和线性距离的**好量子 LDPC 码 (Good QLDPC Codes)**。

### 二分复形 (Bipartite Complex) 与 单纯复形 (Simplicial Complex)

#### 1. 二分复形 (Bipartite Complex)

**【教材定义直译】**
> **例 1.4.1 (二分复形)**. 设 $H$ 为码 $\mathcal{C}$ 的奇偶校验矩阵。那么，$\mathcal{C} = \mathbb{F}_2^n \xrightarrow{H} \mathbb{F}_2^m$ 是一个 1-链复形，且满足其同调群 $H_1(\mathcal{C}) = \ker(H) = \mathcal{C}$。一般地，对于一个二分图，其邻接矩阵的形式为 $\begin{pmatrix} 0 & A \\ A^T & 0 \end{pmatrix}$，我们可以通过取 $A$ 作为边界映射来构建这个复形。当我们在后文中提到“二分图的复形”时，指的就是这个构造。这个定义在相差一个对偶（即取 $A$ 或是 $A^T$ 作为映射）的意义下是良定义的。

**【概念说明】**
这段话实际上是把我们最熟悉的“校验矩阵”直接包装成了链复形的语言。
*   在二分图（如 Tanner 图）中，顶点分为两类：变量节点（比特）和校验节点（方程）。矩阵 $A$ 就是记录比特属于哪个方程的“连接矩阵”。
*   当我们把 $A$ 看作边界算子时，整个纠错码系统就变成了一个 1-链复形。此时，$A$ 的零空间（核，$\ker(A)$）就是所有不触发报警的态，也就是合法码字本身。这就是为什么说同调群恰好等于码空间 $\mathcal{C}$。

#### 2. 单纯复形 (Simplicial Complex)

**【教材定义直译】**
> **例 1.4.2 (单纯复形)**. 给定一个基础元素空间，记为 $[n]$。其幂集 $2^{[n]}$ 的一个子集族 $X \subseteq 2^{[n]}$ 如果是**向下封闭的**（即如果 $x \in X$，那么对于任意的 $y \subseteq x$，都有 $y \in X$），则 $X$ 定义了一个单纯复形。设 $X_j = \{x \in X \mid |x| = j + 1\}$，即所有大小为 $j+1$ 的子集构成的集合。我们构建一个链复形，令向量空间 $C_j := \mathbb{F}_2[X_j]$，边界映射定义为 $\partial_j(x) = \sum_{e \in x} (x \setminus \{e\})$。它的对偶映射也非常自然：$\partial_j^*(x) = \sum (x \cup \{e\})$，其中求和遍历所有满足 $(x \cup \{e\}) \in X_{j+1}$ 的 $e$。一个图 $G=(V, E)$ 可以自然地看作是一个单纯 1-复形，其中 $X_0 = V, X_1 = E$。我们把由“边-顶点关联矩阵”给出映射的复形 $E \to V$，称为与图 $G$ 关联的图复形。

**【概念说明】**
这里给出了单纯复形的纯组合/代数定义。它剥离了几何形状，直接用“集合”来表示复形：
*   **向下封闭**：意思是如果包含了一个三角形面（大小为3的集合），就必须包含它的三条边（大小为2的子集）和三个顶点（大小为1的子集）。
*   **基底 $X_j$ 与 空间 $C_j$**：$j$-单纯形（比如 $X_1$ 是边）就是所有大小为 $j+1$ 的集合。我们用这些集合元素作为基，张成向量空间 $C_j$。
*   **边界映射 $\partial_j$**：操作上就是“从集合里删掉一个元素”。比如对一条边 $\{v_1, v_2\}$ 求边界，结果就是把它变成顶点 $v_1$ 和 $v_2$ 的和。这对应求几何边界。
*   **对偶映射 $\partial_j^*$**：操作上是“往集合里增加一个元素”。比如对一条边求上边缘，就是去寻找所有包含这条边的“三角形面”。
*   **图作为特例**：图只有顶点（$X_0$）和边（$X_1$）。它的边界映射就是我们熟知的**关联矩阵 (Incidence Matrix)**，它把每条边映射到它两端的顶点上。

### 什么是稳定子码 (Stabilizer Codes)？

在引入 CSS 码之前，我们需要先了解量子纠错的核心框架：**稳定子码**。它是经典线性纠错码在量子领域的完美推广，由 Gottesman 提出。

#### 1. 物理图像与泡利群 (Pauli Group)
量子态极易受到环境噪声的破坏（例如发生比特翻转 $X$ 或相位翻转 $Z$）。为了进行纠错，我们考虑由 $n$ 个物理量子比特组成的系统，作用在它上面的基本算子构成了 $n$-比特泡利群 $\mathcal{P}_n$（由 $I, X, Y, Z$ 的 $n$ 重张量积组成，并带有相位 $\pm 1, \pm i$）。

#### 2. 稳定子群 (Stabilizer Group) 的定义
我们从整个泡利群中挑选出一个子集 $\mathcal{S}$，并要求它满足两个核心条件：
*   **阿贝尔群 (Abelian)**：$\mathcal{S}$ 中的任意两个算子必须**相互对易**（即 $S_i S_j = S_j S_i$）。因为量子力学中，只有相互对易的算子才能同时具有确定的本征态，我们才能对它们进行同时测量。
*   **不包含 $-I$**：群内不能有负的恒等算子 ($-I \notin \mathcal{S}$)，否则会导致系统不存在非平庸的物理态。

满足上述条件的群 $\mathcal{S}$ 就是“稳定子群”。由于群很大，我们通常只通过它的**生成元 (Generators)** 来描述它：$\mathcal{S} = \langle S_1, S_2, \dots, S_{n-k} \rangle$。这里的 $n-k$ 个生成元就相当于经典纠错中的 $n-k$ 个校验方程。

#### 3. 码空间 (Code Space)：合法的量子态
就像经典线性码是校验矩阵的零空间一样，量子稳定子码的**码空间 $\mathcal{C}$** 被定义为所有稳定子生成元的**共同 $+1$ 本征态**构成的子空间。
换句话说，如果一个多比特量子态 $|\psi\rangle$ 是一个合法的“码字”，那么对于稳定子群中的任何一个算子 $S \in \mathcal{S}$，都必须有：
$$S |\psi\rangle = |\psi\rangle$$
此时，我们称量子态 $|\psi\rangle$ 被算子 $S$ **稳定 (stabilized)**。

#### 4. 纠错原理与“安检”
*   **非破坏性测量**：当我们怀疑系统发生了错误时，我们不能直接去测量量子比特本身（这会导致波函数坍缩，直接摧毁我们要保护的量子叠加信息），而是**去集体测量稳定子算子 $S_i$ 的本征值**。
*   **校验子 (Syndrome)**：如果系统没有出错，测量 $S_i$ 的结果必定是 $+1$。如果系统发生了一个错误 $E$，且该错误与某个稳定子反对易（$E S_i = -S_i E$），那么测量 $S_i$ 时，本征值就会翻转为 $-1$。这些测量出的 $\pm 1$ 结果组合起来，就是用于诊断具体错误模式的校验子。

**总结：** 稳定子码通过精心挑选一组互相通融（对易）的泡利算子，建立了一套“量子安检门”。只要不触发任何安检报警（本征值全为 $+1$），系统就依然处于安全的码空间中。而接下来我们要看的 **CSS 码**，正是稳定子码中最著名且最易于用经典码来构造的特例。

### CSS 码 (Calderbank-Shor-Steane Codes)

CSS 码是由 Calderbank, Shor 和 Steane 提出的一类特殊的量子稳定子码。它的核心思想是极其优雅的：**将量子纠错问题拆解为两个独立的经典纠错问题**。

在一般量子态中，错误可以分解为比特翻转（$X$ 错误）和相位翻转（$Z$ 错误）。CSS 码利用两个经典的线性码来分别处理这两种错误。

#### 1. 代数定义与正交性条件

假设我们用两个经典的奇偶校验矩阵 $H_X$ 和 $H_Z$ 来定义一个 CSS 码：
*   **$H_X$**：用于检测 $Z$ 错误（$X$ 型稳定子的生成元）。
*   **$H_Z$**：用于检测 $X$ 错误（$Z$ 型稳定子的生成元）。

根据量子力学的基本原理（稳定子必须互相**对易**），$X$ 型校验和 $Z$ 型校验必须满足极其严格的**正交性条件**：
$$H_X H_Z^T = 0$$
这在经典码空间的语言中，等价于要求定义 $Z$ 校验的经典码 $\mathcal{C}_Z$ 的对偶空间完全包含于定义 $X$ 校验的经典码 $\mathcal{C}_X$ 中（即 $\mathcal{C}_Z^\perp \subseteq \mathcal{C}_X$）。

#### 2. 与链复形 (Chain Complex) 的深刻联系

如果您回想我们在前面关于“链复形”的讨论，就会发现 CSS 码的正交性条件 $H_X H_Z^T = 0$ 与链复形的根本性质 $\partial^2 = 0$ 如出一辙！

事实上，**每一个 CSS 码都等价于一个长度为 3 的链复形 (2-chain complex)**：
$$C_2 \xrightarrow{\partial_2} C_1 \xrightarrow{\partial_1} C_0$$

具体对应关系如下：
*   **空间 $C_1$**：放置物理量子比特的向量空间 $\mathbb{F}_2^n$。
*   **空间 $C_2$**：$Z$ 型稳定子的空间（有几个 $Z$ 校验方程就有几维）。
*   **空间 $C_0$**：$X$ 型稳定子的空间（有几个 $X$ 校验方程就有几维）。
*   **边界映射**：我们将 $\partial_2$ 定义为 $H_Z^T$，将 $\partial_1$ 定义为 $H_X$。

那么，$\partial_1 \circ \partial_2 = H_X H_Z^T = 0$，完美满足链复形的定义。
*   在这个复形中，**同调群 $H_1(C) = \ker(H_X) / \text{im}(H_Z^T)$** 恰好就是该 CSS 码的**逻辑算子空间**！

#### 3. 在量子 LDPC 码中的意义

为什么现今绝大多数好量子 LDPC 码（Good QLDPC codes）都是 CSS 码？
因为构造 QLDPC 码的核心难点在于：**我们既要让 $H_X$ 和 $H_Z$ 极其稀疏（满足 LDPC 条件），又要让它们极其严格地满足 $H_X H_Z^T = 0$。** 纯随机生成的稀疏矩阵几乎不可能满足正交性。因此，研究人员求助于前面提到的**复形（Complexes）**结构——利用高维几何拓扑（如单纯复形的边界算子）天生满足 $\partial^2 = 0$ 的特性，来系统性地生成不仅稀疏而且正交的 $H_X$ 和 $H_Z$ 矩阵。

### 补充：从“物理对易”到“代数正交”的推导细节

为什么在物理上要求“稳定子必须互相**对易**”，到了数学公式里就变成了**正交性条件 $H_X H_Z^T = 0$** 呢？这中间的翻译过程非常巧妙。

#### 1. 物理上的反对易关系 (Anti-commutation)
我们知道，单个量子比特上的泡利算子 $X$ 和 $Z$ 是**反对易**的，也就是说：
$$X Z = -Z X$$
但如果它们作用在**不同**的量子比特上（比如第一个比特用 $X$，第二个比特用 $Z$），它们是**对易**的：
$$X_1 Z_2 = Z_2 X_1$$

#### 2. 多比特泡利算子的“碰撞”
现在考虑两个多比特算子，一个是纯 $X$ 型算子 $S_X$（对应 $H_X$ 矩阵的一行），另一个是纯 $Z$ 型算子 $S_Z$（对应 $H_Z$ 矩阵的一行）。

当我们要交换 $S_X$ 和 $S_Z$ 的顺序时，我们只需要看它们在**哪些特定的物理量子比特上发生了“碰撞”**（即在同一个位置，既有 $X$ 作用，又有 $Z$ 作用）。
*   每发生一次碰撞，交换顺序时就会产生一个负号 ($-1$)。
*   如果没有碰撞，或者在某个位置一个是 $X$（或 $Z$）另一个是 $I$（单位阵），交换时不产生负号。

#### 3. 统计负号的数量（偶数与奇数）
如果我们想让总算子 $S_X$ 和 $S_Z$ 整体上是**对易**的（即 $S_X S_Z = + S_Z S_X$），那么交换它们时产生的总负号必须相抵消。
*   这意味着：它们发生**碰撞的次数必须是偶数**！
*   $(-1)^{\text{偶数}} = +1$ （对易）
*   $(-1)^{\text{奇数}} = -1$ （反对易）

#### 4. 翻译成经典二元域 $\mathbb{F}_2$ 上的几何语言
在经典代码中，我们用 $0$ 和 $1$ 的向量来表示算子：
*   $S_X$ 可以记为向量 $v_x = (x_1, x_2, \dots, x_n)$，其中 $x_i=1$ 表示该位置有 $X$，$0$ 表示没有。
*   $S_Z$ 可以记为向量 $v_z = (z_1, z_2, \dots, z_n)$，其中 $z_i=1$ 表示该位置有 $Z$，$0$ 表示没有。

这两个算子在第 $i$ 个位置“发生碰撞”，就等价于 $x_i = 1$ 且 $z_i = 1$。也就是它们的乘积 $x_i \cdot z_i = 1$。
那么，**总碰撞次数**就是两个向量的点积（内积）：
$$\text{碰撞次数} = \sum_{i=1}^n x_i z_i$$

要让碰撞次数是偶数，在模 2 加法（二元域 $\mathbb{F}_2$）下，这就等价于内积为 $0$：
$$v_x \cdot v_z \equiv 0 \pmod 2$$

#### 5. 矩阵形式
既然 $H_X$ 的每一行代表一个 $X$ 稳定子，$H_Z$ 的每一行代表一个 $Z$ 稳定子。要求所有的 $X$ 稳定子都与所有的 $Z$ 稳定子对易，就等于要求 $H_X$ 的**每一行**与 $H_Z$ 的**每一行**的内积都为 $0$。
用矩阵乘法写出来，正是：
$$H_X H_Z^T = 0$$

**总结：** 物理学中两个算子对易 $\iff$ 它们共有的非平庸作用位点数是偶数 $\iff$ 在 $\mathbb{F}_2$ 域中代表它们的经典向量内积为零 $\iff$ 校验矩阵正交。

### 链复形的张量积 (Tensor Product of Chain Complexes)

在代数拓扑和同调代数中，**张量积**是一种将两个低维结构“乘”起来构造高维结构的标准方法（例如，两条线段的张量积是一个正方形）。在量子纠错码中，这是将两个经典码转化为一个量子码的核心技术。

#### 1. 代数定义

设有两个链复形 $(A_\bullet, \partial^A)$ 和 $(B_\bullet, \partial^B)$。它们的张量积复形 $C_\bullet = (A \otimes B)_\bullet$ 包含两个关键部分的定义：

**空间 (Chain Groups)：**
新复形中第 $n$ 级的空间 $C_n$，是由原始复形中所有“级数之和为 $n$”的空间的张量积的直和构成：
$$C_n = \bigoplus_{i+j=n} (A_i \otimes B_j)$$
*例如，新复形的 2-链 $C_2$ 包含：$(A_2 \otimes B_0) \oplus (A_1 \otimes B_1) \oplus (A_0 \otimes B_2)$。*

**边界算子 (Boundary Operator / Differential)：**
对于元素 $a \in A_i$ 和 $b \in B_j$，张量积复形的边界算子 $\partial^{A \otimes B}$ 遵循某种“分部求导（莱布尼茨）法则”：
$$\partial^{A \otimes B}(a \otimes b) = \partial^A(a) \otimes b + (-1)^i a \otimes \partial^B(b)$$
*(注：这里引入了符号 $(-1)^i$，是为了确保新复形的边界算子依然满足 $\partial^2 = 0$ 的根本性质。您可以通过展开 $\partial^2(a \otimes b)$ 验证，正是这个负号让交叉项相互抵消了。)*

#### 2. 在 $\mathbb{F}_2$ 域上的极大简化

在量子纠错（尤其是 QLDPC 码）的语境下，我们绝大多数时候是在二元域 $\mathbb{F}_2$（模 2 算术）上工作。在 $\mathbb{F}_2$ 中，**$+1$ 和 $-1$ 是相等的**（即 $1 + 1 = 0 \implies 1 = -1$）！

这意味着那个讨厌的符号 $(-1)^i$ 直接消失了！边界算子在 $\mathbb{F}_2$ 上变成了纯粹的加法：
$$\partial(a \otimes b) = \partial^A(a) \otimes b + a \otimes \partial^B(b)$$

如果用**矩阵的克罗内克积 (Kronecker Product, 通常也记为 $\otimes$)** 来表示边界映射矩阵，那么这就变成了：
$$\partial^{A \otimes B} = \partial^A \otimes I_B + I_A \otimes \partial^B$$

#### 3. 量子 LDPC 码中的巅峰应用：超图乘积码 (HGP)

为什么要在 QLDPC 的笔记中重点提这个概念？

因为大名鼎鼎的**超图乘积码 (Hypergraph Product Codes, Tillich & Zémor, 2009)** —— 第一个被证明具有有限码率的好 QLDPC 码类 —— **其本质正是两个经典线性码（可视为两个 1-链复形）的张量积！**

*   假设你有两个经典码 $\mathcal{C}_1$ 和 $\mathcal{C}_2$，对应的校验矩阵（边界算子）分别为 $H_1$ 和 $H_2$。
*   把它们看作长度为 2 的短链复形，对它们求张量积。
*   得到的将会是一个长度为 3 的 2-链复形：$C_2 \xrightarrow{\partial_2} C_1 \xrightarrow{\partial_1} C_0$。
*   因为链复形天然满足 $\partial_1 \circ \partial_2 = 0$，这就**自动满足了 CSS 码所要求的严格正交性条件 $H_X H_Z^T = 0$**。

通过链复形的张量积，我们可以毫不费力地把任何两个经典码“拼”成一个完全合法的量子 CSS 码！现今最先进的量子码（如 Lifted Product Codes, Quantum Tanner Codes）依然是在这一基础思想上进行各种拓扑推广。

### Künneth 定理 (Künneth Theorem) 及其证明

在定义了链复形的张量积之后，最核心的问题是：**新复形的同调群是什么？**

**Künneth 定理** 给出了这个问题的完美解答。它告诉我们，两个链复形张量积的同调群，完全可以由它们各自的同调群的张量积来决定。在一般的环上，Künneth 公式会包含一个叫做 Tor 的“挠（Torsion）”项；但是，在量子纠错的语境下，我们绝大多数时候工作在**有限域（如 $\mathbb{F}_2$）或某个域 (Field) 上的向量空间**中。在域上，所有的模都是自由的，Tor 项直接消失，定理变得极其简洁。

#### 1. 域上的 Künneth 定理表述

**定理：** 假设 $(A_\bullet, \partial^A)$ 和 $(B_\bullet, \partial^B)$ 是域 $F$（如 $\mathbb{F}_2$）上的两个链复形。那么，它们的张量积复形 $(A \otimes B)_\bullet$ 的第 $n$ 阶同调群为：
$$H_n(A \otimes B) \cong \bigoplus_{i+j=n} \big( H_i(A) \otimes H_j(B) \big)$$

这意味着，新复形的同调完全是由旧复形同调的张量积“拼凑”而成的。

#### 2. 定理的证明（基于域上的分解）

在域 $F$ 上证明这个定理，最直观的方法是利用线性代数中向量空间的**完全可约性**（即任何子空间都有直补空间）。

**步骤 1：链复形的结构分解 (Splitting)**
由于我们是在向量空间中，对于链复形 $A_\bullet$，它的闭链空间 $Z_i = \ker(\partial_i^A)$ 和边界空间 $B_i = \text{im}(\partial_{i+1}^A)$ 都是子空间。
我们可以将单个**向量空间** $A_i$ 分解为三个部分的直和：
$$A_i \cong B_i \oplus H_i(A) \oplus C_i$$
*   $B_i$ 是边界（像空间）。
*   $H_i(A)$ 代表同调群（补足 $B_i$ 以形成完整的闭链空间 $Z_i$）。
*   $C_i$ 是一些非闭链元素，边界算子 $\partial_i^A$ 会将 $C_i$ 同构映射到下一级的边界 $B_{i-1}$ 上。

您可能会问：**为什么空间分成了三份，但下面链复形却只分成了两份？**
这是因为当我们把不同层级的 $A_i$ 连起来形成整体的**链复形**时，边界算子 $\partial^A$ 会跨层级地把 $C_i$ 和 $B_{i-1}$ “绑定”在一起（$\partial^A: C_i \xrightarrow{\sim} B_{i-1}$）。
因此，这三个空间碎片在复形层面上自然地组合成了**两个相互独立的子链复形**：
1.  **同调复形 $\mathcal{H}(A)$**：在每一级只取 $H_i(A)$ 这一部分。因为 $H_i(A)$ 是闭链，算子作用后直接为零，所以此子复形的边界算子全为 0。
2.  **精确复形 (Exact Complexes) $\mathcal{E}(A)$**：在每一级取 $B_i \oplus C_i$ 这两部分合并。在这个子复形中，算子将 $C_i$ 完美映射到下一级的 $B_{i-1}$，没有任何“遗漏”的闭链，其同调完全为 0。

因此，整个链复形被完美分解为 $A_\bullet \cong \mathcal{H}(A) \oplus \mathcal{E}(A)$。同理，$B_\bullet \cong \mathcal{H}(B) \oplus \mathcal{E}(B)$。

**步骤 2：展开张量积**
根据张量积对直和的分配律，我们将 $A_\bullet \otimes B_\bullet$ 展开：
$$(A \otimes B)_\bullet \cong \big( \mathcal{H}(A) \oplus \mathcal{E}(A) \big) \otimes \big( \mathcal{H}(B) \oplus \mathcal{E}(B) \big)$$
$$= \big( \mathcal{H}(A) \otimes \mathcal{H}(B) \big) \oplus \big( \mathcal{H}(A) \otimes \mathcal{E}(B) \big) \oplus \big( \mathcal{E}(A) \otimes \mathcal{H}(B) \big) \oplus \big( \mathcal{E}(A) \otimes \mathcal{E}(B) \big)$$

**步骤 3：消灭无同调的交叉项**
这里用到了一个关键引理：**一个精确链复形（同调为0）与任何链复形的张量积，依然是精确复形（同调为0）。**
这意味着，在上述四个部分中，除了第一项 $\mathcal{H}(A) \otimes \mathcal{H}(B)$ 之外，其余三项由于包含了精确复形 $\mathcal{E}$，它们对总的同调群没有任何贡献（同调均为 0）。

**步骤 4：得出结论**
因此，原复形张量积的同调，完全等于 $\mathcal{H}(A) \otimes \mathcal{H}(B)$ 这个复形的同调。
由于 $\mathcal{H}(A)$ 和 $\mathcal{H}(B)$ 的边界算子定义上就是全零矩阵，它们张量积的边界算子也是全零矩阵（$\partial = 0 \otimes I + I \otimes 0 = 0$）。
在一个边界算子全为 0 的复形中，它的同调空间就是它本身！
所以，第 $n$ 级的同调空间直接等于级数和为 $n$ 的张量积项之和：
$$H_n(A \otimes B) = \bigoplus_{i+j=n} \big( H_i(A) \otimes H_j(B) \big)$$
证明完毕。 $\blacksquare$

#### 3. Künneth 定理在量子纠错 (HGP码) 中的威力

在前面的笔记中提到，**超图乘积码 (HGP)** 是两个经典码 $C_1$ 和 $C_2$ 对应复形的张量积。
在量子码中，同调群的维数 $\dim H_1$ 就是**逻辑量子比特的数量 $k$**。

根据 Künneth 定理，HGP 码的 $H_1$ 同调群为：
$$H_1(C_1 \otimes C_2) \cong \big( H_1(C_1) \otimes H_0(C_2) \big) \oplus \big( H_0(C_1) \otimes H_1(C_2) \big)$$

如果我们将经典码 $C_1$ 的参数记为 $[n_1, k_1]$（$k_1 = \dim H_1(C_1)$，同时 $\dim H_0(C_1)$ 对应转置矩阵的零度），Künneth 定理实际上直接给出了 HGP 码逻辑比特数 $k$ 的解析计算公式！它完美证明了：**新量子码的逻辑算子，正是旧经典码逻辑算子相互之间的张量积组合。**

### 补充证明：为什么“精确复形 $\otimes$ 任意复形”依然精确？

在 Künneth 定理的证明（步骤 3）中，我们用到了一个关键引理：**一个精确复形 $\mathcal{E}$（即同调群全为 0 的复形）与任何链复形 $B$ 的张量积 $\mathcal{E} \otimes B$，依然是一个精确复形。**

在域（如 $\mathbb{F}_2$）上的向量空间中，这个性质的证明非常直观，其核心在于将精确复形拆解为最简单的“基本积木”。

#### 1. 精确复形的“基本单位” (Elementary Exact Complexes)
由于我们在域上工作，任何精确复形 $\mathcal{E}$ 都可以被完全分解为一系列极其简单的“短复形”的直和。
这种最简单的子复形仅仅在相邻的两个层级非零，并且它们之间的边界算子是**同构映射 (Isomorphism，即满秩的可逆矩阵)**。
形式上，这种基本单位 $\mathcal{E}$ 看起来像这样：
$$ \cdots \to 0 \to V_1 \xrightarrow{\partial_1^{\mathcal{E}}} V_0 \to 0 \to \cdots $$
这里 $V_1$ 和 $V_0$ 是维数相同的向量空间，边界算子 $\partial_1^{\mathcal{E}}$ 是一个同构映射（同构于恒等映射 $I_V$）。显然，它的核为 0，像为整个 $V_0$，所以同调为 0。

#### 2. 转化为验证基本单位
因为张量积对直和满足分配律，即 $(\mathcal{E}_1 \oplus \mathcal{E}_2) \otimes B \cong (\mathcal{E}_1 \otimes B) \oplus (\mathcal{E}_2 \otimes B)$。
所以，我们只需要证明：**上述这种只有两层的“基本精确复形”与任意复形 $B$ 的张量积是精确的**，结论自然就推广到了所有精确复形。

#### 3. 张量积复形中的元素与严格的边界算子
假设我们将上述基本精确复形 $\mathcal{E}$ 与任意复形 $B$ 作张量积。新复形 $C = \mathcal{E} \otimes B$ 的第 $n$ 级空间由两部分的直和组成：
$$ C_n = (V_1 \otimes B_{n-1}) \oplus (V_0 \otimes B_n) $$
我们取 $C_n$ 中的任意元素 $e = (x, y)$，其中 $x \in V_1 \otimes B_{n-1}$，$y \in V_0 \otimes B_n$。

根据张量积复形的边界算子公式（在 $\mathbb{F}_2$ 上无正负号）：
$$\partial_n^C(e) = (\partial^{\mathcal{E}} \otimes I + I \otimes \partial^B)(x) + (\partial^{\mathcal{E}} \otimes I + I \otimes \partial^B)(y)$$

仔细拆解这两部分在各个子空间上的投影：
对于 $x \in V_1 \otimes B_{n-1}$：
- $(\partial_1^{\mathcal{E}} \otimes I_{B_{n-1}})(x)$ 落在 $V_0 \otimes B_{n-1}$ 空间。
- $(I_{V_1} \otimes \partial_{n-1}^B)(x)$ 落在 $V_1 \otimes B_{n-2}$ 空间。

对于 $y \in V_0 \otimes B_n$：
- $(\partial_0^{\mathcal{E}} \otimes I_{B_n})(y) = 0$，因为 $\mathcal{E}$ 在 $V_0$ 之后全是 0。
- $(I_{V_0} \otimes \partial_n^B)(y)$ 落在 $V_0 \otimes B_{n-1}$ 空间。

我们将结果按照新空间 $C_{n-1} = (V_1 \otimes B_{n-2}) \oplus (V_0 \otimes B_{n-1})$ 的直和分量进行合并：
$$ \partial_n^C(x, y) = \big( (I_{V_1} \otimes \partial_{n-1}^B)(x), \; (\partial_1^{\mathcal{E}} \otimes I_{B_{n-1}})(x) + (I_{V_0} \otimes \partial_n^B)(y) \big) $$

#### 4. 为什么同调必定为 0？
同调群定义为 $H_n(C) = \ker(\partial_n^C) / \text{im}(\partial_{n+1}^C)$。我们要证明它为 0，即证明所有的闭链都是边界。

**第一步：闭链条件 (Cycle Condition)**
假设 $e = (x, y)$ 是一个闭链，即 $\partial_n^C(x, y) = (0, 0)$。由上述边界算子公式，我们得到两个等式：
(1) $(I_{V_1} \otimes \partial_{n-1}^B)(x) = 0$
(2) $(\partial_1^{\mathcal{E}} \otimes I_{B_{n-1}})(x) + (I_{V_0} \otimes \partial_n^B)(y) = 0$

由于在 $\mathbb{F}_2$ 上加法和减法等价，方程 (2) 给出：
$$ (\partial_1^{\mathcal{E}} \otimes I_{B_{n-1}})(x) = (I_{V_0} \otimes \partial_n^B)(y) $$

**第二步：构造高一级边界 (Boundary Construction)**
我们要证明 $(x, y)$ 是算子 $\partial_{n+1}^C$ 的像。
考虑高一级的空间 $C_{n+1} = (V_1 \otimes B_n) \oplus (V_0 \otimes B_{n+1})$。
因为 $\partial_1^{\mathcal{E}}: V_1 \to V_0$ 是一个同构映射，所以算子 $(\partial_1^{\mathcal{E}} \otimes I_{B_n})$ 也是一个从 $V_1 \otimes B_n$ 到 $V_0 \otimes B_n$ 的同构映射。
因此，对于已知的元素 $y \in V_0 \otimes B_n$，必然存在唯一的原像 $w \in V_1 \otimes B_n$，使得：
$$ (\partial_1^{\mathcal{E}} \otimes I_{B_n})(w) = y $$

现在，我们在 $C_{n+1}$ 中选取元素 $e' = (w, 0)$。对它应用整体边界算子 $\partial_{n+1}^C$：
$$ \partial_{n+1}^C(w, 0) = \big( (I_{V_1} \otimes \partial_n^B)(w), \; (\partial_1^{\mathcal{E}} \otimes I_{B_n})(w) \big) $$

我们来比对这两个分量：
- **第二个分量**：根据 $w$ 的定义，直接等于 $y$。
- **第一个分量**：为 $(I_{V_1} \otimes \partial_n^B)(w)$。我们需要证明它等于 $x$。
  我们在该分量两侧作用同构映射 $(\partial_1^{\mathcal{E}} \otimes I_{B_{n-1}})$：
  $$ (\partial_1^{\mathcal{E}} \otimes I_{B_{n-1}}) (I_{V_1} \otimes \partial_n^B)(w) = (\partial_1^{\mathcal{E}} \otimes \partial_n^B)(w) $$
  由于张量积中不同因子的算子可以交换顺序：
  $$ (\partial_1^{\mathcal{E}} \otimes \partial_n^B)(w) = (I_{V_0} \otimes \partial_n^B) (\partial_1^{\mathcal{E}} \otimes I_{B_n})(w) $$
  代入 $w$ 的定义，上式等于 $(I_{V_0} \otimes \partial_n^B)(y)$。
  再根据闭链条件中的方程 (2)，该值恰好等于 $(\partial_1^{\mathcal{E}} \otimes I_{B_{n-1}})(x)$。
  所以：
  $$ (\partial_1^{\mathcal{E}} \otimes I_{B_{n-1}}) (I_{V_1} \otimes \partial_n^B)(w) = (\partial_1^{\mathcal{E}} \otimes I_{B_{n-1}})(x) $$
  因为 $(\partial_1^{\mathcal{E}} \otimes I_{B_{n-1}})$ 是同构映射，去壳后得到 $x = (I_{V_1} \otimes \partial_n^B)(w)$。

至此，我们严格得出：
$$ \partial_{n+1}^C(w, 0) = (x, y) = e $$

这完美证明了对于 $C_n$ 中的**任意**一个闭链 $(x, y)$，它都必然是高一级元素 $(w, 0)$ 的整体边界。没有任何闭链能逃脱成为边界的命运。
因此，$\ker(\partial_n^C) = \text{im}(\partial_{n+1}^C)$，同调群完全为 0！

### 定理：张量积复形的码距下界

在分析量子 LDPC 码（如超图乘积码及其变体）时，我们经常需要计算两个复形张量积后的新码距。以下定理给出了当其中一个复形是 1-链复形时，新复形码距的严格下界。

**【定理表述】**
设 $\mathcal{D}$ 为一个 1-链复形（即仅在 $D_1$ 和 $D_0$ 处非零，形式为 $D_1 \xrightarrow{\partial_1^{\mathcal{D}}} D_0$），$\mathcal{C}$ 为任意链复形。令 $\mathcal{E} = \mathcal{C} \otimes \mathcal{D}$ 为它们的张量积复形。
那么，复形 $\mathcal{E}$ 在第 $k$ 级的码距 $d_k(\mathcal{E})$ 满足：
$$d_k(\mathcal{E}) \geq \min \big( d_k(\mathcal{C}), \, d_{k-1}(\mathcal{C}) d_1(\mathcal{D}) \big)$$
此外，如果映射 $\partial_1^{\mathcal{D}}$ 是**满射 (Surjective)**，那么有更强的下界：
$$d_k(\mathcal{E}) \geq d_{k-1}(\mathcal{C}) d_1(\mathcal{D})$$

---

**【定理证明】**

**1. 空间与元素的分解**
根据张量积复形的定义，新复形 $\mathcal{E}$ 在第 $k$ 级的空间为：
$$E_k = (C_k \otimes D_0) \oplus (C_{k-1} \otimes D_1)$$
我们取 $\mathcal{E}$ 中的一个**非平凡闭链** $e = (x, y)$，其中 $x \in C_k \otimes D_0$，$y \in C_{k-1} \otimes D_1$。
“非平凡闭链”意味着 $e \in \ker(\partial_k^{\mathcal{E}})$ 但 $e \notin \text{im}(\partial_{k+1}^{\mathcal{E}})$。我们需要证明其汉明重量 $|e| = |x| + |y|$ 满足定理中的下界。

**2. 闭链条件 (Cycle Condition)**
因为 $\partial_k^{\mathcal{E}}(e) = 0$，根据张量积边界算子公式（在 $\mathbb{F}_2$ 上）：
$$\partial_k^{\mathcal{E}}(x, y) = \big( (\partial_k^{\mathcal{C}} \otimes I_{D_0})(x) + (I_{C_{k-1}} \otimes \partial_1^{\mathcal{D}})(y), \; (\partial_{k-1}^{\mathcal{C}} \otimes I_{D_1})(y) \big) = (0, 0)$$
由此我们得到两个方程：
(1) $(\partial_{k-1}^{\mathcal{C}} \otimes I_{D_1})(y) = 0$
(2) $(\partial_k^{\mathcal{C}} \otimes I_{D_0})(x) = (I_{C_{k-1}} \otimes \partial_1^{\mathcal{D}})(y)$

**3. 分情况讨论**
为了严格界定重量，我们不能简单地给 $e$ 加上边界（这会破坏极小重量假设），而是需要考虑 $y$ 在复形 $\mathcal{C}$ 上的同调性质。

**情形一：$y$ 在 $\mathcal{C}$ 方向上是平凡的（即 $y$ 是边界）**
这意味着存在 $z_1 \in C_k \otimes D_1$，使得 $y = (\partial_k^{\mathcal{C}} \otimes I_{D_1})(z_1)$。
我们知道 $e=(x, y)$ 在复形 $\mathcal{E}$ 中是全局非平凡的。我们将原元素 $e = (x, y)$ 加上一个全局边界 $B' = \partial_{k+1}^{\mathcal{E}}(0, z_1) = \big( (I_{C_k} \otimes \partial_1^{\mathcal{D}})(z_1), (\partial_k^{\mathcal{C}} \otimes I_{D_1})(z_1) \big) = \big( (I_{C_k} \otimes \partial_1^{\mathcal{D}})(z_1), y \big)$，得到一个新的代表元：
$$e' = e + B' = \big( x + (I_{C_k} \otimes \partial_1^{\mathcal{D}})(z_1), \; 0 \big) = (x', 0)$$
这里记 $x' = x + (I_{C_k} \otimes \partial_1^{\mathcal{D}})(z_1)$。

因为 $e'$ 与 $e$ 仅相差一个整体边界 $B'$，而 $e$ 是非平凡闭链，所以 $e'$ 必然也是全局复形 $\mathcal{E}$ 中的非平凡闭链（即 $e' \notin \operatorname{im}(\partial_{k+1}^{\mathcal{E}})$）。

这对于 $x'$ 施加了什么限制呢？我们可以用反证法来推导：
假设存在某个 $w \in C_{k+1} \otimes D_0$ 以及满足闭链条件 $(\partial_k^{\mathcal{C}} \otimes I_{D_1})(v)=0$ 的 $v \in C_k \otimes D_1$，使得 $x'$ 可以写成整体边界第一分量的形式：
$$x' = (\partial_{k+1}^{\mathcal{C}} \otimes I_{D_0})(w) + (I_{C_k} \otimes \partial_1^{\mathcal{D}})(v)$$
那么，如果我们把 $(w, v)$ 作为高一级的元素，对其求整体边界，会得到：
$$\partial_{k+1}^{\mathcal{E}}(w, v) = \big( (\partial_{k+1}^{\mathcal{C}} \otimes I_{D_0})(w) + (I_{C_k} \otimes \partial_1^{\mathcal{D}})(v), \; (\partial_k^{\mathcal{C}} \otimes I_{D_1})(v) \big)$$
代入我们刚才的假设，第一分量正是 $x'$，而第二分量因为 $v$ 的限制条件恰好为 $0$。因此：
$$\partial_{k+1}^{\mathcal{E}}(w, v) = (x', 0) = e'$$
这意味着 $e'$ 成为了整体复形的一个边界（平凡闭链），这就与“$e'$ 是全局非平凡闭链”这一确定事实直接矛盾！
因此，不论怎么选取，$x'$ 都**绝不可能**被表示为 $(\partial_{k+1}^{\mathcal{C}} \otimes I_{D_0})(w) + (I_{C_k} \otimes \partial_1^{\mathcal{D}})(v)$ 的形式。

**【泛函 $\alpha$ 的存在性构造】**
为了提取出 $x'$ 中的非平凡部分，我们需要在对偶空间 $D_0^*$ 中寻找一个合适的线性泛函 $\alpha$。为什么必然存在满足条件的 $\alpha$ 呢？
考虑 $D_0$ 的子空间 $U = \operatorname{im}(\partial_1^{\mathcal{D}})$。我们可以为 $D_0$ 选取一组适应于 $U$ 的基底：先选取 $U$ 的一组基 $\{u_1, \dots, u_m\}$，然后将其扩充为整个 $D_0$ 的基 $\{u_1, \dots, u_m, b_1, \dots, b_p\}$。
对应地，将 $x' \in C_k \otimes D_0$ 沿着这组基底展开：
$$ x' = \sum_{i=1}^m c_{u,i} \otimes u_i + \sum_{j=1}^p c_{b,j} \otimes b_j $$
由于 $(\partial_k^{\mathcal{C}} \otimes I_{D_0})(x') = 0$（它是闭链的分量，且交叉项已经消除），这意味着所有的系数 $c_{u,i}, c_{b,j} \in C_k$ 本身都必定是复形 $\mathcal{C}$ 中的闭链，即 $c_{u,i}, c_{b,j} \in \ker(\partial_k^{\mathcal{C}})$。

如果在第二部分的展开中，所有的系数 $c_{b,j}$ 全都是 $\mathcal{C}$ 中的边界（即存在 $w_j \in C_{k+1}$ 使得 $c_{b,j} = \partial_{k+1}^{\mathcal{C}}(w_j)$），那么 $x'$ 的第二部分和就可以写成 $(\partial_{k+1}^{\mathcal{C}} \otimes I_{D_0})(w)$，其中 $w = \sum_{j=1}^p w_j \otimes b_j$。
同时，对于第一部分和 $\sum_{i=1}^m c_{u,i} \otimes u_i$，因为 $u_i \in U = \operatorname{im}(\partial_1^{\mathcal{D}})$，必定存在 $v_i \in D_1$ 使得 $\partial_1^{\mathcal{D}}(v_i) = u_i$。因此第一部分可以写成 $(I_{C_k} \otimes \partial_1^{\mathcal{D}})(v)$，其中 $v = \sum_{i=1}^m c_{u,i} \otimes v_i$。**更关键的是，因为前面已经推导出 $c_{u,i} \in \ker(\partial_k^{\mathcal{C}})$，所以 $(\partial_k^{\mathcal{C}} \otimes I_{D_1})(v) = \sum_{i=1}^m \partial_k^{\mathcal{C}}(c_{u,i}) \otimes v_i = 0$**。这完美满足了反证法假设中对 $v$ 必须是 $\mathcal{C} \otimes D_1$ 方向闭链的要求。

这会导致 $x'$ 完全可以被表示为 $(\partial_{k+1}^{\mathcal{C}} \otimes I_{D_0})(w) + (I_{C_k} \otimes \partial_1^{\mathcal{D}})(v)$ 且 $v$ 满足闭链条件，从而彻底落入前面反证法已经排除掉的形式！

因此，必定存在至少一个指标 $j^*$，使得系数 $c_{b,j^*}$ 是 $C_k$ 中的**非平凡闭链**（即它不是 $\partial_{k+1}^{\mathcal{C}}$ 的像）。
现在，我们就可以明确地定义所需的线性泛函 $\alpha: D_0 \to \mathbb{F}_2$：令 $\alpha(b_{j^*}) = 1$，并且对其余所有的基向量（包括所有的 $u_i$ 和其他 $b_j$）全部映射为 0。

这个被显式构造出来的 $\alpha$ 完美满足了以下两个严格条件：
1. 因为 $\alpha(u_i) = 0$ 对所有基成立，所以 $\alpha$ 在整个 $U = \operatorname{im}(\partial_1^{\mathcal{D}})$ 上恒为 0。
2. 将算子 $(I_{C_k} \otimes \alpha)$ 作用于 $x'$ 时，由于只有 $b_{j^*}$ 会存活，我们得到复形 $\mathcal{C} \otimes \mathbb{F}_2$ 中的链 $\tilde{x} = (I_{C_k} \otimes \alpha)(x') = c_{b,j^*} \otimes 1$。因为 $c_{b,j^*}$ 是非平凡闭链，$\tilde{x}$ 在复形 $\mathcal{C} \otimes \mathbb{F}_2$ 中必然也是一个非平凡的闭链（即 $\tilde{x} \in \ker(\partial_k^{\mathcal{C}} \otimes I_{\mathbb{F}_2})$ 但 $\tilde{x} \notin \operatorname{im}(\partial_{k+1}^{\mathcal{C}} \otimes I_{\mathbb{F}_2})$）。

注意，$\tilde{x}$ 也可以直接由原变量 $x$ 给出，因为：
$$\tilde{x} = (I_{C_k} \otimes \alpha)(x') = (I_{C_k} \otimes \alpha)\big(x + (I_{C_k} \otimes \partial_1^{\mathcal{D}})(z_1)\big) = (I_{C_k} \otimes \alpha)(x) + 0 = (I_{C_k} \otimes \alpha)(x)$$
（上式的后半部分等于 0 是因为 $\alpha$ 作用在 $\operatorname{im}(\partial_1^{\mathcal{D}})$ 上恒为 0）。
并且根据前面的闭链条件方程(2) $(\partial_k^{\mathcal{C}} \otimes I_{D_0})(x) = (I_{C_{k-1}} \otimes \partial_1^{\mathcal{D}})(y)$，我们可以严格验证 $\tilde{x}$ 确实是 $\mathcal{C} \otimes \mathbb{F}_2$ 上的闭链：
$$(\partial_k^{\mathcal{C}} \otimes I_{\mathbb{F}_2})(\tilde{x}) = (\partial_k^{\mathcal{C}} \otimes I_{\mathbb{F}_2}) (I_{C_k} \otimes \alpha)(x) = (I_{C_{k-1}} \otimes \alpha)(\partial_k^{\mathcal{C}} \otimes I_{D_0})(x) = (I_{C_{k-1}} \otimes \alpha)(I_{C_{k-1}} \otimes \partial_1^{\mathcal{D}})(y) = \big( I_{C_{k-1}} \otimes (\alpha \circ \partial_1^{\mathcal{D}}) \big)(y) = 0$$

因为 $\tilde{x}$ 是 $C_k \otimes \mathbb{F}_2$ 中的非平凡闭链，它必须满足 $\mathcal{C} \otimes \mathbb{F}_2$ 的极小码距下界：$|\tilde{x}| \geq d_k(\mathcal{C} \otimes \mathbb{F}_2) = d_k(\mathcal{C})$。
同时，泛函 $(I_{C_k} \otimes \alpha)$ 相当于提取 $x$ 在 $D_0$ 的特定基底上的分量并求和（系数为 0 或 1）。（为了更直观地理解这一点，我们可以将 $C_k \otimes D_0$ 空间中的元素想象成一个二维的 0-1 矩阵，其中行对应 $D_0$ 的基底，列对应 $C_k$ 的基底。对 $D_0$ 方向作用 $\alpha$，相当于对矩阵的某些行进行求和并压缩成单行。观察该矩阵的任意一列：如果原来全是 0，求和后必然是 0；如果原来有 $\ge 1$ 个 1，求和后的结果在 $\mathbb{F}_2$ 下只能是 0 或 1，必然 $\le$ 原来列中 1 的个数。因此，压缩后向量的总重量绝不会大于原矩阵的总重量）。所以我们有：
$$|x| \geq |(I_{C_k} \otimes \alpha)(x)| = |\tilde{x}|$$

因此，我们得到了严密的重量下界：
$$|e| = |x| + |y| \geq |x| \geq |\tilde{x}| \geq d_k(\mathcal{C})$$

**情形二：$y \notin \operatorname{im}(\partial_k^{\mathcal{C}} \otimes I_{D_1})$**
如果 $y$ 不能写成 $(\partial_k^{\mathcal{C}} \otimes I_{D_1})(z_1)$ 的形式。设 $D_1$ 的一组基底为 $\{v_j\}$，我们可以将 $y \in C_{k-1} \otimes D_1$ 展开为 $y = \sum_j y_j \otimes v_j$，其中 $y_j \in C_{k-1}$。

因为根据闭链条件方程 (1) 有 $(\partial_{k-1}^{\mathcal{C}} \otimes I_{D_1})(y) = 0$，所以 $\sum_j \partial_{k-1}^{\mathcal{C}}(y_j) \otimes v_j = 0$。这表明每一个系数 $y_j$ 都必定满足 $y_j \in \ker(\partial_{k-1}^{\mathcal{C}})$。

根据方程 (2) $(\partial_k^{\mathcal{C}} \otimes I_{D_0})(x) = (I_{C_{k-1}} \otimes \partial_1^{\mathcal{D}})(y)$。将 $y$ 的基底展开代入右侧，我们得到等式：
$$(\partial_k^{\mathcal{C}} \otimes I_{D_0})(x) = \sum_j y_j \otimes \partial_1^{\mathcal{D}}(v_j)$$
由于已知 $y \notin \operatorname{im}(\partial_k^{\mathcal{C}} \otimes I_{D_1})$，这意味着在和式 $y = \sum_j y_j \otimes v_j$ 中，必然存在至少一个 $y_j \notin \operatorname{im}(\partial_k^{\mathcal{C}})$。因此，必然存在某个定义在 $C_{k-1}$ 上的线性泛函 $\phi: C_{k-1} \to \mathbb{F}_2$，满足以下两个性质：
1. 对任意 $c \in \operatorname{im}(\partial_k^{\mathcal{C}})$ 都有 $\phi(c) = 0$。
2. 当算子 $(\phi \otimes I_{D_1})$ 作用于 $y$ 时，对其系数进行收缩后得到的向量 $w_\phi \in \mathbb{F}_2 \otimes D_1$ 非零：
$$w_\phi = (\phi \otimes I_{D_1})(y) = \sum_j \phi(y_j) \otimes v_j \neq 0$$

同时，我们将算子 $(\phi \otimes I_{D_0})$ 作用于上述方程 (2) 推导出的等式的左右两端：
左端为 $(\phi \otimes I_{D_0}) (\partial_k^{\mathcal{C}} \otimes I_{D_0})(x) = (\phi \circ \partial_k^{\mathcal{C}} \otimes I_{D_0})(x)$。由于 $\phi$ 在 $\operatorname{im}(\partial_k^{\mathcal{C}})$ 上恒为 0，左端必然等于 0。
右端为 $(\phi \otimes I_{D_0}) \big( \sum_j y_j \otimes \partial_1^{\mathcal{D}}(v_j) \big) = \sum_j \phi(y_j) \otimes \partial_1^{\mathcal{D}}(v_j)$。利用算子线性性质，右端等于 $(I_{\mathbb{F}_2} \otimes \partial_1^{\mathcal{D}}) \big( \sum_j \phi(y_j) \otimes v_j \big) = (I_{\mathbb{F}_2} \otimes \partial_1^{\mathcal{D}})(w_\phi)$。
等式左右两端对齐，我们得到 $(I_{\mathbb{F}_2} \otimes \partial_1^{\mathcal{D}})(w_\phi) = 0$。

这证明了 $w_\phi \in \ker(I_{\mathbb{F}_2} \otimes \partial_1^{\mathcal{D}})$ 且 $w_\phi \neq 0$。根据复形 $\mathcal{D}$ 的码距定义，$w_\phi$ 的汉明重量至少为 $d_1(\mathcal{D})$。这意味着在和式 $w_\phi = \sum_j \phi(y_j) \otimes v_j$ 中，至少有 $d_1(\mathcal{D})$ 个分量满足 $\phi(y_j) \neq 0$（在 $\mathbb{F}_2$ 中即为 1）。

对于这至少 $d_1(\mathcal{D})$ 个分量，既然 $\phi(y_j) \neq 0$，说明对应的 $y_j \notin \operatorname{im}(\partial_k^{\mathcal{C}})$。再结合前文已经说明的 $y_j \in \ker(\partial_{k-1}^{\mathcal{C}})$，这严格证明了这至少 $d_1(\mathcal{D})$ 个 $y_j$ 都是 $C_{k-1}$ 中的非平凡闭链。因此，它们的重量必然满足 $|y_j| \geq d_{k-1}(\mathcal{C})$。

由于张量积空间中 $y = \sum_j y_j \otimes v_j$ 的总重量等于各正交基底方向上重量的直接求和，我们对 $y$ 的总重量进行放缩：
$$|e| \geq |y| = \sum_j |y_j| \geq d_1(\mathcal{D}) \cdot d_{k-1}(\mathcal{C})$$

综合以上两种情形，对于任意非平凡闭链，其重量至少是两者的最小值，即：
$$d_k(\mathcal{E}) \geq \min \big( d_k(\mathcal{C}), \, d_{k-1}(\mathcal{C}) d_1(\mathcal{D}) \big)$$

**4. 若 $\partial_1^{\mathcal{D}}$ 是满射 (Surjective)**

当映射 $\partial_1^{\mathcal{D}}: D_1 \to D_0$ 是满射时，有 $\operatorname{im}(\partial_1^{\mathcal{D}}) = D_0$。这一条件将彻底排除前文中“情形一”发生的可能性。

回顾**情形一**：假设 $y$ 在 $\mathcal{C}$ 方向上是平凡的，我们通过加上一个整体边界，得到了一个与原同调类等价的非平凡闭链代表元 $e' = (x', 0)$。
因为 $e'$ 是闭链，我们有 $(\partial_k^{\mathcal{C}} \otimes I_{D_0})(x') = 0$。
在满射条件下，由于 $\operatorname{im}(\partial_1^{\mathcal{D}}) = D_0$，算子 $(I_{C_k} \otimes \partial_1^{\mathcal{D}})$ 也是从 $C_k \otimes D_1$ 到 $C_k \otimes D_0$ 的满射。进一步地，利用向量空间的性质，对于任何满足闭链条件 $(\partial_k^{\mathcal{C}} \otimes I_{D_0})(x') = 0$ 的 $x'$，我们总可以在对应的闭链空间中找到原像，即存在 $c \in \ker(\partial_k^{\mathcal{C}}) \otimes D_1$，使得：
$$x' = (I_{C_k} \otimes \partial_1^{\mathcal{D}})(c)$$

现在，我们把元素 $(0, c)$ 作为高一级的链放入整体复形 $\mathcal{E}$ 中求边界。由于 $c$ 在 $\mathcal{C}$ 方向上是闭链，即 $(\partial_k^{\mathcal{C}} \otimes I_{D_1})(c) = 0$，我们得到：
$$\partial_{k+1}^{\mathcal{E}}(0, c) = \big( (I_{C_k} \otimes \partial_1^{\mathcal{D}})(c), \; (\partial_k^{\mathcal{C}} \otimes I_{D_1})(c) \big) = (x', 0) = e'$$

这证明了代表元 $e' = (x', 0)$ 必然是一个完全的整体边界（即平凡同调类）。但这与我们最初“$e'$ 是非平凡闭链”的假设产生了直接的逻辑矛盾！

因此，当 $\partial_1^{\mathcal{D}}$ 为满射时，**情形一永远不可能发生**。任何非平凡的闭链都必定落入**情形二**（即 $y$ 在 $\mathcal{C}$ 方向上非平凡）。从而我们可以直接采纳情形二的重量约束，得到更强的下界：
$$d_k(\mathcal{E}) \geq d_{k-1}(\mathcal{C}) d_1(\mathcal{D})$$
证明完毕。 $\blacksquare$

### 解释：利用定理计算张量积 CSS 码的 $X$ 和 $Z$ 码距

您提供的这段英文是对前文“张量积复形的码距下界”定理物理应用的极佳总结。这段话将纯粹的同调代数定理，直接连接到了量子 CSS 码的两个核心物理指标：$X$ 码距（$d_X$）和 $Z$ 码距（$d_Z$）。

让我们逐句拆解这段话的深刻含义：

#### 1. CSS 码与复形的对应
> "Say the code is given by a short chain $E_2 \to E_1 \to E_0$, $d_X = d_1(\mathcal{E})$ and $d_Z = d_1(\mathcal{E}^*)$."

*   我们在前面的笔记中已经知道，任何一个 CSS 码都可以表示为一个长度为 3 的短链复形 $\mathcal{E}$。
*   **$X$ 码距 ($d_X$)**：物理上指能造成逻辑 $X$ 错误且不触发 $Z$ 校验的最小 $X$ 错误重量。这在代数上精确对应于原链复形 $\mathcal{E}$ 在第 1 级的同调距离，即 $d_1(\mathcal{E})$。
*   **$Z$ 码距 ($d_Z$)**：物理上指最小的非平凡 $Z$ 错误重量。既然 $X$ 和 $Z$ 错误在 CSS 码中是对偶的，那么 $Z$ 码距就自然对应于**对偶复形 (Dual Complex)** $\mathcal{E}^*$（或者叫上链复形）的同调距离，即 $d_1(\mathcal{E}^*)$。

#### 2. 张量积的对偶性质
> "We have seen earlier that $\mathcal{E}^* = (\mathcal{C} \otimes \mathcal{D})^* \cong \mathcal{C}^* \otimes \mathcal{D}^*$."

*   这里指出了一个极其优美的代数性质：**张量积的对偶，同构于对偶的张量积**。
*   如果我们的整体复形 $\mathcal{E}$ 是由两个较小的复形 $\mathcal{C}$ 和 $\mathcal{D}$ 做张量积构成的（例如超图乘积码 HGP），那么它的对偶复形 $\mathcal{E}^*$ 根本不需要重新费力计算，它直接等于 $\mathcal{C}^* \otimes \mathcal{D}^*$。

#### 3. “一鱼两吃”：定理的双重应用
> "Thus, all we need to do is to apply this theorem twice, once for $(\mathcal{C}, \mathcal{D})$ and once for $(\mathcal{C}^*, \mathcal{D}^*)$."

*   这是这段话的核心结论。我们刚刚辛辛苦苦证明了一个关于张量积复形码距下界的定理：只要知道分量复形的码距，就能给出它们张量积复形的码距。
*   **第一次应用**：将定理应用于原复形 $(\mathcal{C}, \mathcal{D})$，我们可以立刻得到整体复形 $\mathcal{E}$ 的码距下界，这就直接给出了 **$X$ 码距 $d_X$** 的下界。
*   **第二次应用**：因为对偶复形 $\mathcal{E}^*$ 也是一个张量积（$\mathcal{C}^* \otimes \mathcal{D}^*$），我们可以**将同一个定理原封不动地再用一次**！只需把公式里的参数换成对偶复形 $\mathcal{C}^*$ 和 $\mathcal{D}^*$ 的码距，我们就能立刻得到整体对偶复形的码距下界，这完美地给出了 **$Z$ 码距 $d_Z$** 的下界。

**总结：**
这段话点明了张量积复形方法的强大之处：依靠极度的代数对称性，我们用**同一个定理**，通过代入原复形和对偶复形的参数，就能优雅且严密地同时计算出量子码最难搞定的两个物理指标（$d_X$ 和 $d_Z$）。

### 什么是 `supp` (Support / 支撑集)？

在阅读纠错码、代数拓扑或图论文献时，`supp` 是 **Support（支撑集）** 的缩写。它的基本含义是**一个数学对象中“非零元素”所在位置的集合**。

#### 1. 向量的支撑集 (General Definition)
假设我们有一个定义在二元域 $\mathbb{F}_2$ 上的向量 $x = (x_1, x_2, \dots, x_n)$。
那么 $x$ 的支撑集定义为所有取值不为零（即为 1）的坐标索引构成的集合：
$$\operatorname{supp}(x) = \{i \mid x_i \neq 0\} = \{i \mid x_i = 1\}$$

**联系与区别**：我们之前反复提到的**汉明重量 (Hamming Weight) $|x|$**，其实就是支撑集的大小（元素个数），即 $|x| = |\operatorname{supp}(x)|$。重量是一个数字，而支撑集是一个具体的集合。

#### 2. 在超图/复形中的具体含义：$\operatorname{supp}(\partial_1^*(v))$
您提到在超图 (Hypergraph) 的上下文中看到了 $\operatorname{supp}(\partial_1^*(v))$。结合我们前面关于“上链复形”的知识，我们可以极其精确地拆解它的几何意义：

*   **$v$**：通常代表一个特定的**顶点 (Vertex)**，在代数上是一个基底向量（属于 0-链空间 $C_0$）。
*   **$\partial_1^*$ (上边缘算子)**：我们前面提到过，$
\partial^*$ 的作用是沿着对偶方向“升级”。在图或超图中，算子 $\partial_1^*$ 作用于顶点 $v$ 时，会得到一个在“边空间” ($C_1$) 中的向量。这个向量的特点是：**如果某条边包含了顶点 $v$，那么该边对应的坐标就是 $1$，否则就是 $0$。**
*   **$\operatorname{supp}(\partial_1^*(v))$**：对刚才得到的那个边空间向量取支撑集，实际上就是把所有坐标为 1 的边提取出来，打包成一个集合。

**【几何结论】**
在图或超图中，**$\operatorname{supp}(\partial_1^*(v))$ 严谨的几何意义就是：“与顶点 $v$ 直接相连（关联）的所有边构成的集合”**。这在图论中也常被称为顶点 $v$ 的邻接边集 (Incidence set) 或星形集 (Star)。

**【在 QLDPC 码中的物理意义】**
由于超图经常被用来表示纠错码（顶点和边分别对应比特和方程）：
*   如果 $v$ 代表一个**物理量子比特**，那么 $\operatorname{supp}(\partial_1^*(v))$ 就是**这个量子比特所参与的所有校验方程的集合**。
*   如果这是一个 $d$-正则图，那么对于任意的 $v$，这个集合的大小（即度数）必定等于 $d$，即 $|\operatorname{supp}(\partial_1^*(v))| = d$。

### 什么是超图 (Hypergraph)？

结合您提供的文献段落（来源于 Tillich-Zémor 2014 年关于超图乘积码的经典论文 [TZ14]），我们来详细定义并解释“超图”在这个具体语境下的含义。

#### 1. 一般意义上的超图定义
在普通的图 (Graph) 中，一条“边”只能连接**两个**顶点。但在**超图 (Hypergraph)** 中，限制被打破了：一条“超边 (Hyperedge)”可以像一个“圈”或“集合”一样，同时“框住”或包含**任意数量**的顶点。
*   **普通图**：边 $e = \{v_1, v_2\}$
*   **超图**：超边 $e = \{v_1, v_2, v_3, v_4, \dots\}$

#### 2. 在 1-链复形与经典码中的对应关系
文献指出，考虑一个 1-链复形 $\mathcal{C}$，形式为 $C_1 \xrightarrow{\partial_1} C_0$。在这个复形中构造纠错码 $C$ 时，它的校验矩阵就是 $\partial_1$。此时：
*   **$C_1$ 的基底 (Basis of $C_1$)**：代表所有的**物理比特（变量节点）**。
*   **$C_0$ 的基底 (Basis of $C_0$)**：代表所有的**校验方程（校验节点）**。

#### 3. 将纠错码转化为超图 (The Hypergraph of the Code)
这段话给出了一种极其优美的几何视角：我们如何把一个纯代数的奇偶校验矩阵 $\partial_1$ 画成一个超图？

*   **超图的顶点 (Vertices)**：在这个视角下，超图的基本“顶点”通常指的是 $C_1$ 的基底（即所有的物理比特）。（*注：文献原文中说“vertices are the basis of $C_0, C_1$” 可能是在描述 Tanner 图的二分图结构，即把比特和校验全当成图的顶点；但在纯超图视角下，习惯上以比特为顶点，以校验为超边。我们以超边定义为准。*）
*   **超图的超边 (Hyperedges)**：文献给出了严格的定义——超边是所有形如 $\operatorname{supp}(\partial_1^*(v))$ 的集合。其中 $v$ 是 $C_0$ 的基底（一个校验顶点）。

**【推导与物理图像】**
根据我们上一节对 `supp` 的解释：
1. 我们选中一个校验方程 $v \in C_0$。
2. 应用上边缘算子 $\partial_1^*(v)$，我们在 $C_1$ 空间中得到了一个向量，这个向量在所有参与了校验 $v$ 的比特位置上都是 1。
3. 取支撑集 $\operatorname{supp}(\partial_1^*(v))$，我们得到的就是**一个具体包含了所有参与该校验方程的比特的集合**。

**结论：**
在这个超图中，每一个**物理比特**就是一个**顶点**；而每一个**校验方程**就是一条**超边**，这条超边像一个套索一样，把所有参与该方程计算的物理比特全部“套”在了一起。

超图乘积码 (Hypergraph Product Codes) 正是将两个这样由经典码构成的超图，通过类似于直积的方式“相乘”，从而构造出了天生满足 CSS 正交性条件的二维量子码复形。

### 定理：超图乘积码 (HGP) 的具体参数计算 [TZ14]

在有了“张量积复形码距下界”和“Künneth 定理”这两个强大的代数工具后，我们可以非常轻松地推导出 Tillich 和 Zémor (2014) 关于超图乘积码参数的经典定理。

**【定理 2.1.4 表述】**
设 $\mathcal{A}$ 是一个 1-链复形，其边界映射 $\partial_1^\mathcal{A}$ 由奇偶校验矩阵 $H \in \mathbb{F}_2^{(n-c) \times n}$ 给出，并且假设 $H$ 是**满射 (Surjective)** 的（即矩阵行满秩）。
令 $\mathcal{E} = \mathcal{A} \otimes \mathcal{A}^*$ 为 $\mathcal{A}$ 与其对偶复形 $\mathcal{A}^*$ 的张量积复形。

那么，新构造出的量子超图乘积码 $\mathcal{E}$ 具有以下参数：
1.  **物理量子比特数 (码长)**: $n_1(\mathcal{E}) = n^2 + (n-c)^2$
2.  **逻辑量子比特数 (维数)**: $r_1(\mathcal{E}) = c^2$
3.  **码距**: $X$ 码距和 $Z$ 码距均满足下界 $d_1(\mathcal{E}) \ge d_1(\mathcal{A})$ 以及 $d_1(\mathcal{E}^*) \ge d_1(\mathcal{A})$

*(注：这里 $n$ 是经典码的码长，$n-c$ 是满秩的校验方程数，因此经典码的逻辑维数 / 线性无关的闭链数为 $n - (n-c) = c$。)*

---

**【定理证明】**

**1. 码距 $d_1(\mathcal{E})$ 的证明**
首先，由于复形 $\mathcal{E} = \mathcal{A} \otimes \mathcal{A}^*$，其对偶复形 $\mathcal{E}^* = (\mathcal{A} \otimes \mathcal{A}^*)^* \cong \mathcal{A}^* \otimes \mathcal{A} \cong \mathcal{E}$。这意味着这个量子码是**自对偶的 (Self-dual)**，$X$ 错误和 $Z$ 错误具有完全对称的结构。因此，我们只需要证明 $d_1(\mathcal{E}) \ge d_1(\mathcal{A})$ 即可。

我们分析对偶复形 $\mathcal{A}^*$。在将其视为标准的链复形时（指标反转），它具有形式：$A_0^* \to A_1^* \to 0$。
它的第 0 级同调群（对应于逻辑算子）为：
$$H_0(\mathcal{A}^*) = A_1^* / \operatorname{im}(\partial_1^*)$$
因为经典码存在逻辑比特（即 $c > 0$），这意味着 $\partial_1^*$ 不是满射，$\operatorname{im}(\partial_1^*) \neq A_1^*$。因此，必然存在 $A_1^*$ 中的某个基底向量，它不属于 $\operatorname{im}(\partial_1^*)$。
这意味着在 $\mathcal{A}^*$ 的第 0 级，存在重量仅为 1 的非平凡同调类代表元（单个基底向量）。因此，对偶复形在第 0 级的码距极其微小：
$$d_0(\mathcal{A}^*) = 1$$

现在，我们直接调用前面刚刚证明过的**“满射条件下的更强下界定理”**（即定理 2.1.3）。
令定理中的 $\mathcal{C} = \mathcal{A}^*$，$\mathcal{D} = \mathcal{A}$。因为已知 $\mathcal{A}$ 的边界算子 $\partial_1^\mathcal{A}$ 是满射，我们直接代入下界公式：
$$d_1(\mathcal{E}) = d_1(\mathcal{A}^* \otimes \mathcal{A}) \ge d_0(\mathcal{A}^*) \cdot d_1(\mathcal{A})$$
因为 $d_0(\mathcal{A}^*) = 1$，我们立刻得到：
$$d_1(\mathcal{E}) \ge 1 \cdot d_1(\mathcal{A}) = d_1(\mathcal{A})$$
这证明了张量积构造后，量子码的码距至少能够保持与原经典码的码距一样大！

**2. 物理比特数 $n_1(\mathcal{E})$ 的证明**
根据张量积复形的空间定义，新复形 $\mathcal{E} = \mathcal{A} \otimes \mathcal{A}^*$ 的第 1 级空间为：
$$E_1 = \big( A_1 \otimes (\mathcal{A}^*)_0 \big) \oplus \big( A_0 \otimes (\mathcal{A}^*)_1 \big)$$
关键在于理解对偶复形 $\mathcal{A}^*$ 的层级指标（作为链复形，指标应当递减）。由于原复形 $\mathcal{A}$ 是 $A_1 \to A_0$，它的对偶复形自然是上链复形 $A_0^* \to A_1^*$。当我们将其视为标准的链复形时，层级发生反转：第 0 级空间其实是 $(\mathcal{A}^*)_0 = A_1^*$，第 1 级空间是 $(\mathcal{A}^*)_1 = A_0^*$。

代入上式，我们得到：
$$E_1 = (A_1 \otimes A_1^*) \oplus (A_0 \otimes A_0^*)$$
其中，$A_1$ 的维数为 $n$，$A_0$ 的维数为 $n-c$。对偶空间的维数等于原空间（$\dim A^* = \dim A$）。
对其求维数（即新码的物理量子比特数 / 码长）：
$$n_1(\mathcal{E}) = \dim(A_1)\dim(A_1^*) + \dim(A_0)\dim(A_0^*) = n \cdot n + (n-c) \cdot (n-c) = n^2 + (n-c)^2$$
这完美符合了定理中关于物理比特数的结论。

**3. 逻辑比特数 $r_1(\mathcal{E})$ 的证明**
根据 Künneth 定理公式，整体复形的逻辑维数（Betti 数）可以通过分量复形计算：
$$r_1(\mathcal{E}) = r_1(\mathcal{A}^*)r_0(\mathcal{A}) + r_0(\mathcal{A}^*)r_1(\mathcal{A})$$

这里巧妙利用了**满射**条件：因为 $\partial_1^\mathcal{A}$ 是满射，所以 $\operatorname{im}(\partial_1^\mathcal{A}) = A_0$。因此 $H_0(\mathcal{A}) = A_0 / \operatorname{im}(\partial_1^\mathcal{A}) = 0$。这意味着 $r_0(\mathcal{A}) = 0$。

这使得 Künneth 公式的第一项直接消失！
$$r_1(\mathcal{E}) = 0 + r_0(\mathcal{A}^*)r_1(\mathcal{A})$$

分别计算剩余的两项：
*   $r_1(\mathcal{A})$：这是原经典码的逻辑维数。因为码长为 $n$，满秩校验方程有 $n-c$ 个，所以 $\dim \ker(\partial_1^\mathcal{A}) = n - (n-c) = c$。即 $r_1(\mathcal{A}) = c$。
*   $r_0(\mathcal{A}^*)$：这是对偶复形的第 0 级同调维数。根据定义，$H_0(\mathcal{A}^*) = (\mathcal{A}^*)_0 / \operatorname{im}(\partial_1^*) = A_1^* / \operatorname{im}(\partial_1^*)$。由于 $\partial_1^\mathcal{A}$ 矩阵为 $H$，$\partial_1^*$ 对应 $H^T$。因为 $H$ 满秩，$H^T$ 的列向量线性无关，其像空间的维数就是 $\operatorname{rank}(H^T) = n-c$。而空间 $A_1^*$ 的总维数为 $n$。因此，商空间的维数为 $n - (n-c) = c$。即 $r_0(\mathcal{A}^*) = c$。

代入公式，我们得到量子码的逻辑比特数：
$$r_1(\mathcal{E}) = c \cdot c = c^2$$
证明完毕。 $\blacksquare$

### 为什么同调群的维数等于逻辑量子比特数？

您的疑问非常深刻。您指出：**“同调群的维度不是指非平凡错误的等价类维度吗？”** —— 这个理解是 **100% 正确** 的！

要理解为什么“非平凡错误的维度”恰好等于“逻辑比特的数量”，我们需要将物理图像与代数结构连接起来。这中间的核心思想是：**在量子纠错中，所谓“未被发现的严重错误”，其实就是“合法的逻辑操作”。**

让我们梳理这条逻辑链条：

#### 1. 错误 vs. 逻辑操作
*   **什么是稳定子（平凡错误）？** $\text{im}(\partial_2)$ 中的元素。它们操作了物理比特，但对码空间中的逻辑态没有任何实质影响（相当于乘了 $+1$）。
*   **什么是非平凡错误？** $\ker(\partial_1) \setminus \text{im}(\partial_2)$ 中的元素（即同调群 $H_1$ 的非零元素）。它们不触发校验子报警（在 $\ker$ 中），却又不是稳定子。这意味着什么？这意味着**它们把一个合法的逻辑态，悄无声息地变成了另一个合法的逻辑态！**
*   **视角的反转**：如果这个改变是自然噪声造成的，我们称之为“逻辑错误”（灾难）；但如果我们是故意施加这个操作的呢？那它就是我们在码空间里进行运算的 **逻辑算子 (Logical Operator)**！

#### 2. 维数代表“独立逻辑算子的个数”
在二元域 $\mathbb{F}_2$ 上，同调群 $H_1$ 的维数（即 Betti 数）如果是 $k$，这意味着在同调群（逻辑算子空间）中，存在 **$k$ 个线性独立的基底**。

在 CSS 码的语境下（例如 $H_1$ 对应逻辑 $X$ 算子）：
*   $\dim(H_1) = k$ 意味着，我们能找到 $k$ 个相互独立的、全局的**逻辑 $\bar{X}$ 算子**：$\bar{X}_1, \bar{X}_2, \dots, \bar{X}_k$。
*   同理，由于庞加莱对偶性（或 CSS 码的正交配对），我们也能在对偶空间（上同调群 $H^1$）中找到 $k$ 个相互独立的、全局的**逻辑 $\bar{Z}$ 算子**：$\bar{Z}_1, \bar{Z}_2, \dots, \bar{Z}_k$。

#### 3. 算子个数与逻辑比特数的对应
在量子力学中，**“一个量子比特”的代数定义是什么？**
一个量子比特，就是一对满足特定对易关系的泡利算子（即 $\bar{X}$ 和 $\bar{Z}$ 反对易，且它们与其他的比特算子对易）。

既然我们的纠错码系统能够提供 $k$ 对相互独立的逻辑操作 $(\bar{X}_1, \bar{Z}_1), \dots, (\bar{X}_k, \bar{Z}_k)$，这就说明在这个巨大的物理比特空间（$n$ 个物理比特）中，我们精确地划分出了一个内部结构，这个结构**完全等效于 $k$ 个不受噪声影响的完美量子比特**。

#### 总结：自由度的计算 (Counting Degrees of Freedom)
我们也可以用纯计算自由度的方式来理解：
*   总共有 $n$ 个物理比特（对应空间 $C_1$，维度 $n$）。
*   我们施加了 $\text{rank}(H_X)$ 个独立的 $X$ 校验，和 $\text{rank}(H_Z)$ 个独立的 $Z$ 校验。
*   在稳定子码中，每一个独立的校验方程都会将系统的希尔伯特空间自由度减半（即消耗 1 个维度的自由度）。
*   剩下的自由度（逻辑比特数 $k$）就是：
    $$k = n - \text{rank}(H_X) - \text{rank}(H_Z)$$
*   而根据同调代数的秩-零化度定理：
    $$\dim H_1 = \dim(\ker H_X) - \dim(\text{im} H_Z^T) = (n - \text{rank}(H_X)) - \text{rank}(H_Z)$$

因此，**同调群的维数 $\dim H_1$ 在数值上精确等于系统剩余的自由度 $k$**。这也就是为什么我们把 $\dim H_1$ 称为这个量子纠错码的“逻辑量子比特数”。

### 推导：为什么 HGP 码的码距是 $O(\sqrt{N})$？

在阅读文献时，您看到的这句话：“Taking $c=\Theta(n)$ yields a constant rate quantum LDPC code with distance growing as $O(\sqrt{N})$” （注意：文献中有时会混用符号，这里的 $\sqrt{N}$ 指的是**量子码总码长**的平方根），它是通过将经典**好码 (Good Codes)** 的参数代入 HGP 公式中推导出来的。

让我们一步步来看这个极其巧妙的换算：

#### 1. 前提：经典的“好码” (Good Classical LDPC Codes)
利用二分扩展图 (Bipartite expanders of constant degree)，我们可以构造出极好的经典 LDPC 码（如 Sipser-Spielman 码或 Gallager 码）。
对于这种长度为 $n$ 的经典“好码”，它的逻辑比特数 $c$ 和码距 $d$ 都可以随着码长 $n$ **线性增长**。即存在常数，使得：
*   经典逻辑比特：$c = \Theta(n)$
*   经典码距：$d = \Theta(n)$

#### 2. 代入 HGP 量子码的参数公式
现在，我们把这种经典“好码”作为基码 $\mathcal{A}$，投入到我们在上一个定理中证明的 HGP 机器 $\mathcal{E} = \mathcal{A} \otimes \mathcal{A}^*$ 中。
我们将量子码的总码长记为 $N_Q$，逻辑比特数记为 $K_Q$，码距记为 $D_Q$。

根据上一个定理的公式：
*   **量子码长**：$N_Q = n^2 + (n-c)^2$。因为 $c = \Theta(n)$，所以 $n-c$ 依然是与 $n$ 同阶的量（或为 0）。所以总物理比特数 $N_Q = \Theta(n^2)$。
*   **量子逻辑比特**：$K_Q = c^2$。因为 $c = \Theta(n)$，所以 $K_Q = \Theta(n^2)$。
*   **量子码距**：$D_Q \ge d$。因为 $d = \Theta(n)$，所以量子码距 $D_Q = \Theta(n)$。

#### 3. 变量代换（见证奇迹的时刻）
现在，我们要撇开经典码的长 $n$，单纯观察这个**量子码自身**的参数（即 $N_Q, K_Q, D_Q$ 之间的关系）：

*   **码率 (Rate)**：
    量子码的码率为 $K_Q / N_Q = \Theta(n^2) / \Theta(n^2) = \text{Constant}$（一个大于 0 的常数）。这就解释了文献前半句的 “constant rate”。
*   **码距与码长的关系**：
    由于 $N_Q = \Theta(n^2)$，两边开根号，我们得到 **$n = \Theta(\sqrt{N_Q})$**。
    将这个关系代入到量子码距公式 $D_Q = \Theta(n)$ 中，立刻得到：
    $$D_Q = \Theta(\sqrt{N_Q})$$

#### 总结
这就是 $\sqrt{N}$ 的来源！
因为我们是用两个一维的经典码（长度为 $n$）通过类似面积的“乘积”方式张成了一个二维的量子码（总面积/码长 $N_Q \approx n^2$）。但是，错误要想在这张二维的网上构成一个不被察觉的“逻辑错误闭环”，它只需要横穿其中一个维度即可（长度为 $n$）。因此，**二维网的总面积是 $N_Q$，而横穿它的最短路径（码距）自然就是 $\sqrt{N_Q}$。**

### 补充：原文没写 $d=\Theta(n)$，这个条件是怎么来的？

在阅读文献时，我们注意到原话是：“Taking $c=\Theta(n)$ yields a constant rate quantum LDPC code with distance growing as $O(\sqrt{n})$. these are achievable by using bipartite expanders of constant degree”。作者似乎只假定了 $c=\Theta(n)$。

其实，$d=\Theta(n)$ 的条件隐藏在原话的后半句 **“using bipartite expanders”**（使用二分扩展图）中。这涉及经典编码理论中的一个“行话”：

在经典纠错码中，基于常数度数扩展图构造的 LDPC 码（如著名的 Sipser-Spielman 码）有一个数学铁律：**只要图的扩展性质足够好，这种图导出的经典码必然是一个“好码 (Good Code)”**。

而代数编码理论中对“经典好码”的严格定义，就是必须**同时**满足：
1.  **常数码率**（即逻辑比特 $c$ 随码长 $n$ 线性增长，$c = \Theta(n)$）
2.  **常数相对码距**（即码距 $d$ 随码长 $n$ 线性增长，**$d = \Theta(n)$**）

所以，当作者说“我们可以用二分扩展图来实现”时，在学者的默认语境里，这句话就等价于声明：“我已经拿到了一族天然同时满足 $c = \Theta(n)$ 和 $d = \Theta(n)$ 的经典基码”。这就是为什么我们可以在后续推导中理直气壮地代入 $d = \Theta(n)$。

### LSV 拉马努金图与 LSV 拉马努金复形 (LSV Ramanujan Graphs and Complexes)

在阅读前沿的量子 LDPC 码（尤其是基于**高维扩展图 HDX** 的高级构造，如量子 Tanner 码）文献时，常常会遇到 **LSV 拉马努金复形** 这一概念。它是经典拉马努金图向高维拓扑结构的巅峰推广。

#### 1. 什么是拉马努金图 (Ramanujan Graph)？
我们在前面讨论过，好的扩展图需要“第二大特征值 $\lambda_2$ 尽可能小”。但在图论中，特征值不可能无限小。
*   **Alon-Boppana 界**：数学上证明，对于任何节点数趋于无穷的 $d$-正则图，其第二大特征值的绝对值不可能明显小于 $2\sqrt{d-1}$。
*   **拉马努金图的定义**：如果一个 $d$-正则图的第二大特征值恰好达到了这个理论极限，即严格满足 $\lambda(G) \le 2\sqrt{d-1}$，我们就称其为**拉马努金图**。
*   **意义**：拉马努金图是**理论上最完美的扩展图**，具有最优的连通性和信息混合速度。它能用最少的边实现最强的全局强健性。

#### 2. 什么是 LSV 拉马努金复形 (LSV Ramanujan Complex)？
随着量子纠错技术的发展（比如 CSS 码需要 $\partial^2 = 0$ 的多层级复形结构），一维的图已经不够用了。我们需要把“最优扩展性”推广到二维甚至更高维的单纯复形 (Simplicial Complexes) 上。
*   **LSV 的诞生**：2005年，三位数学家 Alexander Lubotzky, Howard Samuels 和 Uzi Vishne (简称 **LSV**) 成功构造出了一族高维的单纯复形，因此被命名为 **LSV 拉马努金复形**。
*   **数学本质**：它们是局部域上射影一般线性群 $PGL_d(F)$ 的 Bruhat-Tits 建筑 (Buildings) 的有限商空间。这是一种将代数群论、数论与拓扑几何深度结合的精妙构造。
*   **高维扩展性质 (High-Dimensional Expansion)**：在 LSV 复形中，不仅仅是“顶点和边”构成的 1-骨架（即 LSV 拉马努金图）是极佳的扩展图；更关键的是，它局部看来的任何一个侧面——比如以某个顶点为中心的“链接图 (Link graph)”——也都表现出拉马努金图的最优谱扩展性质 (Spectral Expansion)。

#### 3. 为什么量子 LDPC 码极度依赖 LSV 复形？
如果说超图乘积码 (HGP) 只是简单地把两张普通一维扩展图“正交相乘”，从而得到了码距为 $O(\sqrt{N})$ 的二维代码；那么基于 LSV 拉马努金复形的构造，则是直接利用了高维空间天然交织的致密网络：
*   **打破码距诅咒**：借助于 LSV 复形极强的**局部谱扩展性质 (Local Spectral Expansion)**，研究人员构造出了诸如张量张量复形 (Tensor Tensor Complexes) 和量子 Tanner 码。这使得量子纠错的码距终于打破了开根号的限制，实现了与码长成正比的突破（即 $d = \Theta(N)$ 的好量子 LDPC 码）。
*   **局域高效解码**：由于 LSV 复形的每个局部都是完美的扩展图，局部错误无法轻易隐藏。这保证了诸如小集翻转 (Small-Set Flip) 这种简单的局部解码器，就能高效地纠正任何规模小于码距的物理错误。

### 定理 2.1.5：LSV 复形的同调码距与“偏科”问题 [EKZ20]

在前一节我们介绍了 LSV 拉马努金复形，现在我们来看看基于它构造出的链复形在码距上的具体表现。Evra, Kaufman 和 Zémor 在 2020 年给出了其同调码距的严格界限。

**【定理 2.1.5 表述】**
对于 $d$ 维的 LSV 复形 $\mathcal{X}$，在任意层级 $1 \leqslant j \leqslant d-1$，其原复形和对偶复形的码距满足：
*   **原复形码距**：$d_j(\mathcal{X}) = \Omega(\log n_j(\mathcal{X}))$
*   **对偶复形码距**：$d_j(\mathcal{X}^*) = \Omega(n_j(\mathcal{X}^*))$

此外，对于 $j=1, 2$，其逻辑维数 $r_j(\mathcal{X}) > 0$（保证了非平凡逻辑比特的存在）。

#### 【核心物理解读与后续说明】

这段定理和它后面的文字揭示了直接使用 LSV 复形构造量子码时面临的一个巨大尴尬，以及解决这个尴尬的通用方案：

**1. 极度的“偏科”现象 (Asymmetry)**
从定理的界限可以看出，LSV 复形在两个对偶方向上的表现是极其不对称的：
*   一方面，它的**对偶码距非常完美**，随码长呈**线性增长 ($\Omega(n)$)**。如果将其对应到 $Z$ 错误上，这意味着代码对 $Z$ 噪声有极其强大的抵抗力。
*   另一方面，它的**原码距非常拉胯**，随码长仅仅呈**对数级增长 ($\Omega(\log n)$)**。对应到 $X$ 错误上，这意味着极少量的 $X$ 错误就能摧毁逻辑态。

**2. 构造出的量子码 $Q_r$**
由于 CSS 码的总码距取决于 $X$ 码距和 $Z$ 码距中的**最小值**。当我们利用这种复形在层级 $r=1, 2$ 上直接提取量子码时，得到的量子码 $Q_r$ 的参数大致为：
$$ [[n, 1, O((\log n)^r), O(n)]] $$
这里明确指出了问题：虽然我们有线性级别的 $Z$ 码距，但 **$X$ 码距（乃至总码距）实在是太小了**，仅仅是多项式对数级别。这使得它根本算不上是一个“好”的量子 LDPC 码。

**3. 解决方案：张量积与“平衡” (Balancing Procedure)**
为了弥补这种致命的偏科，文中提出了一个通用的解决框架：**与一个“好经典码”进行张量积操作**。
*   我们手里已经有了一个“偏科”的复形（一侧是 $\log n$，另一侧是 $n$）。
*   我们再拿一个参数均衡的经典好码（具有常数码率和线性码距 $d = \Theta(n)$）。
*   将它们进行**张量积**。根据我们在前面学过的“张量积复形码距下界”定理，张量积能够将较弱的码距与经典好码的强码距结合。通过这种所谓的 **Balancing（平衡）** 操作，我们可以强行把那个可怜的对数级 $X$ 码距“拉抬”上来，最终构造出在 $X$ 和 $Z$ 两个方向上都具有很高容错能力的高性能量子 LDPC 码。这也是后来著名的 **平衡乘积码 (Balanced Product Codes)** 的核心思想起源。

### 定理 2.1.6：利用张量积提升复形的码距 [EKZ20]

这部分定理正式给出了我们在上一节提到的“平衡操作 (Balancing Procedure)”的严格数学保证：通过将一个可能“偏科”的复形与一个经典码作张量积，我们可以有效提升其弱侧的码距。

**【定理 2.1.6 表述】**
设 $\mathcal{X}$ 是一个 2-链复形，$\mathcal{C}$ 是一个具有**满射边界映射 (surjective boundary map)** 的 1-链复形。
令 $\mathcal{E} = \mathcal{X} \otimes \mathcal{C}$ 为它们的张量积复形。
那么，张量积复形 $\mathcal{E}$ 的码距满足如下下界：
1.  **原复形方向 (对应 $X$ 码距)**: $d_2(\mathcal{E}) \geqslant d_1(\mathcal{X})d_1(\mathcal{C})$
2.  **对偶复形方向 (对应 $Z$ 码距)**: $d_1(\mathcal{E}^{\ast}) \geqslant d_1(\mathcal{X}^{\ast})$

---

**【定理证明】**

这一定理的证明极其巧妙，它完全建立在我们之前推导过的“张量积复形码距下界”（即定理 2.1.3）的“一鱼两吃”式应用上。

**第一部分：证明** $d_2(\mathcal{E}) \geqslant d_1(\mathcal{X})d_1(\mathcal{C})$

我们直接对原张量积复形 $\mathcal{E} = \mathcal{X} \otimes \mathcal{C}$ 应用定理 2.1.3，考虑计算第 2 级的码距 $d_2(\mathcal{E})$。

因为已知 1-链复形 $\mathcal{C}$ 的边界映射是**满射**的，我们可以直接触发定理 2.1.3 中的**“更强下界”**条件。

直接代入满射条件下的下界公式 $d_k(\text{整体}) \ge d_{k-1}(\text{复形1}) d_1(\text{复形2})$：
$$d_2(\mathcal{E}) = d_2(\mathcal{X} \otimes \mathcal{C}) \geqslant d_{2-1}(\mathcal{X}) \cdot d_1(\mathcal{C}) = d_1(\mathcal{X})d_1(\mathcal{C})$$

第一部分得证。

**第二部分：证明** $d_1(\mathcal{E}^{\ast}) \geqslant d_1(\mathcal{X}^{\ast})$

我们对对偶复形 $\mathcal{E}^{\ast} = (\mathcal{X} \otimes \mathcal{C})^{\ast} \cong \mathcal{X}^{\ast} \otimes \mathcal{C}^{\ast}$ 再次应用定理 2.1.3，计算第 1 级的码距。

在这个方向上，我们不能保证 $\mathcal{C}^{\ast}$ 也是满射的，所以必须使用包含 $\min$ 的通用下界公式：
$$d_1(\mathcal{E}^{\ast}) \geqslant \min \big( d_1(\mathcal{C}^{\ast}), \, d_0(\mathcal{C}^{\ast})d_1(\mathcal{X}^{\ast}) \big)$$

现在我们需要确定 $\mathcal{C}^{\ast}$ 的参数：

1. $\mathcal{C}$ 是 1-链复形，边界映射 $\partial_1: C_1 \to C_0$ 是满射。它的对偶复形 $\mathcal{C}^{\ast}$ 的映射方向为 $C_0^{\ast} \xrightarrow{\partial_1^{\ast}} C_1^{\ast}$。当我们把它作为“第2个分量复形”套用进标准 1-链复形 $D_1 \to D_0$ 的公式时，定理公式中的 $d_1$ 级对应的是起点 **$C_0^{\ast}$** 的位置！在这个位置，我们要找的是映射 $\partial_1^{\ast}$ 的核 $\ker(\partial_1^{\ast})$。因为原映射 $\partial_1$ 是满射（矩阵行满秩），它的对偶映射 $\partial_1^{\ast}$（即转置矩阵）必定是**单射 (Injective，列满秩)**。单射意味着它的核里只有 0（没有任何非平凡的闭链）。按照约定，当不存在非平凡错误时，码距定义为无穷大，即 **$d_1(\mathcal{C}^{\ast}) = \infty$**。
2. 在第 0 级（对应终点 $C_1^{\ast}$ 的位置），正如我们在证明超图乘积码 (HGP) 参数时所讨论过的，此时映射进入该空间后不再有向下的约束，其同调群包含所有未被 $\operatorname{im}(\partial_1^{\ast})$ 覆盖的基向量。因此对偶复形在最末端的极小码距为 **$d_0(\mathcal{C}^{\ast}) = 1$**。

将这两个值代入 $\min$ 公式中：
$$d_1(\mathcal{E}^{\ast}) \geqslant \min \big( \infty, \, 1 \cdot d_1(\mathcal{X}^{\ast}) \big) = d_1(\mathcal{X}^{\ast})$$

第二部分得证。 $\blacksquare$

---

#### 【核心物理解读：平衡操作的成功】
结合上一节关于 LSV 复形的“偏科”问题（定理 2.1.5），这个定理展现了魔法般的疗效：
假设 $\mathcal{X}$ 是那个偏科的 LSV 复形，它的 $d_1(\mathcal{X}^{\ast})$ 很大（线性级 $\Omega(n)$），但是 $d_1(\mathcal{X})$ 极小（对数级 $\Omega(\log n)$）。

现在，我们把它与一个经典的**好码 $\mathcal{C}$** 做张量积。好码 $\mathcal{C}$ 具有很大的码距 $d_1(\mathcal{C}) = \Theta(n)$。
根据定理 2.1.6 算出的新量子码 $\mathcal{E}$：

**新 $Z$ 码距**：$d_1(\mathcal{E}^{\ast}) \geqslant d_1(\mathcal{X}^{\ast})$。原本巨大的 $Z$ 码距**被完好无损地保留了下来**。

**新 $X$ 码距**：$d_2(\mathcal{E}) \geqslant d_1(\mathcal{X}) \cdot d_1(\mathcal{C})$。原本拉胯的对数级 $X$ 码距，**被乘上了一个巨大的系数 $\Theta(n)$**，被强行“拉抬”到了非常高的水平！

这就是为什么“张量积 (Tensor Product)”在量子 LDPC 码的构造中是一项无可替代的终极武器。

### 关键释疑：为什么定理算出的下标不一样（$d_2$ 和 $d_1$），但它们能构成 CSS 码？

在阅读定理 2.1.6 时，一个非常敏锐的观察是：我们算出的原复形码距是 **$d_2(\mathcal{E})$**，对偶复形码距却是 **$d_1(\mathcal{E}^*)$**，它们的**下标看起来不一样**。

既然构造 CSS 码要求 $X$ 错误和 $Z$ 错误必须发生在一群同样的物理量子比特上，那这两个看似不同的下标，怎么能用来定义同一个量子码呢？

答案隐藏在**对偶复形的指标反转**之中。

#### 1. 复形的层级与对偶空间的对应

首先，我们来看张量积复形 $\mathcal{E} = \mathcal{X} \otimes \mathcal{C}$ 的总长度。
*   $\mathcal{X}$ 是 2-链复形（最大层级为 2）。
*   $\mathcal{C}$ 是 1-链复形（最大层级为 1）。
*   它们的张量积 $\mathcal{E}$ 是一个 **3-链复形**，其序列为：
    $$E_3 \xrightarrow{\partial_3} E_2 \xrightarrow{\partial_2} E_1 \xrightarrow{\partial_1} E_0$$

现在，我们对其取对偶。上链复形（箭头反向，指标递增）为：
$$E_0^* \xrightarrow{\partial_1^*} E_1^* \xrightarrow{\partial_2^*} E_2^* \xrightarrow{\partial_3^*} E_3^*$$

为了在数学上统一处理，当我们把这个上链复形重新写成一个**标准的链复形 $\mathcal{E}^*$（箭头向左，指标递减）**时，必须对其层级指标进行**反转映射**。对于总长度为 3 的复形，反转公式为：
$$(\mathcal{E}^*)_j = E_{3-j}^*$$

让我们逐一对应：
*   第 0 级：$(\mathcal{E}^*)_0 = E_{3}^*$
*   **第 1 级：$(\mathcal{E}^*)_1 = E_{2}^*$**
*   第 2 级：$(\mathcal{E}^*)_2 = E_{1}^*$
*   第 3 级：$(\mathcal{E}^*)_3 = E_{0}^*$

#### 2. 奇迹般的会师：其实它们都在 $E_2$ 上！

现在我们再回头看定理 2.1.6 给出的两个码距：
1.  **$d_2(\mathcal{E})$**：这是原复形在第 2 级的同调距离。它发生在空间 **$E_2$** 上。
2.  **$d_1(\mathcal{E}^*)$**：这是对偶复形在第 1 级的同调距离。根据上面的反转公式，它发生在空间 $(\mathcal{E}^*)_1 = $ **$E_2^*$** 上！

**结论：虽然下标写的是 2 和 1，但它们完完全全是对着同一个物理层级 —— 即 $E_2$ 空间！**

#### 3. 提取出的量子 CSS 码

既然都发生在 $E_2$ 上，我们就可以直接在这个层级上定义量子码：
*   **物理量子比特**放置在 $E_2$ 的基底上。
*   **$X$ 校验 (检测 $Z$ 错误)**由映射 $\partial_2: E_2 \to E_1$ 给出。其对应的 $X$ 码距就是定理中的 $d_2(\mathcal{E})$，被拉抬到了 $d_1(\mathcal{X})d_1(\mathcal{C})$。
*   **$Z$ 校验 (检测 $X$ 错误)**由对偶映射 $\partial_3^*: E_2^* \to E_3^*$ 给出。其对应的 $Z$ 码距就是定理中的 $d_1(\mathcal{E}^*)$，保留了极好的 $d_1(\mathcal{X}^*)$。

通过张量积，我们完美地把 $\mathcal{X}$ 原本弱势的 $X$ 码距提升了上来，同时保留了它强势的 $Z$ 码距，并且巧妙地利用层级的移位（$1+1=2$ 和 $1+0=1$ 转换后等于 $3-1=2$），让它们在 $E_2$ 层级完美相遇，直接诞生了一个强大的量子 LDPC 码！

### 推论 2.1.7 (距离平衡 / Distance balancing)

结合前面介绍的 Künneth 定理（计算逻辑维数）和定理 2.1.6（计算码距下界），我们可以直接得出一个极具威力的推论。这个推论为我们提供了一个现成的“量子码升级公式”。

**【推论表述】**
设 $Q$ 是一个参数为 $[[n, K, d_x, d_z]]$ 的量子 CSS 码。设 $C$ 是一个参数为 $[m, k, d]$ 的经典码，且其边界映射是**满射 (surjective map)**。
令 $Q' = (Q \otimes C)_3^1$。
那么，新量子码 $Q'$ 的参数为：
$$ [[ O(nm), \; kK, \; d \cdot d_x, \; d_z ]] $$

---

**【推论证明】**

我们将量子码 $Q$ 视为一个 2-链复形 $\mathcal{X} = (X_2 \to X_1 \to X_0)$，物理比特位于 $X_1$（维数为 $n$），逻辑维数 $\dim H_1(\mathcal{X}) = K$，$X$ 码距 $d_1(\mathcal{X}) = d_x$，$Z$ 码距 $d_1(\mathcal{X}^*) = d_z$。
我们将经典码 $C$ 视为一个 1-链复形 $\mathcal{C} = (C_1 \to C_0)$，物理比特位于 $C_1$（维数为 $m$），逻辑维数 $\dim H_1(\mathcal{C}) = k$，经典码距 $d_1(\mathcal{C}) = d$。由于映射满射，$\dim H_0(\mathcal{C}) = 0$。

构造张量积复形 $\mathcal{E} = \mathcal{X} \otimes \mathcal{C}$，并在空间 $E_2$ 上提取新的量子码 $Q'$。

**1. 新物理比特数 (码长) $N' = O(nm)$**
根据张量积空间的定义，新码的物理量子比特位于 $E_2$ 空间：
$$E_2 = (X_2 \otimes C_0) \oplus (X_1 \otimes C_1) \oplus (X_0 \otimes C_2)$$
因为 $\mathcal{C}$ 是 1-链复形，$C_2 = 0$，最后一项消失。所以：
$$\dim(E_2) = \dim(X_2)\dim(C_0) + \dim(X_1)\dim(C_1)$$
已知 $\dim(X_1) = n$ 且 $\dim(C_1) = m$。同时，$\dim(X_2)$ 通常不超过 $n$（校验方程数不可能超过比特数太多，除非有大量冗余），$\dim(C_0) < m$。因此，总维数以 $n \cdot m$ 为主导：
$$N' \leqslant n \cdot m + n \cdot m = O(nm)$$

**2. 新逻辑比特数 $K' = kK$**
新量子码的逻辑算子空间对应于 $H_2(\mathcal{E})$。根据域上的 Künneth 定理：
$$H_2(\mathcal{E}) = \big( H_2(\mathcal{X}) \otimes H_0(\mathcal{C}) \big) \oplus \big( H_1(\mathcal{X}) \otimes H_1(\mathcal{C}) \big) \oplus \big( H_0(\mathcal{X}) \otimes H_2(\mathcal{C}) \big)$$
*   因为 $\mathcal{C}$ 的映射是满射，$H_0(\mathcal{C}) = 0$，第一项直接消失。
*   因为 $C_2 = 0$，$H_2(\mathcal{C}) = 0$，第三项也消失。

因此，只剩下极其干净的中间项：
$$H_2(\mathcal{E}) \cong H_1(\mathcal{X}) \otimes H_1(\mathcal{C})$$
对其取维数，立刻得到：
$$K' = \dim(H_1(\mathcal{X})) \times \dim(H_1(\mathcal{C})) = K \times k = kK$$

**3. 新 $X$ 码距和 $Z$ 码距**
这就是我们刚才在定理 2.1.6 中证明过的核心结论。
由于 $\mathcal{C}$ 满射，根据定理 2.1.6：
*   **新 $X$ 码距**（在 $E_2$ 级）：$d_2(\mathcal{E}) \geqslant d_1(\mathcal{X}) \cdot d_1(\mathcal{C}) = d_x \cdot d$
*   **新 $Z$ 码距**（在 $E_2^*$ 级）：$d_1(\mathcal{E}^*) \geqslant d_1(\mathcal{X}^*) = d_z$

综上，新量子码 $Q'$ 的参数严格满足 $[[ O(nm), \; kK, \; d\cdot d_x, \; d_z ]]$。证明完毕。 $\blacksquare$

---

#### 【物理意义：彻底治愈“偏科”】

回想我们在前面定理 2.1.5 中遇到的尴尬：基于高维扩展图的 LSV 复形提取出的量子码 $Q_r$，其参数存在极度的“偏科”。它的参数大致为 $[[n, 1, O((\log n)^r), O(n)]]$，即 $X$ 码距只有可怜的 $O((\log n)^r)$。

为了在 [EKZ20] 中得到理想的构造，我们正是利用了上述“距离平衡”推论。我们将偏科的 $Q_r$ 与一个经典的扩展图好码 $C$ 进行张量积。
关键的技巧在于**经典码长度 $m$ 的精妙选取**：我们并不让 $m$ 简单地等于 $n$，而是特意选取 $m = n / (\log n)^r$。同时，这个经典扩展图好码的参数为 $[m, O(m), O(m)]$。

将这些参数代入我们的推论公式，见证奇迹的时刻到了！新量子码 $Q'$ 的参数变成了：
*   **新物理比特数 (总码长 $N$)**：$N = nm = n^2 / (\log n)^r$。
*   **新逻辑比特数**：$K' = 1 \cdot O(m) = O(m)$。
*   **新 $Z$ 码距**：依然保留原本的 $O(n)$。
*   **新 $X$ 码距**：原本拉胯的 $d_x = O((\log n)^r)$，现在被乘上了经典码距 $d = O(m) = O(n / (\log n)^r)$。乘积结果刚好抵消：$O((\log n)^r) \cdot O(n / (\log n)^r) = O(n)$！

此时，$X$ 和 $Z$ 方向上的码距在 $O(n)$ 这个量级上实现了**完美平衡**！新量子码的整体参数为：
$$[[ nm, \; O(m), \; O(n), \; O(n) ]]$$

最后，我们将其换算为关于总码长 $N$ 的表达式：
由于 $N = n^2 / (\log n)^r$，可以反推出 $n = \sqrt{N (\log n)^r}$。在多项式对数级别下，$\log n \approx \frac{1}{2} \log N$，因此 $n = \Theta(\sqrt{N(\log N)^r}$。。

通过这种参数极其讲究的“平衡操作”，原本严重偏科的量子码，直接变成了一个总码距严格突破 $O(\sqrt{N})$ 限制（多乘了一个多项式对数因子）的强大平衡量子码！这就是“距离平衡”在现代量子码构造中的核心威力。

### 推论 2.1.8 (维数提升 / Dimension boosting)

在 [PK21] (Panteleev and Kalachev, 2021) 的工作中，定理 2.1.6 及其推论被再次极其巧妙地应用，用来**以牺牲部分码距为代价，大幅提升量子码的码率（逻辑比特数）**。这个技巧基于以下引理：

**【推论表述】**
设 $Q$ 和 $C$ 与前文定义相同：$Q$ 是参数为 $[[n, K, d_x, d_z]]$ 的量子 CSS 码，$C$ 是参数为 $[m, k, d]$ 的经典码（且边界映射满射）。
我们将 $Q$ 与 $C$ 进行张量积，**取其对偶 (Dual)**，然后再与 $C$ 进行一次张量积。记最终提取出的量子码为：
$$ Q' = \big( (Q \otimes C)_3^* \otimes C \big)_3 $$
那么，新量子码 $Q'$ 的参数为：
$$ [[ O(nm^2), \; k^2K, \; d \cdot d_z, \; d \cdot d_x ]] $$

---

**【推论证明】**

这个证明的优美之处在于它不需要重新计算任何复形同调，只需要像搭积木一样**连续调用两次前文的推论 2.1.7**，并在中间穿插一次物理对偶操作：

**步骤 1：第一次张量积 (提升 $X$ 码距)**
令 $Q_1 = (Q \otimes C)_3$。根据推论 2.1.7，我们直接写出 $Q_1$ 的参数：
$$ Q_1 = [[ O(nm), \; kK, \; d \cdot d_x, \; d_z ]] $$

**步骤 2：取对偶 (Dual 操作，翻转 $X$ 和 $Z$)**
对于任何一个 CSS 码，取对偶操作在物理上极其简单：它就是直接把 $X$ 型校验矩阵和 $Z$ 型校验矩阵互换（即 $X \leftrightarrow Z$）。
这会导致逻辑比特数不变、物理码长不变，但是 **$X$ 码距和 $Z$ 码距会完全对调**。因此，对偶码 $Q_1^*$ 的参数为：
$$ Q_1^* = [[ O(nm), \; kK, \; d_z, \; d \cdot d_x ]] $$
注意，此时 $Q_1^*$ 的 $X$ 码距变成了 $d_z$，$Z$ 码距变成了 $d \cdot d_x$。

**步骤 3：第二次张量积 (提升新的 $X$ 码距)**
现在，我们将对偶后的量子码 $Q_1^*$ 再次作为基码，与经典的满射好码 $C$ 结合：令 $Q' = (Q_1^* \otimes C)_3$。
我们把 $Q_1^*$ 的参数当作新的输入，**第二次**套用推论 2.1.7 的公式：
*   **新码长**：$O(nm) \times m = O(nm^2)$。
*   **新逻辑维数**：$kK \times k = k^2K$。
*   **新 $X$ 码距**：等于 $Q_1^*$ 的 $X$ 码距乘以 $d$，即 $d_z \cdot d = d \cdot d_z$。
*   **新 $Z$ 码距**：保留 $Q_1^*$ 的 $Z$ 码距，即 $d \cdot d_x$。

综上所述，最终得到的量子码 $Q'$ 的参数严格为 $[[ O(nm^2), \; k^2K, \; d \cdot d_z, \; d \cdot d_x ]]$。证明完毕。 $\blacksquare$

---

#### 【物理意义：用距离换空间】

如果我们将 $Q$ 视为某个初始的量子码，将 $C$ 视为具有恒定码率（$k = \Theta(m)$）的经典好码。
经过这样一次“张量积 $\to$ 翻转 $\to$ 张量积”的操作后：
*   **逻辑比特数** 经历了一个 $k^2$ 级别的巨大膨胀（从 $K$ 放大到了 $O(m^2) K$）。由于总码长膨胀了 $O(m^2)$ 倍，这意味着系统依然保持甚至提升了整体的**宏观码率 (Rate)**。
*   **码距被拉平**：原本可能是 $d_x$ 和 $d_z$ 的距离，现在都统一乘上了一个常数系数（经典码距 $d$），变成了交叉对等的形式。虽然距离随着码长的增加可能在渐近意义上增长较慢（“at the cost of the distance”），但这使得研究人员能够构造出逻辑比特密度极高的**高码率好量子 LDPC 码 (High-rate Good QLDPC Codes)**。

### 定义 2.2.1：群表示 (Group Representation)

**【教材定义直译】**
群 $H$ 的一个**表示 (Representation)** 是一个二元组 $(\rho, V)$，其中 $V$ 是某个域 $\mathbb{F}$ 上的向量空间，$\rho$ 是一个群同态 $\rho : H \to GL(V)$，这里 $GL(V)$ 是 $V$ 到自身的自同构（即可逆线性映射）构成的乘法群。
换句话说，如果 $V \cong \mathbb{F}^n$（即 $V$ 是 $n$ 维向量空间），那么 $\rho$ 就是把群元素映射为 $n \times n$ 的可逆矩阵（矩阵的元素属于 $\mathbb{F}$）。

在接下来的内容中，我们主要会遇到两种表示：
1.  **自然表示 (Natural representation)**：对于对称群 $\operatorname{Sym}(l)$，其自然表示 $(\rho_{nat}, \mathbb{F}^l)$ 定义为 $\rho_{nat}(\sigma)e_i = e_{\sigma \cdot i}$，其中 $\{e_1, \cdots, e_l\}$ 是向量空间 $V = \mathbb{F}^l$ 的一组基。
2.  **正则表示 (Regular representation)**：对于任意有限群 $H$，令 $R := \mathbb{F}_2[H]$（群代数/群环）。集合 $\{g \mid g \in H\}$ 构成了空间 $R$ 的一组基。**左正则表示**定义为 $\rho_{reg}(h)g = hg$，而相应的**右正则表示**变体定义为 $\rho_{reg}(h)g = gh^{-1}$。

---

**【概念说明与拆解】**

*   **群表示的核心思想**：群通常是抽象的代数结构（比如对称操作、置换规则）。“表示论”的目的，就是给这些抽象的群元素穿上“线性代数的外衣”——也就是把**群元素变成矩阵**，把**群的乘法变成矩阵乘法**。这样我们就可以利用强大的线性代数工具来研究抽象群。
*   **群同态 $\rho$**：这意味着 $\rho(g_1 \cdot g_2) = \rho(g_1) \cdot \rho(g_2)$，矩阵的乘法完美保留了群本身的乘法结构。
*   **自然表示的直观理解**：比如 $\operatorname{Sym}(3)$ 是三个元素的置换群。自然表示就是把空间设为 3 维，基底为 $e_1, e_2, e_3$。一个置换操作 $\sigma$ 作用在向量上，就等同于直接交换这些基底向量的位置（这对应于**置换矩阵 Permutation Matrix**）。
*   **正则表示的强大之处**：正则表示不依赖于具体的几何空间，而是**直接拿群里的所有元素来当作向量空间的基底**。
    *   **空间大小**：如果群 $H$ 有 $|H|$ 个元素，这个表示的向量空间维数就是 $|H|$。
    *   **作用方式**：群元素 $h$ 作用在基底元素 $g$ 上，其实就是群里的乘法 $hg$。因为乘完之后 $hg$ 依然是群里的元素（也就是另一个基底），这相当于 $h$ 把所有的基底进行了一次完美的重新排列（没有遗漏也没有重复）。
    *   **左与右**：左乘就是 $hg$；而在定义右作用时，为了满足群表示的同态结合律法则 $\rho(a)\rho(b) = \rho(ab)$，必须定义为乘以逆元 $gh^{-1}$。
*   **在量子 LDPC 码中的应用**：在构造高级量子码（如 **提升乘积码 Lifted Product Codes** 或量子 Tanner 码）时，我们经常利用群代数 $\mathbb{F}_2[H]$ 的正则表示，将一个小规模的“基础图”通过群的对称性极其优雅地展开（Lift）成一个巨大的、高度对称的复形结构。

### 定义 2.2.2：群 $H$ 在复形上的作用 ($H$-action on a complex)

**【教材定义直译】**
设 $\mathcal{C}$ 是一个链复形，$H$ 是一个有限群。群 $H$ 在复形上的作用由一组同态映射 $\{ \rho_i \}$ 给出，其中 $\rho_i : H \to GL(C_i)$。对于任意元素 $x \in C_i$，我们通常将群作用简写为 $h \cdot x = \rho_i(h)x$。

这个定义的核心条件是：群作用必须与边界映射**相容 (compatible)**，即：对于任意群元素 $g$ 和复形元素 $a$，都必须满足：
$$ g \cdot (\partial a) = \partial(g \cdot a) $$

虽然上述定义适用于任何代数表示，但在后续的讨论中，我们将施加一个严格的限制：所有的 $\rho_i$ 实际上必须映射到**置换矩阵 (permutation matrices)** 的集合中。这使得群 $H$ 只能置换（重新排列）基底元素，而严格禁止了诸如 $h \cdot e_1 = e_2 + e_3$ 这样的线性组合作用。

---

**【概念说明与拆解】**

*   **什么是“作用在整体复形上”？**
    我们知道链复形是由一层层的向量空间 $C_i$ 和连接它们的边界算子 $\partial$ 构成的。群 $H$ 作用在整体复形上，意味着对于复形的**每一层空间** $C_i$，我们都分配了一个相应的群表示 $\rho_i$。群元素可以同时在所有的层级中“移动”向量。

*   **核心条件 $g \cdot (\partial a) = \partial(g \cdot a)$ 的几何意义：**
    这在数学上被称为图表交换 (Commutativity) 或等变性 (Equivariance)。它的直观意思是：**先平移再求边界，等于先求边界再平移**。
    举个几何的例子：假设 $a$ 是一个三角形面，$\partial a$ 是它的三条边。你先让群元素 $g$ 把这个三角形“旋转”到一个新的位置（$g \cdot a$），再求新三角形的边界 $\partial(g \cdot a)$；它的结果必须绝对等同于先求出原三角形的三条边 $\partial a$，然后再把这三条边集体“旋转”过去 $g \cdot (\partial a)$。这个极其严苛的条件保证了**群的对称性操作绝不会撕裂或破坏复形本身的拓扑连通结构**。

*   **为什么必须限制为“置换矩阵”？**
    *   **置换矩阵的特性**：每一行每一列都只有一个 1，其余都是 0。它作用在基底向量（在码的语境下，基底就是具体的物理量子比特或校验方程）上时，效果仅仅是“移形换位”——比如把“第 1 号比特”挪到了“第 2 号比特”的位置，或者把“校验方程 A”映射成了“校验方程 B”。
    *   **拒绝线性叠加**：如果不加这个限制，允许出现线性组合（如 $h \cdot e_1 = e_2 + e_3$），就意味着群作用把一个孤立的物理比特“劈开”成了两个比特的叠加。这在纯代数中是合法的，但在量子 LDPC 码的**几何图构造（如图的提升 Lift / Tanner 图）**中是荒谬且无法物理实现的。我们希望群的作用仅仅是在图（网络）上进行整体的对称性“排列”，绝不能改变图节点的绝对数量或产生分裂。因此，只有纯粹的置换才符合物理直觉。

### 例 2.2.3：在图上的作用 (Action on a graph)

**【教材定义直译】**
在复形 $\mathcal{F}$ 是一个图复形 (graph complex) 的情况下，这个相容性条件（即 $g \cdot (\partial a) = \partial(g \cdot a)$）等价于要求该群作用是一个**图同构 (graph isomorphism)**。
为了更直观地看出这一点，假设 $e = (u,v)$ 是一条连接顶点 $u$ 和 $v$ 的边。对于任意的群元素 $g \in H$，有：
$$g \cdot (\partial e) = g \cdot (u+v) = g \cdot u + g \cdot v = \partial(g \cdot e)$$
因此，必然有 $g \cdot e = (g \cdot u, g \cdot v)$。

---

**【概念说明与拆解】**

*   **图复形的边界 $\partial$**：在图（即 1-维单纯复形）中，一条边 $e$ 的边界就是它的两个端点。在 $\mathbb{F}_2$ 域（模 2 加法）下，我们写成 $\partial e = u + v$。
*   **代数推导对应的几何直觉**：
    *   **等式左边 $g \cdot (\partial e)$**：先求出边 $e$ 的两个端点 $u$ 和 $v$，然后用群元素 $g$ 分别把这两个顶点“平移/变换”到新的位置 $g \cdot u$ 和 $g \cdot v$。
    *   **等式右边 $\partial(g \cdot e)$**：先把整条边 $e$ 整体“平移/变换”成一条新边 $g \cdot e$，然后再去求这条新边的两个端点。
    *   **二者相等意味着什么？** 这意味着被变换后的新边 $g \cdot e$，它的两端连接的恰好就是变换后的新顶点 $g \cdot u$ 和 $g \cdot v$！
*   **为什么这就是“图同构”？**
    图同构的核心要求就是**保持顶点之间的连通关系完全不变**：变换前被边 $e$ 相连的两个点，变换后对应的新点必然被对应的新边相连。上面的代数推导完美地展现了这一点。这表明群 $H$ 在图上的作用仅仅是对图进行对称的“翻转”、“旋转”或“整体重排”，绝不会把原本连着的边扯断，也不会凭空把没连着的点连起来。这种保持拓扑结构不变的优良对称性，是我们后续通过“群提升 (Lift)”构造大型高度对称纠错码的基础。

### 定义 2.2.4：商复形 (Quotient Complex)

**【教材定义直译】**
对于带有群 $H$ 作用的链复形 $\mathcal{C}$，我们定义一个新的链复形 $\mathcal{C}/H$，其每一层的空间定义为：
$$ (\mathcal{C}/H)_i = C_i/H := C_i / \langle v - h \cdot v \mid v \in C_i, h \in H \rangle $$
换句话说，我们正在**将每个向量的轨道 (orbits) 收缩为一个单点**。

观察到，对于任意的 $v \in C_i$：
$$ \partial(v - h \cdot v) = \partial v - \partial(h \cdot v) = \partial v - h \cdot \partial(v) \equiv 0 \in C_{i-1}/H $$
因此，得益于群作用与边界映射的相容性，边界映射在这些商空间上是**良定义 (well-defined)** 的。

---

**【概念说明与拆解】**

*   **什么是“商 (Quotient)”与等价类？**
    在代数中，“商空间”就是强行规定一些原本不同的元素“相等”。
    这里的定义 $C_i / \langle v - h \cdot v \rangle$，意思是我们在原空间 $C_i$ 中模掉（除以）所有形如 $v - h \cdot v$ 的元素。换言之，我们在新空间中强行令 $v - h \cdot v = 0$，即 **$v = h \cdot v$**。
    这意味着：如果在原复形中，元素 $A$ 可以通过群元素 $h$ 的作用变成元素 $B$（即 $B = h \cdot A$），那么在商复形 $\mathcal{C}/H$ 中，**$A$ 和 $B$ 就被视为同一个点**。

*   **几何直觉：轨道的收缩 (Shrinking Orbits)**
    “轨道 (Orbit)”指的是一个基底元素在群 $H$ 的所有可能操作下能到达的所有位置的集合。
    想象一个高度对称的大型几何体（比如一个圆柱体或一个由无数重复单元拼成的巨大网络）。“取商复形”的操作，就相当于把所有因为对称性而长得一模一样的部分**“折叠”或“卷起”**，最后浓缩成一个微小但包含了所有连接规则的“基础单元 (Base Space)”。
    在量子 LDPC 码的“提升 (Lift)”操作中，商复形对应的往往是那个小小的**“基础图 (Base Graph)”**，而原复形 $\mathcal{C}$ 则是基础图展开后的“提升图 (Lifted Graph)”。

*   **边界映射的“良定义性 (Well-definedness)”**
    在数学中，当我们强行把多个不同的点打包成一个等价类时，必须检查一个逻辑漏洞：**原来不同的点，求完边界之后，会不会跑到不同的等价类里去？** 如果那样，新空间的边界映射就自相矛盾了。
    这就是最后那行等式的作用：$\partial(v - h \cdot v) = \partial v - h \cdot \partial(v)$。它利用了上一节强调的核心条件——**群作用与边界算子的交换律 $\partial(h \cdot v) = h \cdot \partial(v)$**。
    这个推导说明，如果在当前层 $A$ 和 $B$ 被视为同一个元素，那么求边界后，$\partial A$ 和 $\partial B$ 依然相差一个群作用，因此在下一层的商空间 $C_{i-1}/H$ 中，它们依然会被视为同一个元素。这就保证了整个商复形的结构是逻辑严密、完美自洽的。

### 定义 2.2.5：自由作用 (Free action)

为了更好地研究商复形，我们通常会将注意力集中在一类性质极好的群作用上，这类作用被称为**自由作用 (Free actions)**。

**【教材定义直译】**
如果对于任何层级 $i$ 以及任何基底元素 $v \in C_i$，当满足 $h \cdot v = v$ 时，必然推出 $h = e$（$e$ 为群的单位元），那么这个群作用就被称为**自由的 (free)**。

---

**【概念说明与拆解】**

*   **“自由 (Free)”的字面含义**：
    在数学中，“自由作用”意味着群 $H$ 中的任何非平凡元素（即除了什么都不干的单位元 $e$ 之外的操作），**绝对不可能把任何一个基底元素固定在原地**。只要你施加了一个真实的动作 $h \neq e$，所有的顶点、边或面都必须移动到新的位置。

*   **为什么排除 $h \cdot v = v$？ (不动点)**
    如果存在 $h \neq e$ 使得 $h \cdot v = v$，那么 $v$ 就被称为这个群作用的**不动点 (Fixed point)**。在几何图或复形上，这意味着你在进行旋转或翻转时，某个顶点刚好在旋转轴或对称面上，导致它自身没有发生位移。

*   **自由作用在量子码（Lifted Codes）中的重要性**：
    1.  **尺寸的严格膨胀**：如果群 $H$ 在基础复形上的作用是自由的，那么每个轨道的长度（即一个元素在群作用下能生成的所有不同元素的数量）就严格等于群的阶数 $|H|$。这意味着，当我们将商复形（基础图）重新展开（Lift）成大复形时，每个基础节点都会完美地分裂成刚好 $|H|$ 个新节点。这使得我们可以极其精确地计算出大复形的物理比特数和校验方程数。
    2.  **避免拓扑奇点**：没有不动点意味着在做“商”操作时，空间不会发生局部的“扭结”或产生几何上的“奇点”。展开后的复形（即覆盖空间 Covering Space）在局部上长得和基础复形一模一样，只是全局变得更大了。这完美契合了纠错码设计中**保持局部连接结构不变（LDPC 性质不变）**的要求。

### 引理 2.2.6：自由作用下的复形分解 (Decomposition of Complex under Free Action)

我们将首先证明，自由作用会退化为一系列正则表示的集合，从而得到商空间 $V/H$ 的结构。然后，我们将观察取商后的边界映射的结构。

**【教材引理直译】**
假设群 $H$ 作用在复形 $\mathcal{C}$ 上，且在每个层级 $C_i$ 上的作用都是**自由的 (free)**。那么对于每一个 $i$：
*   **i)** $C_i \cong C_i/H \otimes \mathbb{F}_2[H]$。
*   **ii)** 群作用分解为 $\rho_i(h) = I \otimes \rho_{reg}(h)$。
*   **iii)** 边界映射 $\partial_i : C_i \to C_{i-1}$ 具有形式 $\partial'_i \otimes \rho_{reg}(s_i)$，其中 $\partial'_i : C_i/H \to C_{i-1}/H$ 是商复形上的边界映射。

---

**【引理证明】**

设 $x$ 是任意基底元素。那么 $h \cdot x = g \cdot x$ 意味着 $(g^{-1}h) \cdot x = x$。因为作用是自由的（没有不动点），我们必然得到 $g^{-1}h = e$，即 $g = h$。因此，$x$ 的轨道（即集合 $\{h \cdot x \mid h \in H\}$）的大小严格等于 $\vert{}H\vert{}$。这也意味着商空间的维数为 $C_i/H$ 的维数 $= \dim C_i / \vert{}H\vert{}$。

设 $b$ 是商空间 $C_i/H$ 的一个基向量。因为 $b$（当被视为 $C_i$ 中的元素时）的轨道大小为 $\vert{}H\vert{}$，我们可以通过群作用得到完全不同（线性无关）的基向量，记为 $b \otimes h := h \cdot b$。

现在来看群元素 $g$ 的作用：
$$ g \cdot (b \otimes h) = g(h \cdot b) = (gh) \cdot b = b \otimes (gh) $$
这说明，该作用在第一个分量（商空间位置）上是恒等映射 (identity)，在第二个分量（群代数副本）上是正则表示 (regular)。这就证明了 ii)。

此外，假设在单位元 $e$ 处的边界映射展开为 $\partial_i(v \otimes e) = \sum_w (w \otimes s_i(v,w))$，其中 $s_i(v,w) \in H$。那么对于任意的 $h \in H$，利用群作用与边界算子的相容性（交换律）：
$$ \partial_i(v \otimes h) = \partial_i(h \cdot (v \otimes e)) = h \cdot \partial_i(v \otimes e) = h \cdot \sum_w (w \otimes s_i(v,w)) = \sum_w (w \otimes h s_i(v,w)) $$

因此，如果我们按照字典序 (lex order) 重新排列 $C_i$ 的基底（先按基础图节点 $v$ 排序，再按群元素 $h$ 排序），那么边界算子 $\partial_i$ 的矩阵在区块 $(v,w)$ 处的值，正是群元素 $s_i(v,w)$ 对应的正则表示矩阵 $\rho_{reg}(s_i(v,w))$。

于是，我们在商复形上定义边界映射为 $\partial'_i(v) = \sum_w g_{v,w} w$，引理得证。 $\blacksquare$

---

**【概念说明与拆解】**

这个引理是量子提升码 (Lifted Product Codes) 和量子 Tanner 码等极其重要的高级构造中，最核心的代数定理之一！它揭示了“对称性”如何完美地转化为“矩阵块”。

*   **结构的直积化 (Tensor Product Structure)**：结论 i) 告诉我们，只要群作用是“自由的”，原来的大空间 $C_i$ 就完美地变成了“基础图（商空间）”与“群代数”的张量积。物理上这意味着：大图上的每一个物理比特（或校验），都可以被唯一地拆解为一个坐标对 $(b, h)$ —— 其中 $b$ 告诉你它在“基础图”的哪个位置（哪个轨道），$h$ 告诉你它在这个轨道里的哪一个“副本”。

*   **作用的局部化**：结论 ii) 中 $I \otimes \rho_{reg}(h)$ 非常规整。群元素的操作绝不会把基础图位置 $b$ 的比特挪到位置 $b'$ 去（第一部分是恒等算子 $I$），它只会在同一个位置 $b$ 的各个内部副本之间进行轮换（第二部分是群的正则表示 $\rho_{reg}$）。

*   **边界映射的块矩阵结构 (Block Matrix)**：结论 iii) 直接给出了大复形（提升后的量子码）校验矩阵的“图纸”。
    引理证明了，大矩阵 $\partial_i$ 可以被划分为一个个区块 (blocks)。每一个区块代表基础图中的两个节点 $(v, w)$ 之间的连接。
    *   如果基础图中 $v$ 和 $w$ 有连接，这个区块就是一个 $\vert{}H\vert{} \times \vert{}H\vert{}$ 的置换矩阵（具体为某个群元素的正则表示矩阵 $\rho_{reg}(s)$）。
    *   如果没有连接，这个区块就是一个全零矩阵。
    
    **这就把经典纠错码中常用的“多项式矩阵 / 准循环 (Quasi-Cyclic) 结构”，在坚实的代数拓扑基础上，完美推广到了任意非阿贝尔群！** 所有的提升码，本质上都是在基础图上给每条边分配一个群元素，然后用置换矩阵替换这些边来展开成大图。

### 补充注记：关于群表示中“左作用”与“右作用”的细节纠偏

如果您仔细推敲引理 2.2.6 的证明，会发现一个代数符号上的微妙瑕疵，这在处理**非阿贝尔群 (Non-abelian groups)** 时极其关键。

**1. 发现矛盾：右乘的出现**
在证明的推导中，我们得到了边界算子的展开式：
$$ \partial_i(v \otimes h) = \sum_w (w \otimes h s_i(v,w)) $$
请注意，这里群元素 $s_i(v,w)$ 是乘在 $h$ 的**右边**的（即操作为 $h \mapsto h s_i$）。

**2. 与“正则表示”定义的冲突**
回顾我们在【定义 2.2.1】中对正则表示的严格定义：
*   **左正则表示 (Left regular representation)**: $\rho_{L}(g) x = g x$ （乘在左边）
*   **右正则表示 (Right regular representation)**: $\rho_{R}(g) x = x g^{-1}$ （乘在右边，且必须取逆元，以保证同态的结合律 $\rho(ab)=\rho(a)\rho(b)$）

既然推导出的操作是右乘 $s_i(v,w)$，即 $x \mapsto x s_i$，我们可以把它改写为 $x \mapsto x (s_i^{-1})^{-1}$。这就完美对应了 **$s_i^{-1}$ 的右正则表示**！

**3. 结论与学术界行文习惯**
*   **严格地说**：原文中所写的“矩阵的区块为 $\rho_{reg}(s_i(v,w))$”是不够严谨的。如果默认 $\rho_{reg}$ 是左正则表示，这就完全错了；它应当被写成 **$\rho_{R}(s_i(v,w)^{-1})$**，或者在某些约定下写成左正则表示的转置。
*   **为什么教材/论文经常这么写？** 因为在早期研究中（包括绝大多数经典的 LDPC 码和早期的提升码），人们使用的都是**阿贝尔群（交换群，如 $\mathbb{Z}_m$ 或 $\mathbb{Z}_2^k$）**。在交换群中，左乘和右乘完全等价（$hs = sh$），并且在二元域群代数中经常有元素等于其自身的逆。因此作者们往往会省略这个细节。
*   **在量子 Tanner 码中的重要性**：当我们将理论推广到**非阿贝尔群**（这也是获得线性码距好量子码的核心途径）时，**左与右绝不能混淆**！通常的做法是：让基础图的群作用从**左边**乘（保持复形的对称性），而将基础图边上的“校验群元素”从**右边**乘（形成对应的矩阵块）。这种左右作用的相互对易性（$(hg)s = h(gs)$），是保证大复形结构自洽的代数基石。

### 定义 2.2.7：带号矩阵 / 符号矩阵 (Signed matrices)

**【教材定义直译】**
设 $H$ 为一个有限群，且令 $R = \mathbb{F}_2[H]$（即定义在二元域上的群代数）。我们称任何定义在 $R$ 上的矩阵为**带号矩阵 (signed matrix)**。

---

**【概念说明与拆解】**

*   **什么是定义在群代数 $R$ 上的矩阵？**
    在经典的线性纠错码中，奇偶校验矩阵的元素要么是 0，要么是 1（即定义在 $\mathbb{F}_2$ 上）。
    而“带号矩阵”的元素不再是简单的数字，而是**群代数 $\mathbb{F}_2[H]$ 中的元素**。这意味着，矩阵里的每一个非零项，实际上都“携带”了一个群元素 $g \in H$（或者群元素的线性组合）。

*   **为什么叫“带号 (Signed)”？**
    在图论中，普通的邻接矩阵或关联矩阵只用 1 表示“有连接”，用 0 表示“无连接”。
    如果在每条边上额外分配一个群元素（即给这条边赋予一个特定的“权重”、“方向”或“相位”），这种图在数学上被称为**带压图 (Voltage Graph)** 或带号图。对应地，记录这种连接关系的矩阵就被称为“带号矩阵”。它不仅告诉你“相连”，还告诉你“通过哪个群元素相连”。

*   **与引理 2.2.6 的无缝衔接**：
    上一节我们证明了，自由作用下的商复形，其边界矩阵展开后是由一个个 $\rho_{reg}(g)$ 矩阵块组成的。
    带号矩阵正是这种“块矩阵”的高度浓缩表达！我们不需要在草稿纸上画出庞大的 $|H| \times |H|$ 置换矩阵块，而是直接在一个小矩阵对应的位置上写下群元素 $g$。这个小小的带号矩阵，就是大复形（提升码）的“压缩图纸 (Base Matrix)”。


### 定义 2.2.8：提升矩阵 (Lifted matrix)

**【教材定义直译】**
令 $R = \mathbb{F}_2[H]$，且设 $X = \sum_{i,j} x_{ij}E_{ij}$ 是一个元素 $x_{ij} \in R$ 的带号矩阵。给定群 $H$ 在 $\mathbb{F}_2$ 上的一个表示 $(\rho, V)$，我们可以定义**提升矩阵 (lifted matrix)** 为：
$$ \rho(X) = \sum_{i,j} E_{ij} \otimes \rho(x_{ij}) $$
利用群表示的性质 $\rho(gh) = \rho(g)\rho(h)$，可以证明 $\rho(XY) = \rho(X)\rho(Y)$。

---

**【概念说明与拆解】**

*   **从“压缩图纸”到“宏伟建筑” (Lifting 操作)**：
    如果说上一节的“带号矩阵”是一张浓缩的小型设计图纸，那么“提升 (Lift)”操作就是具体的施工过程。
    *   $E_{ij}$ 是指示位置的单位矩阵（只有第 $i$ 行第 $j$ 列是 1，其余为 0）。
    *   张量积 $\otimes$ 在这里表现为矩阵的分块扩展（Kronecker 积）。
    *   如果我们把 $\rho$ 取为维数为 $|H| \times |H|$ 的正则表示（即置换矩阵），那么 $\rho(X)$ 的操作极其直观：**把原来小小的带号矩阵 $X$ 里的每一个群元素项 $x_{ij}$，直接替换成一个巨大的 $|H| \times |H|$ 的置换矩阵块 $\rho(x_{ij})$**。零元素则替换成 $|H| \times |H|$ 的全零矩阵块。这样，一个很小的带号矩阵，就“膨胀”成了一个巨型且极其稀疏的物理奇偶校验矩阵。

*   **极其重要的同态性质 $\rho(XY) = \rho(X)\rho(Y)$**：
    这绝不仅仅是一个普通的代数公式，它是**量子提升码 (Lifted Codes) 得以存在的代数基石**！
    回顾一下，构造量子 CSS 码的终极噩梦在于必须满足极其严苛的正交性条件：$H_X H_Z^T = 0$。
    这个同态性质告诉我们：**只要我们在小小的“带号矩阵”层面上（即在群代数 $R$ 上），设计出满足 $X \cdot Y = 0$ 的小矩阵，那么当我们把它们提升（展开）成巨大无比的物理矩阵后，它们必然自动满足 $\rho(X)\rho(Y) = 0$！**

*   **降维打击的设计思路**：
    这赋予了编码理论家一种“降维打击”的能力。我们不再需要直接去几万乘几万的庞大物理空间里艰难地寻找相互正交的巨型稀疏矩阵；我们只需要在一个非常小的维度（比如 $5 \times 5$ 的带号矩阵）上做代数方程求解。只要小图纸上的等式成立，大建筑就绝对不会崩塌。这是多项式矩阵 / 准循环 LDPC 码思想向非交换群最完美的推广。

### 构造带有自由作用的图 —— 提升 (Lifting)

**【教材前提设定】**
设 $G = (V,E)$ 为一个无向图，我们在顶点集 $V$ 上人为设定一个排序 (ordering)。根据约定，如果 $u \leqslant v$，我们将边记为 $(u,v) \in E$。令 $H$ 为对称群 $\operatorname{Sym}(l)$ 的一个子群。

### 定义 2.2.9：图的 $(H,l)$-提升 ($(H,l)$-lift of a graph)

**【教材定义直译】**
无向图 $G = (V,E)$ 的一个 **$(H,l)$-标签 (labelling)** 是一个映射函数 $s : E \to H \subseteq \operatorname{Sym}(l)$。
由此生成的**提升图 (lifted graph)** $G(s) = (V',E')$ 是一个拥有 $l$ 份顶点副本的图。其顶点集为 $V' = V \times [l]$。对于原图 $E$ 中的每一条边 $(u,v)$，我们在提升图 $E'$ 中，将顶点 $(u,i)$ 与顶点 $(v,s(u,v) \cdot i)$ 相连。

需要注意的是，如果一个图是较小基础图的 $(H,l)$-提升，那么群 $H$ 会在这个大图的顶点上产生一个自然的群作用：$h \cdot (u,i) = (u, h \cdot i)$。
**当群 $H$ 是交换群 (commutative) 时**，这个作用可以自然地扩展到边上：设 $e' = ((u,i), (v,s(e)i))$ 是大图中的一条边，那么群元素 $h$ 作用在边上必须满足
$$h \cdot e' = ((u, h \cdot i), (v, h \cdot s(e)i)) = ((u, h \cdot i), (v, s(e)(h \cdot i))) \in E'$$

---

**【概念说明与拆解】**

*   **为什么要给顶点排序 ($u \leqslant v$)？**
    因为这是一张无向图，边没有天然的方向。但是，当我们给边分配一个群元素（相当于一种变换操作）时，操作是有方向性的（从 $u$ 到 $v$ 是操作 $g$，从 $v$ 到 $u$ 就必须是逆操作 $g^{-1}$）。人为对顶点进行排序，是为了在数学书写上统一“基准方向”，避免群元素的分配出现歧义。

*   **几何图像：什么是“提升 (Lift)”？**
    想象原图 $G$ 是画在地板上的几个点和几条线。
    1.  **复制 (Copies)**：现在我们在原图的每个点上方，垂直建起一座有 $l$ 层楼的高塔。原来的一点 $u$，变成了 $l$ 个点 $(u, 1), (u, 2), \dots, (u, l)$。这些垂直方向上的点集在数学上被称为**纤维 (Fibers)**。
    2.  **连线 (Connecting)**：原图上 $u$ 和 $v$ 之间有一条边。在立体的高塔中，我们要把 $u$ 塔的房间和 $v$ 塔的房间连起来。怎么连？由这条边上的标签 $s(u,v)$ 决定！因为 $s(u,v)$ 是一个置换操作，它告诉我们：“把 $u$ 塔的第 $i$ 层，连接到 $v$ 塔的第 $s(u,v) \cdot i$ 层”。
    这样，一张简单的地板图，就被“拔高”成了错综复杂的立体网络（提升图）。这就是著名量子 LDPC 码——**提升乘积码 (Lifted Product Codes)** 的几何基础。

*   **为什么强调群 $H$ 必须是“交换群” (Commutative / Abelian)？**
    这是一个极其深刻的代数细节！
    在大图中，群元素 $h$ 想要作为一个“全局对称性”去平移大图（即 $h$ 作用在顶点上变成 $(u, h \cdot i)$）。
    为了保证平移后的图和原来长得一样（即原本连着的边，平移后依然合法地连着），原边 $((u,i), (v,s(e)i))$ 平移后，其终点必须合法。
    平移后的新起点是 $(u, h \cdot i)$。根据提升图的连线规则，这个新起点应该连接到的合法终点是：对新起点应用边标签 $s(e)$，也就是 $(v, s(e) \cdot (h \cdot i))$。
    但实际平移操作直接把原来的终点变为了 $(v, h \cdot (s(e) \cdot i))$。
    要让这两者完美吻合，**必须要求 $s(e) \cdot h = h \cdot s(e)$**！
    如果 $H$ 是阿贝尔群（交换群），这自然成立。但如果是**非阿贝尔群 (Non-abelian groups)**，这个等式通常不成立。这就是为什么早期的提升码只研究循环群，而后来发展出基于非交换群的量子 Tanner 码时，必须极其小心地将群作用区分为“左乘”和“右乘”的根本原因（见引理 2.2.6 补充注记）。

### 例 2.2.10：图提升的矩阵表示 (Graph lifting)

**【教材例题直译】**
设 $A_G$ 为原图 $G$ 的邻接矩阵。给定一个 $(H,l)$-标签，我们定义其**带号邻接矩阵 (signed adjacency matrix)** 为：将原邻接矩阵中第 $(u,v)$ 项的 $1$ 替换为群元素标签 $s(u,v)$ 后得到的矩阵。
$$A_G(s) = \sum_{(u,v) \in E(G)} s(u,v)E_{u,v}$$
其中 $E_{u,v}$ 是一个仅在第 $(u,v)$ 项非零且值为 1 的指示矩阵。
很容易看出，提升图 $G(s)$ 的完整物理邻接矩阵 $A_{G(s)}$，恰好等于对带号邻接矩阵应用自然表示：
$$A_{G(s)} = \rho_{nat}(A_G(s))$$

---

**【概念说明与拆解】**

*   **从普通邻接矩阵到“带号邻接矩阵” $A_G(s)$**：
    普通图的邻接矩阵里只有 $0$ 和 $1$，$1$ 代表“两点相连”。在带号图里，我们把这个 $1$ 升级成了群元素 $s(u,v)$。这个矩阵就是我们在【定义 2.2.7】中提到的“浓缩设计图纸”。它仅仅等于原图的大小（比如原图有 5 个点，它就是 $5 \times 5$ 的矩阵），但里面装的不再是数字，而是群元素。

*   **见证“提升 (Lift)”的矩阵化实现**：
    根据【定义 2.2.8】，我们把这张浓缩图纸进行物理展开。
    自然表示 $\rho_{nat}$ 会把每一个群元素 $s(u,v)$ 都转化为一个 $l \times l$ 的**置换矩阵 (Permutation Matrix)**。
    因此，操作 $\rho_{nat}(A_G(s))$ 就是把原来那个 $5 \times 5$ 小矩阵里的每一个群元素项，都直接替换成一个 $l \times l$ 的置换矩阵块；把 $0$ 替换成 $l \times l$ 的全零矩阵块。
    最终得到的是一个 $(5l) \times (5l)$ 的巨大分块矩阵。这个巨型矩阵，正是那张充满了高塔纤维的立体“提升图 $G(s)$”的真实物理邻接矩阵！

*   **总结**：
    这一个小节极其优雅地将图论里的“图提升（几何操作）”和代数里的“矩阵群表示（代数操作）”完全等价了起来。几何上拉高建塔、错综连线的复杂过程，在代数上仅仅是一个简单的 $\rho_{nat}$ 映射替换而已。这为量子 LDPC 码的奇偶校验矩阵构造提供了极大的便利。

### 前置概念：$H$-排序 ($H$-ordering)

在证明群作用如何传递到更复杂的 Tanner 构造之前，我们需要定义一个基础概念：**$H$-排序 ($H$-ordering)**。

**【概念说明】**
假设 $G$ 是一个 $d$-正则图，并且群 $H$ 作用在图 $G$ 上。
对于图中的任何一个顶点 $v$，它有 $d$ 个邻居。我们人为地给这些邻居编号（从 $1$ 到 $d$），记为 $ord_v(i)$，表示“顶点 $v$ 的第 $i$ 个邻居”。
所谓 **$H$-排序**，就是要求这个局部编号规则必须与群 $H$ 的全局对称性**完美相容**：
$$ ord_{hv}(i) = h \cdot ord_v(i) $$
**物理直觉**：这就像是给每个顶点周围的边贴上 1 到 $d$ 的数字标签。如果你站在顶点 $v$ 看向“1号边”，然后用群操作 $h$ 把你整个视野平移（或旋转）到了新的顶点 $h \cdot v$，此时你视野里正前方的边（平移过来的那条边），在新的顶点处也必须正好被称为“1号边”。局部视角的标号在全局对称变换下是不变的。

---

### 引理 2.2.11：自由作用在 Tanner 构造上的诱导

**【教材引理直译】**
如果我们有一个群 $H$ 在基础图 $L$ 上的**自由作用 (free action)**，并且配备了上述的 **$H$-排序**，那么对于任意的局部码 $C_0$，利用它们构造出的 **Tanner 图 $T(L, C_0)$ 上也必然存在群 $H$ 的自由作用**。

---

**【引理证明与核心逻辑拆解】**

为了证明这一点，我们需要明确 Tanner 图 $T(L, C_0)$ 的结构，并验证群作用在这个新图上既是合法的（图同态），又是自由的（无不动点）。

**1. Tanner 图的二分图结构**
在 $T(L, C_0)$ 这个二分图中：
*   **左侧顶点**：对应于原图 $L$ 的**边 (edges)**，记为 $e = (v, v')$。
*   **右侧顶点**：对应于“原图顶点与局部约束的组合”，记为 $(v, q)$，其中 $v$ 是 $L$ 的顶点，$q$ 是局部码 $C_0$ 的第 $q$ 个校验约束。

**2. 定义群 $H$ 在 Tanner 图上的自然作用**
我们要让群 $H$ 作用在这个巨大的二分图上，最自然的方式是：
*   **作用在右侧顶点**：$h \cdot (v,q) := (h(v), q)$。也就是说，群作用只“挪动”顶点 $v$ 的位置，而**绝不改变**局部约束的编号 $q$。
*   **作用在左侧顶点 (边)**：群作用在边 $e=(v,v')$ 上的效果和在原图 $L$ 上完全一样，即 $h \cdot e = (h(v), h(v'))$。由于群在原图 $L$ 上的作用是图同态，被变换后的边必定是原图中一条真实存在的边。

**3. 验证连边的合法性 (Graph Homomorphism)**
关键挑战来了：如果左侧的边 $e$ 原本与右侧的顶点 $(v, q)$ 之间**有连线**，那么在群元素 $h$ 作用后，变换后的新边 $h \cdot e$ 与变换后的新顶点 $(h(v), q)$ 之间**还能合法连线吗**？
*   **原图中的连线意义**：边 $(v,v')$ 与 $(v,q)$ 相连，物理上意味着这条边参与了顶点 $v$ 处的第 $q$ 个约束。代数上，这意味着存在某个局部序号 $k$（属于该约束的作用域），使得 $ord_v(k) = v'$。
*   **平移后的等式**：因为我们拥有完美的 **$H$-排序**，对上式两边同时施加群作用 $h$，得到 $h \cdot ord_v(k) = h(v')$，即 $ord_{h(v)}(k) = h(v')$。
*   **平移后的连线意义**：这个新等式翻译回物理图像就是：在新的顶点 $h(v)$ 处，那个带有同样序号 $k$ 的邻居，恰好就是 $h(v')$！既然局部序号 $k$ 没变，它必然依然参与第 $q$ 个约束。因此，边 $(h(v), h(v'))$ 必须与 $(h(v), q)$ 相连。
这证明了群作用是 Tanner 图的一个完美的**图同态 (Graph Homomorphism)**，即平移绝不会破坏 Tanner 图原有的连接规则。

**4. 验证自由性 (Freeness)**
*   因为群 $H$ 在基础图 $L$ 上的作用是自由的，意味着对于 $h \neq e$，$h(v) \neq v$ 且 $h(e) \neq e$。
*   因此，在 Tanner 图中，右侧节点 $h(v, q) = (h(v), q) \neq (v, q)$，左侧节点 $h(e) \neq e$。所有节点都不会在非平庸群操作下固定不动。
*   既然节点上是自由的，且这个图同态保持了二分图的左右结构（边不能翻转变成顶点），所以群作用在整个 Tanner 图（包括顶点和边）上都是严格**自由的**。 $\blacksquare$

**【总结】**
这个引理是构造“高度对称的量子 Tanner 码”的桥梁。它保证了：只要我们找了一个拥有良好群对称性的“骨架图”，在上面挂满各种复杂的局部校验规则（展开为 Tanner 图），那个宏观的群对称性不仅不会丢失，还会完美地“遗传”给整个复杂的校验网络。

### 定义与构造：平衡乘积复形 (Balanced Product of Complexes)

在有了前面关于“群作用”和“商复形”的所有术语铺垫之后，我们终于可以引入由 Breuckmann 和 Eberhardt 在 [BE21a] 中提出的核心概念：**平衡乘积 (Balanced Product)**。它是平衡乘积量子 LDPC 码的代数基石。

#### 1. 空间的“平衡商”定义

假设我们有两个都带有群 $H$ 作用的链复形 $\mathcal{C}$ 和 $\mathcal{D}$。
*(严格在代数上，我们要求群 $H$ 在 $\mathcal{C}$ 上是**左作用**，在 $\mathcal{D}$ 上是**右作用**；但由于我们在很多经典构造中主要关注群 $H$ 是**交换群 (commutative/abelian)** 的情况，左右作用相互等价，因此可以忽略这个方向性差异。)*

我们定义一个全新的“平衡张量积商空间”：
$$ C_i \otimes_H D_j := C_i \otimes D_j / \langle h \cdot v \otimes w - v \otimes h \cdot w \rangle $$

**【物理与几何直觉】**
这里的核心在于那个被模掉的等价关系：$h \cdot v \otimes w \equiv v \otimes h \cdot w$。
它的意思是：**群元素 $h$ 可以在张量积符号 $\otimes$ 左右自由“穿梭”或“转移”**。如果你在左边空间施加了一个对称性操作 $h$，它完全等价于你什么都不做，而是跑到右边空间去施加这个相同的操作。两边的对称性被这个张量积操作“平衡”地锁死了。

**【与普通商空间的联系】**
这其实是上一节提到的“单回复形商空间 $C_i/H$”极其自然的推广。因为你可以把 $C_i/H$ 视为 $C_i \otimes \mathbb{F}_2 / H$，其中群 $H$ 在域 $\mathbb{F}_2$ 上的作用是平凡的（即 $h \cdot 1 = 1$）。这样，左边出来的群元素转移到右边就被“吃掉”了，这就退化成了我们前面讨论的普通商复形。

#### 2. 构造整体平衡乘积复形 $\mathcal{C} \otimes_H \mathcal{D}$

构造方法非常简单：直接照搬普通“张量积复形”的整个图纸（包括空间的直和以及边界算子的分部求导法则），**仅仅把每一个出现普通张量积空间 $C_i \otimes D_j$ 的地方，都强制替换为上述的平衡商空间 $C_i \otimes_H D_j$**。

#### 3. 验证边界映射的良定义性 (Well-definedness)

在数学中，一旦我们人为定义了等价类（即强行规定 $h \cdot v \otimes w$ 和 $v \otimes h \cdot w$ 是同一个东西），我们就必须接受一个严峻的拷问：**当你对它们求边界（导数）时，结果还会是同一个东西吗？**如果不是，这个复形的几何结构就崩溃了。

幸运的是，得益于“群作用与边界算子的相容性（交换律）”，这完全成立。推导过程如下：

$$ \begin{aligned}
\partial(h \cdot v \otimes w) &= \partial(h \cdot v) \otimes w + (h \cdot v) \otimes \partial(w) \quad && \text{张量积复形的标准边界公式 (在 } \mathbb{F}_2 \text{ 上无负号)} \\
&= h \cdot \partial(v) \otimes w + h \cdot v \otimes \partial(w) \quad && \text{利用相容性：先求边界再作用 = 先作用再求边界} \\
&\equiv \partial(v) \otimes h \cdot w + v \otimes h \cdot \partial(w) \quad && \text{利用新空间的平衡等价关系：将左边的 } h \text{ 全部“转移”到右边} \\
&= \partial(v) \otimes h \cdot w + v \otimes \partial(h \cdot w) \quad && \text{在右侧再次利用相容性：将 } h \cdot \partial(w) \text{ 还原为 } \partial(h \cdot w) \\
&= \partial(v \otimes h \cdot w) \quad && \text{这正好是 } v \otimes h \cdot w \text{ 的边界展开式！}
\end{aligned} $$

**结论：**因为 $\partial(h \cdot v \otimes w) \equiv \partial(v \otimes h \cdot w)$ 在目标商空间中严格成立，这意味着无论你选择哪个代表元去求边界，最终得到的都是同一个合法的边界元素。因此，平衡乘积复形上的边界映射是**完美且良定义**的！

### 定义与构造：扭曲乘积 (Twisted Product)

在 [HHO21] 中定义的**扭曲乘积 (Twisted Product)** 的灵感来源于微分几何中的**纤维丛 (Fiber bundles)** 概念，这也是其名称的由来。不过，我们完全可以直接在链复形的层面上定义它，而无需纠结于底层的拓扑结构。

#### 1. 前置设定与扭曲映射 $\varphi$
为了定义扭曲乘积，我们需要以下设定：
*   两个链复形 $\mathcal{B}$（通常被称为底复形 Base）和 $\mathcal{F}$（通常被称为纤维复形 Fiber）。
*   一个群 $H$，并且群 $H$ 能够作用在 $\mathcal{F}$ 上。
*   **扭曲映射 (Twist map) $\varphi$**：这是一个将集合 $E = \{(b, a) \mid a \in \operatorname{supp}(\partial b) \; b \in B_1, a \in B_0\}$ 映射到群 $H$ 的函数。

**物理与几何对应：** 如果我们将 $\mathcal{B}$ 看作是对应于 1-链复形的二分图，那么集合 $E$ 实际上就是这个图 $\mathcal{B}$ 的**边集**。请注意，这个映射完全等同于我们在前面提到的**给图加上标签 (labeling)**（即给每条边分配一个群元素）。

#### 2. 扭曲乘积复形 $\mathcal{B} \otimes_\varphi \mathcal{F}$ 的构造
在此设定下，我们如下定义扭曲乘积 $\mathcal{B} \otimes_\varphi \mathcal{F}$：
*   **空间 (Spaces)**：与通常的张量积完全相同（即对应层级张量积的直和）。
*   **边界映射 (Boundary maps)**：发生扭曲。
    *   从 $B_i F_1 \to B_i F_0$ 的映射是**未扭曲的 (untwisted)**，即它仅仅是 $\operatorname{id} \otimes \partial_\mathcal{F}$。
    *   从 $B_1 F_i \to B_0 F_i$ 的映射是**扭曲的 (twisted)**。设 $f$ 是 $F_i$ 的一个基底元素，那么该映射定义为：

$$ \partial(b \otimes f) = \sum_{a \in \partial b} a \otimes (\varphi(b,a) \cdot f) $$

将这两部分结合起来，例如，我们可以得到映射 $\partial_2 : B_1 F_1 \to B_1 F_0 \oplus B_0 F_1$，其定义为：

$$ \partial_2(b \otimes f) = b \otimes \partial(f) + \sum_{a \in \partial b} a \otimes (\varphi(b,a) \cdot f) $$

我们可以类似地写出 $\partial_1$ 的表达式。

**【注记】**：
1. 当扭曲映射 $\varphi$ 是平凡的（即所有边都被映射到群的单位元，相当于没有发生扭曲）时，这个边界映射就退化成了普通张量积的标准边界映射。在这个意义上，扭曲乘积是张量积的一种直接推广。
2. 上述给出的具体公式仅在底复形 $\mathcal{B}$ 是 **1-链复形** 时才有效。

### 引理 2.2.12：扭曲乘积的合法性 [HHO21, Prop. 2.2]

**【引理表述】**
上述定义确实给出了一个合法的链复形，即满足 $\partial_1\partial_2 = 0$。

**【引理证明】**
设 $b$ 和 $f$ 分别是 $B_1$ 和 $F_1$ 的基底元素。我们直接对它们连续应用两次边界映射，在二元域 $\mathbb{F}_2$ 上（加减法等价）进行展开：

$$ \begin{aligned} \partial_1\partial_2(b \otimes f) &= \partial_1 \left(b \otimes \partial(f)\right) + \partial_1\left(\sum_{a \in \partial b} a \otimes (\varphi(b,a) \cdot f)\right) \\ &= \sum_{a \in \partial b} a \otimes (\varphi(b,a) \cdot \partial(f)) + \left(\sum_{a \in \partial b} a \otimes \partial\left(\varphi(b,a) \cdot f\right)\right) \\ &= \sum_{a \in \partial b} a \otimes \Big( \varphi(b,a) \cdot \partial(f) + \partial\left(\varphi(b,a) \cdot f\right) \Big) \\ &= \sum_{a \in \partial b} a \otimes 0 \\ &= 0 \end{aligned} $$

**解析最后一步为什么等于 0：**
这里括号内的项为 $\varphi(b,a) \cdot \partial(f) + \partial\left(\varphi(b,a) \cdot f\right)$。由于我们在二元域上工作，加法等于减法，这等价于要求 $\varphi(b,a) \cdot \partial(f) = \partial\left(\varphi(b,a) \cdot f\right)$。这正是我们在前面反复强调的**群作用与边界算子的相容性**（群作用必须和边界映射交换）。由于该相容性成立，各项两两抵消，结果必然为 0。证明完毕。 $\blacksquare$

### 引理 2.2.13：扭曲乘积与提升复形的同构关系

令 $\mathcal{B}(\varphi)$ 表示使用扭曲映射 $\varphi$ 作为标签的**提升 1-链复形 (lifted 1-complex)**。具体来说，该提升复形的空间为 $B(\varphi)_i = B_i \times H$，其边界映射定义为 $\partial_i(b,h) = \sum_{a \in \partial b} (a, \varphi(b,a)h)$。

**【引理 2.2.13 表述】**
映射 $\eta : B_iF_j \to B(\varphi)_iF_j/H$ 定义为 $\eta(b \otimes f) = (b,e) \otimes f$，它是复形之间的一个**同构 (isomorphism)**。也就是说：
$$ \mathcal{B} \otimes_\varphi \mathcal{F} \cong (\mathcal{B}(\varphi) \otimes \mathcal{F})/H $$

---

**【引理证明】**

*(注：为了使证明严密，我们首先明确群 $H$ 在张量积 $\mathcal{B}(\varphi) \otimes \mathcal{F}$ 上的作用与商空间等价关系。群 $H$ 在 $\mathcal{B}(\varphi)$ 上的作用定义为右乘的逆 $g \cdot (b, h) = (b, hg^{-1})$，在 $\mathcal{F}$ 上为左作用 $g \cdot f = gf$。因此，在平衡商复形 $(\mathcal{B}(\varphi) \otimes \mathcal{F})/H$ 中，其等价关系严格定义为 $(b, hg^{-1}) \otimes (gf) \sim (b, h) \otimes f$。)*

在每个空间上，我们可以定义其逆映射为 $\eta^{-1}\big((b,h) \otimes f\big) = b \otimes hf$。
**（良定义性验证）**：由于该映射的定义域是商空间，必须证明它与等价类中代表元的选取无关。假设我们选取等价类中的另一个代表元 $(b, hg^{-1}) \otimes (gf)$，代入 $\eta^{-1}$ 的定义得到 $b \otimes (hg^{-1})(gf) = b \otimes hf$。计算结果完全一致，这证明了 $\eta^{-1}$ 是**良定义的 (well-defined)**。显然 $\eta^{-1}\eta(b \otimes f) = \eta^{-1}((b,e) \otimes f) = b \otimes ef = b \otimes f$，即 $\eta^{-1}\eta$ 是恒等映射。

现在，在取商之前，$B(\varphi)_iF_j$ 的基底由 $(b,h) \otimes f$ 构成。利用上述等价关系，并特取 $g = h$，我们有：
$$ (b,h) \otimes f = (b, eh) \otimes f \sim (b,e) \otimes hf $$
这就是商空间中元素的化简规则。因此，形如 $((b,e) \otimes f)$ 的元素可以作为商空间 $B(\varphi)_iF_j/H$ 的基底，对于这些基底，有 $\eta\eta^{-1}((b,e) \otimes f) = ((b,e) \otimes f)$。

为了验证这是复形之间的同构，我们还需要检查它与边界映射是否满足**交换性 (commutativity)**，即需要验证以下交换图表：

$$
\begin{matrix}
b \otimes f & \xrightarrow{\quad \partial \quad} & b \otimes \partial f + \sum_{a \in \partial b} a \otimes \varphi(b,a)f \\
\big\downarrow \eta & & \big\downarrow \eta \\
(b,e) \otimes f & \xrightarrow{\quad \partial' \quad} & (b,e) \otimes \partial f + \partial(b,e) \otimes f
\end{matrix}
$$

根据定义，两行映射结果的第一项求和显然是相等的。

对于第二项的推导如下：
$$ \begin{aligned} \partial(b,e) \otimes f &= \sum_{a \in \partial b} (a, \varphi(b,a)) \otimes f & \text{根据提升复形的边界映射定义} \\ &= \sum_{a \in \partial b} (a,e) \otimes \varphi(b,a)f & \text{根据商空间中的等价关系替换} \\ &= \eta\left(\sum_{a \in \partial b} a \otimes \varphi(b,a)f\right) \end{aligned} $$

因此，上下两条路径的结果完全一致，该图表是交换的。证明完毕。 $\blacksquare$

### 提升乘积 (Lifted Product) 的定义与定理 2.2.14

**【阅读提示：关于文献中符号记法与代数结构的严密重构】**
在原文献中，作者在定义 2.2.4 把 $\mathcal{C}/H$ 定义为 $\mathbb{F}_2$ 上的商复形，边界映射 $\partial(v - h\cdot v) = 0$ 直接把群元（Twist 信息）给抹除掉了。然而，在接下来的定理中，作者却写出了 $\mathcal{C}/H \otimes_{\mathbb{F}_2[H]} \mathcal{D}/H$ 这样的式子。这是一个严重的代数语法错误（符号滥用）：既然 $\mathcal{C}/H$ 的矩阵元素都在 $\mathbb{F}_2$ 中，它根本就不具备 $\mathbb{F}_2[H]$-模的结构，无法直接在环 $\mathbb{F}_2[H]$ 上做张量积。

为了填补这个逻辑断层，我们必须先严密重构定义，将带有群元的 $\mathbb{F}_2[H]$-模结构显式地写出来，然后再构造同构映射并验证交换图。以下是定理 2.2.14 第一个同构 $LP(\mathcal{C}, \mathcal{D}) \cong \mathcal{C}/H \otimes_\varphi \mathcal{D}$ 的严密证明。

#### 第一步：纠正符号，严密定义“符号商复形” (Signed Quotient Complex)

为了能在 $\mathbb{F}_2[H]$ 上做张量积，我们必须把丢失的群元找回来。
设 $\mathcal{C}$ 带有 $H$ 的自由作用。我们定义一个全新的链复形 $\tilde{\mathcal{C}}$（即 Lifted Product 实际操作的那个“Signed Matrix”复形）：
*   **空间**：$\tilde{\mathcal{C}}_i$ 是以 $\mathcal{C}_i/H$（轨道代表元）为基底，在群环 $R = \mathbb{F}_2[H]$ 上生成的自由模。
*   **边界映射**：$\tilde{\partial}^{\mathcal{C}}: \tilde{\mathcal{C}}_i \to \tilde{\mathcal{C}}_{i-1}$。对于基底元素 $c \in \mathcal{C}_i/H$，定义：

$$ \tilde{\partial}^{\mathcal{C}}(c) = \sum_{c' \in \partial c} c' \cdot s_{\mathcal{C}}(c, c') $$

这里 $s_{\mathcal{C}}(c, c') \in H$ 是如何定义的呢？由于原复形 $\mathcal{C}$ 带有群 $H$ 的自由作用，根据**引理 2.2.6**，空间可以分解为 $\mathcal{C}_i \cong \mathcal{C}_i/H \otimes \mathbb{F}_2[H]$。在原复形 $\mathcal{C}$ 中对单位元轨道代表元 $c \otimes e$ 求边界时，其展开式自然具有 $\partial(c \otimes e) = \sum_{c'} c' \otimes s_{\mathcal{C}}(c, c')$ 的形式。这个 $s_{\mathcal{C}}(c, c')$ 就是由原复形边界映射的正则表示块结构中提取出来的、连接这两个轨道代表元的唯一群元素。它在代数上扮演了给商复形“打标签”的作用。

由引理 2.2.6 的实质内涵可知，原复形 $\mathcal{C}$ 其实就是对 $\tilde{\mathcal{C}}$ 应用右正则表示 $\rho_{reg}$ 的结果：

$$ \mathcal{C} \cong \rho_{reg}(\tilde{\mathcal{C}}) $$

同理，我们也构造复形 $\tilde{\mathcal{D}}$，其边界映射带有对应的群元 $s_{\mathcal{D}}(d, d')$。

#### 第二步：重构 Lifted Product，并定义同构映射 $\eta$

现在的 Lifted Product 拥有了完全合法的代数定义：

$$ LP(\mathcal{C}, \mathcal{D}) = \rho_{reg}\left( \tilde{\mathcal{C}} \otimes_{\mathbb{F}_2[H]} \tilde{\mathcal{D}} \right) $$

**解析 $LP$ 的空间结构**：
在环 $\mathbb{F}_2[H]$ 上的张量积 $\tilde{\mathcal{C}}_i \otimes_{\mathbb{F}_2[H]} \tilde{\mathcal{D}}_j$ 的基底是 $c \otimes d$（其中 $c \in \mathcal{C}_i/H, d \in \mathcal{D}_j/H$）。
对其应用 $\rho_{reg}$ 后，空间从 $\mathbb{F}_2[H]$-模退化为 $\mathbb{F}_2$-向量空间，其基底扩充为 $c \otimes d \otimes h$（其中 $h \in H$）。

**解析 Twisted Product 的空间结构**：
目标空间是 $\mathcal{C}/H \otimes_\varphi \mathcal{D}$。它的基底是 $c \otimes y$，其中 $c \in \mathcal{C}_i/H$，且 $y \in \mathcal{D}_j$。
根据引理 2.2.6，$\mathcal{D}_j \cong \mathcal{D}_j/H \otimes \mathbb{F}_2[H]$。所以 $y$ 可以唯一表示为 $d \otimes h$（代表轨道 $d$ 上的元素被 $h$ 作用）。

**定义同构映射 $\eta$**：
定义线性映射 $\eta: LP(\mathcal{C}, \mathcal{D}) \to \mathcal{C}/H \otimes_\varphi \mathcal{D}$ 为：

$$ \eta(c \otimes d \otimes h) = c \otimes (d \otimes h) \equiv c \otimes y $$

这里的结合律变换，在代数上完美对应了把群元 $h$ 的信息从张量积的“公共载体”转移给了纤维 $\mathcal{D}$。由于这仅仅是基底的重新组合，$\eta$ 显然是双射。

#### 第三步：验证边界映射的交换图 (Commutativity)

这是全场最核心的一步：我们要证明在 Lifted 侧算出来的边界，通过 $\eta$ 映射过去，和在 Twisted 侧直接算出来的边界完全相等。

**1. 在 Lifted Product 侧计算边界：**
在 $\mathbb{F}_2[H]$ 上的张量积边界映射为 $\tilde{\partial}^{\mathcal{C}} \otimes id + id \otimes \tilde{\partial}^{\mathcal{D}}$。对其应用 $\rho_{reg}$ 后，作用在基底 $c \otimes d \otimes h$ 上：

$$ \partial^{LP}(c \otimes d \otimes h) = \rho_{reg}(\tilde{\partial}^{\mathcal{C}}(c) \otimes d)(h) + \rho_{reg}(c \otimes \tilde{\partial}^{\mathcal{D}}(d))(h) $$

代入 $\tilde{\partial}$ 的定义（注意 $\rho_{reg}(g)(h) = gh$）：

$$ \partial^{LP}(c \otimes d \otimes h) = \sum_{c' \in \partial c} c' \otimes d \otimes (s_{\mathcal{C}}(c, c')h) + \sum_{d' \in \partial d} c \otimes d' \otimes (s_{\mathcal{D}}(d, d')h) $$

**2. 将结果通过 $\eta$ 映射到 Twisted 侧：**
对上式两端应用映射 $\eta$（将群元 $h$ 绑定给 $d$ 形成 $\mathcal{D}$ 中的元素）：

$$ \eta(\partial^{LP}) = \sum_{c' \in \partial c} c' \otimes \underbrace{(d \otimes s_{\mathcal{C}}(c, c')h)}_{\text{项 A}} + \sum_{d' \in \partial d} c \otimes \underbrace{(d' \otimes s_{\mathcal{D}}(d, d')h)}_{\text{项 B}} $$

**3. 在 Twisted Product 侧直接计算边界：**
在 Twisted 侧，基底为 $c \otimes y$（其中 $y = d \otimes h \in \mathcal{D}_j$）。对于扭曲乘积 $\mathcal{C}/H \otimes_\varphi \mathcal{D}$，其对应的扭曲映射 $\varphi$ 在代数上正是由底空间商复形的标签 $s_{\mathcal{C}}$ 诱导的，即 $\varphi(c, c') = s_{\mathcal{C}}(c, c')$。
其 Twisted 边界算子定义为：

$$ \partial^{TP}(c \otimes y) = \sum_{c' \in \partial c} c' \otimes (s_{\mathcal{C}}(c, c') \cdot y) + c \otimes \partial^{\mathcal{D}}(y) $$

现在我们来比对两边的结果：
*   **对比第一项**：在 Twisted 侧，群元 $s_{\mathcal{C}}(c, c')$ 作用在 $y = (d \otimes h)$ 上，根据群作用定义，结果是 $d \otimes (s_{\mathcal{C}}(c, c')h)$。这与上面 $\eta$ 映射过来的 **“项 A”** 完全相等！
*   **对比第二项**：在 Twisted 侧，对 $y = (d \otimes h)$ 求复形 $\mathcal{D}$ 内部的边界 $\partial^{\mathcal{D}}$。根据引理 2.2.6 的结构 $\partial^{\mathcal{D}} = \partial' \otimes \rho_{reg}$，结果正是 $\sum_{d'} d' \otimes (s_{\mathcal{D}}(d, d')h)$。这与上面 $\eta$ 映射过来的 **“项 B”** 完全相等！

**结论：**

$$ \eta \circ \partial^{LP} = \partial^{TP} \circ \eta $$

交换图严格闭环。至此，我们将作者随意写下的 $\mathcal{C}/H \otimes_{\mathbb{F}_2[H]} \mathcal{D}/H$ 赋予了严密的“符号模”语义，并用无可辩驳的代数步骤证明了 Lifted Product 和 Twisted Product 的同构关系。

### 2.2.5 乘积总结 (Summary of the products)

由于表格容易导致长公式被强制换行，我们将三种核心乘积的特征总结如下：

*   **提升乘积 (Lifted Product)**
    *   **适用条件**：两者均为 1-链复形，且带有阿贝尔群 $H$ 的自由作用。
    *   **定义**：$\rho_{reg}(\mathcal{C}/H \otimes_{\mathbb{F}_2[H]} \mathcal{D}/H)$
*   **扭曲乘积 (Twisted Product)**
    *   **适用条件**：$\mathcal{C}$ 是 1-链复形且带有群 $H$ 的自由作用；$\mathcal{D}$ 带有群 $H$ 的任意作用。
    *   **定义**：$(\mathcal{C} \otimes \mathcal{D})/H$
*   **平衡乘积 (Balanced Product)**
    *   **适用条件**：$\mathcal{C}, \mathcal{D}$ 均带有群 $H$ 的任意作用。
    *   **定义**：$(\mathcal{C} \otimes \mathcal{D})/H$

#### [HHO21] 的构造方法
构造一个大小为 $n$ 的集合 $\vert{}B_1\vert{}$ 以及大小为 $m = 3n/4$ 的集合 $B_0$。对于每个 $a \in B_0$，我们以 $\Delta/n$ 的概率选取一条边 $(b,a)$，其中 $b \in B_1$，且 $\Delta = \Theta(\log^2 n)$。将 $B_0$ 划分为 $\Theta(\log n)$ 个大小相等的桶 (buckets)，对于第 $j$ 个桶，从 $\{0, \cdots, \sqrt{l} - 1\}$ 中均匀采样选取 $\varphi_j$，其中 $l = \Theta(n)$ 为奇数。通过以下方式构造扭曲映射：以 $1/2$ 的概率令 $\varphi(b,a) = 0$；如果 $a$ 属于第 $j$ 个桶，则以 $1/2$ 的概率令 $\varphi(b,a) = \varphi_j$。最终的量子码为 $\mathcal{B} \otimes_\varphi \mathcal{F}$，其中 $\mathcal{F} : \mathbb{Z}_l \to \mathbb{Z}_l$ 是 $l$ 个顶点上的循环图 (cycle graph) 所关联的图复形。

#### [PK21] 的构造方法
取一个具有 $2n$ 个顶点的 $d$-正则扩展图 $G$，并选取一个“好”经典码 $C_0$。构造 $G$ 的一个随机 $\mathbb{Z}_l$-提升，其中 $l = 2^n - 1$。这里的“随机提升”意味着对于每条边 $e \in E(G)$，$s(e)$ 是从 $\mathbb{Z}_l$ 中均匀随机选取的。由于 $\mathbb{Z}_l$ 是阿贝尔群，提升图 $G(s)$ 具有 $\mathbb{Z}_l$ 的自由作用；根据引理 2.2.11，Tanner 图 $\mathcal{B} := T(G(s), C_0)$ 也具有该自由作用。令 $\mathcal{F}$ 为 $l$ 个顶点上的循环图所关联的图复形。如果 $l$ 为奇数，那么 $\mathbb{Z}_l$ 在 $\mathcal{F}$ 上的自然作用是自由的。最终构造出的码由提升乘积给出：$LP(\mathcal{B}, \mathcal{F}) = \rho_{reg}(\mathcal{B}/H \otimes_{\mathbb{F}_2[\mathbb{Z}_l]} \mathcal{F}/H)$。

### 补充视角：从层级图的“提升”看三种乘积 (Lifts on Layered Graphs)

如果我们不想过多地谈论抽象的“群作用”，我们其实可以用一种稍微不同的、更具几何直觉的视角来统一审视这三种乘积。这种视角类似于 [TZ14] 中将张量积定义为 Tanner 图的超图乘积。

#### 1. 前置设定与张量积的“三层楼”结构
假设 $\mathcal{B}$ 和 $\mathcal{F}$ 都是 1-链复形，并且它们边界映射对应的二分图上都带有群元素标签 $s_B$ 和 $s_F$。我们用 $\hat{\mathcal{B}}$ 和 $\hat{\mathcal{F}}$ 来表示它们展开后的**提升复形 (Lifted complexes)**。

如果我们对未提升的原复形做普通的张量积 $\mathcal{B} \otimes \mathcal{F}$，得到的是一个长度为 3 的复形。在图的视角下，这是一个拥有**三层节点的层级图 (Layered graph)**：
*   **第一层 (顶层)**：$B_1 \times F_1$ （双重物理比特层）
*   **第二层 (中间层)**：$B_1 \times F_0 \cup B_0 \times F_1$ （$X$ 校验与 $Z$ 校验层）
*   **第三层 (底层)**：$B_0 \times F_0$ （双重校验层 / 稳定子的稳定子）

#### 2. 三种乘积的几何重构
在这种层级图的视角下，三种复杂的代数乘积可以被极其直观地重新定义：

*   **提升乘积 (Lifted Product)**：
    **“先张量积，再整体提升”**。
    你可以看作是先构造了上述的 $\mathcal{B} \otimes \mathcal{F}$ 三层图，然后再把这个三层图整体提升为 $\widehat{\mathcal{B} \otimes \mathcal{F}}$。
    其中，群元素标签是如何分配的呢？
    *   如果在两层之间是 $\mathcal{B}$ 发生了降级（即从 $B_1 \to B_0$，例如边 $B_1F_0 \to B_0F_0$ 或 $B_1F_1 \to B_0F_1$），我们就**使用 $\mathcal{B}$ 原有的边标签 $s_B$**。
    *   如果是 $\mathcal{F}$ 发生了降级（即从 $F_1 \to F_0$，例如边 $B_0F_1 \to B_0F_0$ 或 $B_1F_1 \to B_1F_0$），我们就**使用 $\mathcal{F}$ 原有的边标签 $s_F$**。

*   **扭曲乘积 (Twisted Product)**：
    **“底图不变，只提升纤维图，再做张量积”**。
    它可以极其简洁地表示为 $\mathcal{B} \otimes \hat{\mathcal{F}}$。这意味着我们把底复形 $\mathcal{B}$ 留在原地，只把纤维复形 $\mathcal{F}$ 展开成高塔，然后再把它们相乘。

*   **平衡乘积 (Balanced Product)**：
    **“分别提升，张量积后合并等价类”**。
    平衡乘积复形 $\mathcal{E}$ 可以这样理解：我们先分别把 $\mathcal{B}$ 和 $\mathcal{F}$ 都提升成巨大的复形 $\hat{\mathcal{B}}$ 和 $\hat{\mathcal{F}}$，然后对它们做张量积得到 $\hat{\mathcal{B}} \otimes \hat{\mathcal{F}}$。但这个图太大了（包含了 $|H|^2$ 倍的冗余），所以我们通过一个等价关系将其“压缩”回 $\mathcal{E}$：
    $$[(u,h), (v,g)] \sim [(u,e), (v,gh^{-1})]$$
    这完美对应了我们前面提到的“平衡商”操作——即允许群元素在左边节点 $u$ 和右边节点 $v$ 之间来回转移，抵消掉多余的对称度。

**总结**：这个视角剥离了繁琐的群同态证明，直接从图的“复制、连线、折叠”这些纯粹的图论操作出发，给出了高级量子码构造的直观且高度统一的几何图像。

### 3. 分析与解码：扭曲乘积的 Künneth 公式

既然我们已经看到了许多基于复形的代数构造，下一步就是详细分析这些码的参数。理解线性码的第一步是计算其**维数 (Dimension)**，即**码率 (Rate)**。

由于这些构造具有高度的代数结构，我们可以将代码（即同调或上同调群）分解为其组成部分。这不仅能让我们轻松计算维数，也为后续**界定码距 (Distance)** 和**设计解码算法**提供了极大帮助。特别地，我们将证明一个推广版的 **Künneth 公式**，它不仅适用于普通的张量积，还适用于更广泛的平衡乘积和扭曲乘积。

#### 1. 扭曲乘积的 Künneth 公式 (Lemma 3.0.1)

我们首先来看扭曲乘积的情况。之所以单独拿出来讨论，原因有二：一是它的结论非常完美，定理形式与普通张量积完全一致；二是我们在后续的解码中会大量用到它。

**【引理 3.0.1 表述】**
设 $\mathcal{E} = \mathcal{B} \otimes_\varphi \mathcal{F}$ 为扭曲乘积复形，并假设纤维复形的同调群 $H_0(\mathcal{F})$ 和 $H_1(\mathcal{F})$ 拥有这样一组代表元：**扭曲映射 $\varphi$ 作用在它们上面等同于恒等作用 (Identity)**。
那么，新复形的第 1 级同调群满足完美的张量分解：
$$ H_1(\mathcal{E}) \cong \big(H_1(\mathcal{B}) \otimes H_0(\mathcal{F})\big) \oplus \big(H_0(\mathcal{B}) \otimes H_1(\mathcal{F})\big) $$
*(注：文献中简写为 $H_1(\mathcal{B})H_0(\mathcal{F}) \oplus H_0(\mathcal{B})H_1(\mathcal{F})$)*

**【引理 3.0.1 的完整证明】**
本引理的证明，实际上需要分为三部分：1、建立从右到左的映射，并且证明真的是个映射（良定义）；2、证明这个映射是单射；3、证明这个映射是满射。实际上原文的证明只是在说明3，甚至没有说明1，换言之，3的说明也不完全。所以笔者这里考虑补上1和2的证明。

**第一部分：重新定义映射 $\Phi$ 及其良定义性验证**

根据原文，复形 $\mathcal{B}$ 和 $\mathcal{F}$ 都是 1-复形（1-complexes）。
我们提取 Lemma 3.0.1 中被 Twist 映射 $\varphi$ 作用为恒等映射（acts by identity）的特殊代表元张成的子空间：
令 $V = \text{span}\{v_i\} \subset F_0$。
令 $U = \text{span}\{u_i\} \subset F_1$。

我们重新定义映射 $\Phi$，它的定义域只对 $\mathcal{B}$ 取同调等价类，而对 $\mathcal{F}$ 直接取具体的向量空间 $V$ 和 $U$：
$$ \Phi: \left(H_1(\mathcal{B}) \otimes V\right) \oplus \left(H_0(\mathcal{B}) \otimes U\right) \to H_1(\mathcal{E}) $$
其映射法则为，对于给定的同调类 $[b] \in H_1(\mathcal{B}), [a] \in H_0(\mathcal{B})$，以及固定的向量 $v \in V, u \in U$：
$$ \Phi([b] \otimes v + [a] \otimes u) = [b \otimes v + a \otimes u] $$

**为什么这个新定义是绝对严密的？（良定义性验证）**
我们现在来验证，这个重新定义的 $\Phi$ 是否真的不受代表元选取的影响。
1. **对于 $\mathcal{F}$ 端（纤维端）**：我们彻底抛弃了等价类 $[v]$ 和 $[u]$，输入被严格限制为具体的向量 $v \in V$ 和 $u \in U$。因为不存在“更换代表元”的操作，所以直接免疫了“Twist 后边界无法抵消”的致命问题。
2. **对于 $\mathcal{B}_1$ 端（基底 1-阶端）**：我们需要考虑更换 $[b]$ 的代表元。但由于 $\mathcal{B}$ 是 1-复形，它最高只有 1 阶空间，这意味着 $B_2 = 0$。因此，1-阶循环 $b$ 根本没有任何上一级的边界（$\partial w_2$）可以加减。同调类 $[b]$ 里有且只有它自己这一个元素！代表元是唯一的，良定义性显然成立。
3. **对于 $\mathcal{B}_0$ 端（基底 0-阶端）**：这是全场唯一需要，且真正能够更换代表元的地方。对于 $[a] \in H_0(\mathcal{B})$，我们可以将其更换为 $a' = a + \partial w$（其中 $w \in B_1$）。我们将新的代表元代入 $\Phi$：
$$ \Phi([a + \partial w] \otimes u) = [(a + \partial w) \otimes u] = [a \otimes u + \partial w \otimes u] $$
我们需要证明，多出来的这部分 $\partial w \otimes u$ 在总复形 $\mathcal{E}$ 中是一个纯边界，从而在 $H_1(\mathcal{E})$ 中被模掉。我们在总复形 $\mathcal{E}$ 中取 2-阶元素 $w \otimes u \in B_1 \otimes F_1$，并对它求 Twisted 边界算子：
$$ \partial_{\mathcal{E}}(w \otimes u) = (\partial w) \otimes \varphi(u) + w \otimes \partial u $$
奇迹（也就是作者预设的前提）在这里发生：因为 $u \in U = \text{span}\{u_i\}$，根据 Lemma 3.0.1 的强制规定，$\varphi$ 对其表现为恒等映射，所以 $\varphi(u) = u$。因为 $u$ 是 1-阶循环，所以 $\partial u = 0$。代入上述条件，等式瞬间净化为：
$$ \partial_{\mathcal{E}}(w \otimes u) = (\partial w) \otimes u $$
这完美证明了差值 $\partial w \otimes u$ 就是 $\mathcal{E}$ 中的一个边界。因此 $[a' \otimes u] = [a \otimes u]$，映射对 $\mathcal{B}_0$ 代表元的选取是严格独立的！

**结论：**
通过将映射的定义域从名义上的 $H_0(\mathcal{F})$ 和 $H_1(\mathcal{F})$ 降维回具体的、被 $\varphi$ 锁定的子空间 $V$ 和 $U$，我们彻底绕开了 Twist 映射导致的交叉项灾难。这才是这个证明在底层代数结构上唯一合法的映射路径。

**第二部分：映射的单射性 (Injectivity)**

为了证明 $\Phi$ 是单射，我们需要证明 $\ker(\Phi) = 0$。假设存在元素 $X \in \left(H_1(\mathcal{B}) \otimes V\right) \oplus \left(H_0(\mathcal{B}) \otimes U\right)$，满足 $\Phi(X) = 0$。我们可以将 $X$ 写为：
$$X = \sum_{i} [b_i] \otimes v_i + \sum_{j} [a_j] \otimes u_j$$
由于 $\Phi(X) = 0$，这意味着它在总复形 $\mathcal{E}$ 中的代表元 $x = \sum_i b_i \otimes v_i + \sum_j a_j \otimes u_j \in \mathcal{E}_1$ 是一个纯边界。因此，必定存在某个 2-阶链 $W \in \mathcal{E}_2$，使得 $\partial_{\mathcal{E}}(W) = x$。

设 $\{e\}$ 为基底空间 $B_1$ 的一组线性无关的基底。由于 $\mathcal{E}_2 = B_1 \otimes F_1$，任何 $W$ 都可以被唯一地展开为：
$$W = \sum_{e \in B_1} e \otimes f_e \quad (f_e \in F_1)$$
对 $W$ 施加 Twisted 边界算子展开：
$$\partial_{\mathcal{E}}(W) = \sum_e \partial_{\mathcal{E}}(e \otimes f_e) = \sum_e \left( e \otimes \partial^\mathcal{F} f_e + \sum_{k \in \partial^\mathcal{B} e} k \otimes \varphi(e,k)f_e \right)$$
将 $\partial_{\mathcal{E}}(W) = x$ 代入，并在直和空间 $\mathcal{E}_1 = (B_1 \otimes F_0) \oplus (B_0 \otimes F_1)$ 中分别对比这两个独立的分量：

**步骤 1：处理 $B_1 \otimes F_0$ 分量**
$$\sum_i b_i \otimes v_i = \sum_e e \otimes \partial^\mathcal{F} f_e$$
由于 $\{e\}$ 是 $B_1$ 的基底，我们可以将左侧的每个 $b_i \in B_1$ 均用基底线性表出：$b_i = \sum_e c_{i,e} e$。代入左侧并交换求和次序：
$$\sum_i \left( \sum_e c_{i,e} e \right) \otimes v_i = \sum_e e \otimes \left( \sum_i c_{i,e} v_i \right)$$
令 $v'_e = \sum_i c_{i,e} v_i$。由于 $v_i \in V$ 且 $V$ 是线性空间，故 $v'_e \in V$。现在等式变为：
$$\sum_e e \otimes v'_e = \sum_e e \otimes \partial^\mathcal{F} f_e$$
因为基底 $\{e\}$ 是线性无关的，我们可以逐项对比张量积在 $F_0$ 上的系数：
$$ \text{对于所有的 } e，必有 \quad v'_e = \partial^\mathcal{F} f_e $$
此时，$v'_e \in V$，而 $\partial^\mathcal{F} f_e \in \text{Im}(\partial^\mathcal{F})$。由于非平凡代表元空间 $V$ 与纯边界空间交集为零，即 $V \cap \text{Im}(\partial^\mathcal{F}) = \{0\}$，上述等式要成立，两边必须同时等于 0。因为 $v'_e = 0$，左侧总和 $\sum_e e \otimes v'_e = 0$。映射回定义域，这证明了 $X$ 的第一部分严格为零：$\sum_i [b_i] \otimes v_i = 0$。
因为 $\partial^\mathcal{F} f_e = 0$，这说明对于所有的 $e$，均有 $f_e \in \ker(\partial^\mathcal{F}) = Z_1(\mathcal{F})$。由于 $\mathcal{F}$ 是 1-复形，$U$ 是 $H_1(\mathcal{F})$ 的全量代表元空间，故严格有 $U = Z_1(\mathcal{F})$。因此，严格得出 $f_e \in U$。

**步骤 2：处理 $B_0 \otimes F_1$ 分量**
$$\sum_j a_j \otimes u_j = \sum_e \sum_{k \in \partial^\mathcal{B} e} k \otimes \varphi(e,k)f_e$$
根据步骤 1 的严密结论，$f_e \in U$。
根据 Lemma 3.0.1 的强制规定，$\varphi$ 对 $U$ 中的元素作用严格为恒等映射。因此，$\varphi(e,k)f_e = f_e$。将此结果代回等式右侧：
$$\text{右侧} = \sum_e \sum_{k \in \partial^\mathcal{B} e} k \otimes f_e$$
将 $f_e$ 作为公因式提取到张量积右侧：
$$\text{右侧} = \sum_e \left( \sum_{k \in \partial^\mathcal{B} e} k \right) \otimes f_e = \sum_e (\partial^\mathcal{B} e) \otimes f_e$$
现在等式化简为：
$$\sum_j a_j \otimes u_j = \sum_e (\partial^\mathcal{B} e) \otimes f_e$$
我们将此等式映射到定义域的商空间 $H_0(\mathcal{B}) \otimes U$ 中：左侧映射为 $\sum_j [a_j] \otimes u_j$。右侧映射为 $\sum_e [\partial^\mathcal{B} e] \otimes f_e$。根据同调群的基本定义，任何边界元素的同调类恒为 0，即 $[\partial^\mathcal{B} e] = 0$。因此，右侧在商空间中恒等于 0。这证明了 $X$ 的第二部分严格为零：$\sum_j [a_j] \otimes u_j = 0$。

**单射结论**
通过步骤 1 和步骤 2 的完全展开与代数消元，我们证明了：如果 $\Phi(X) = 0$，则 $X$ 的两个直和分量必须分别严格等于 0。即 $X = 0 + 0 = 0$。故映射 $\Phi$ 的核 $\ker(\Phi) = \{0\}$，$\Phi$ 是单射证明完毕。

**第三部分：映射的满射性 (Surjectivity)**
实际就是要说明，$H_1(\mathcal E)$中每一个等价类$[y]$都能找到对应的分解方案，即要分解为$y=\sum a\otimes u+\sum b\otimes v+\partial_2^{\mathcal E}(z)$，其中$u$以及$v$必须分别属于$U$和$V$，并且$a$和$b$分别属于${\rm ker}(\partial_0^{\mathcal B})$和${\rm ker}(\partial_1^{\mathcal B})$。
既然我们现在具体操作的是这个代表元向量 $y \in E_1 = B_1F_0 \oplus B_0F_1$，我们可以将 $y$ 展开写成张量积空间基底的线性组合：
$$ y = \sum_b b \otimes g_b + \sum_a a \otimes f_a $$
其中 $b$ 和 $a$ 分别遍历 $B_1$ 和 $B_0$ 的**自然基底向量 (basis vectors)**，且 $g_b \in F_0$，$f_a \in F_1$。

对于每一个基底 $b \in B_1$，我们可以将向量 $g_b \in F_0$ 唯一地分解为：
$$ g_b = \partial(h_b) + z_b $$
这里 $\partial(h_b)$ 属于边界空间 $\operatorname{im}(\partial^F_1)$，而 $z_b$ 则是属于同调群代表元生成的空间，即在前提假设下，所有的扭曲映射作用在 $z_b$ 上都有 $\varphi(b,a)z_b = z_b$。

现在，我们构造一个辅助元素 $x = \sum_b b \otimes h_b \in E_2$。我们给原代表元 $y$ 加上这个辅助元素所生成的整体边界（根据定义，这相当于我们在 $[y]$ 的等价类中选取了另一个代表元，完全不会改变 $y$ 在同调群 $H_1(\mathcal{E})$ 中的同调类性质）：
$$ \begin{aligned}
y + \partial(x) &= \sum_b b \otimes g_b + \sum_a a \otimes f_a + \sum_b \partial(b \otimes h_b) \\
&= \sum_b b \otimes (g_b + \partial h_b) + \sum_a a \otimes f_a + \sum_b \sum_{a \in \partial b} a \otimes \varphi(b,a)h_b \\
&= \sum_b b \otimes z_b + \sum_a a \otimes f'_a
\end{aligned} $$
其中，我们在合并同类项后定义了新的系数 $f'_a = f_a + \sum_{b: a \in \partial b} \varphi(b,a)h_b$。我们将这个新的闭链代表元记为 $y' = y + \partial(x)$。

由于 $y'$ 与 $y$ 属于同一个同调类，并且 $y$ 是闭链（即 $\partial(y) = 0$），所以必然也有 $\partial(y') = 0$。根据扭曲乘积复形的边界算子定义，我们对 $y'$ 计算整体边界：
$$ \begin{aligned}
\partial(y') &= \sum_b \partial(b \otimes z_b) + \sum_a \partial(a \otimes f'_a) \\
&= \sum_b \left( b \otimes \partial(z_b) + \sum_{a \in \partial b} a \otimes (\varphi(b,a) \cdot z_b) \right) + \sum_a \left( a \otimes \partial(f'_a) \right)
\end{aligned} $$
由于 $z_b$ 是同调群 $H_0(\mathcal{F})$ 的代表元，它必定是一个闭链，即 $\partial(z_b) = 0$。上式简化为：
$$ \begin{aligned}
\partial(y') &= \sum_b \sum_{a \in \partial b} a \otimes \varphi(b,a)z_b + \sum_a a \otimes \partial(f'_a)
\end{aligned} $$
由于前面定义的 $z_b$ 满足前提条件，使得 $\varphi(b,a)z_b = z_b$，上述第一项的求和可以化简，并交换求和次序（按 $a$ 合并同类项）：
$$ \sum_b \sum_{a \in \partial b} a \otimes z_b = \sum_a \sum_{b: a \in \partial b} a \otimes z_b = \sum_a a \otimes \left(\sum_{b: a \in \partial b} z_b\right) $$
代回原式，我们得到：
$$ \partial(y') = \sum_a a \otimes \left( \sum_{b: a \in \partial b} z_b + \partial(f'_a) \right) = 0 $$
由于所有的基底 $a$ 是线性无关的，要求总和为 0，必须要求括号内的每一项均为 0。即对任意的基底 $a \in B_0$，都有：
$$ \sum_{b: a \in \partial b} z_b + \partial(f'_a) = 0 $$
这等价于：
$$ \sum_{b: a \in \partial b} z_b = \partial(f'_a) $$
注意到等式左边是一个关于 $z_b$（也就是 $H_0(\mathcal{F})$ 代表元）的线性组合，而右边是一个边界项 $\partial(f'_a)$。由于 $H_0(\mathcal{F})$ 的代表元集合在定义上与边界空间 $\operatorname{im}(\partial^F)$ 是线性无关的（即交集为 0），要使这个等式成立，等式两边必须同时为 0！

这意味着：
1.  对任意基底 $a$，都有 $\partial(f'_a) = 0$，这说明所有的 $f'_a$ 都是纤维复形中的闭链，即 $f'_a \in \ker(\partial^F_1)=U$（注意$\mathcal F$是$1$-复形）。
2.  对任意基底 $a \in B_0$，都有 $\sum_{b: a \in \partial b} z_b = 0$。下面用张量积算子给出严格的代数结论：

考虑组合项 $Z=\sum_b b \otimes z_b \in B_1 \otimes F_0$。
对其施加部分边界算子 $(\partial_1^\mathcal{B} \otimes I_{\mathcal{F}}) : B_1 \otimes F_0 \to B_0 \otimes F_0$：
$$ (\partial_1^\mathcal{B} \otimes I_{\mathcal{F}})(Z) = \sum_b \partial_1^\mathcal{B}(b) \otimes z_b = \sum_b \left( \sum_{a \in \partial b} a \right) \otimes z_b $$
改变求和次序并提取基底 $a$：
$$ (\partial_1^\mathcal{B} \otimes I_{\mathcal{F}})(Z) = \sum_a a \otimes \left( \sum_{b: a \in \partial b} z_b \right) $$
代入上述推导结果 $\sum_{b: a \in \partial b} z_b = 0$，直接得到：
$$ (\partial_1^\mathcal{B} \otimes I_{\mathcal{F}})(Z) = 0 $$
这在代数上严格证明了 $Z \in \ker(\partial_1^\mathcal{B} \otimes I_{\mathcal{F}})={\rm ker}(\partial_1^{\mathcal B})\otimes F_0$。再注意到$Z\in B_0\otimes V$，所以$Z\in{\rm ker}(\partial_1^{\mathcal B})\otimes V$。

从而：$$y'=Z+\sum_aa\otimes f_a'$$且其中$a\in B_0={\rm ker}(\partial_0^{\mathcal B}),f_a'\in {\rm ker}(\partial_1^{\mathcal F})=U,Z\in{\rm ker}(\partial_1^{\mathcal B})\otimes V$（注意$Z$的那个字面上的拆解中的$b$并不直接符合属于${\rm ker}(\partial_1^{\mathcal B})$这个要求）。

#### 2. 从“同调”到“闭链”的转化 (Corollary 3.0.2)

同调群的概念在计算**码率 (Rate)** 时非常完美（因为维数可以通过同构直接相加乘），但在计算**码距 (Distance)** 时却束手无策，因为汉明重量（码距）在同构映射下是**不守恒的**。为了计算码距和解码，我们需要极其具体的向量表达式。

**【推论 3.0.2 表述】**
固定 $H_i(\mathcal{B})$ 和 $H_i(\mathcal{F})$ 的任意一组线性无关的代表元（$i \in \{0, 1\}$），并要求 $\mathcal{F}$ 的代表元在扭曲集合 $\varphi$ 下保持不变。
那么，整体复形 $\mathcal{E}$ 的闭链空间（核空间）可以精确地显式写出：
$$ \ker(\partial_1^\mathcal{E}) = {\rm span}(\{ x_1 \otimes y_0 + x_0 \otimes y_1 + \partial_2^\mathcal{E}(z) \}) $$
（原文中没有这个${\rm span}$，后果就是集合中将只有直集态的少量和，显然是不对的）
其中，
*   $x_i$ 遍历 $H_i(\mathcal{B})$ 代表元的线性生成空间。
*   $y_i$ 遍历 $H_i(\mathcal{F})$ 代表元的线性生成空间。
*   $z \in E_2$ 代表整体边界空间的任意前像。

一个核向量 $v$ 是**非平凡的 (non-trivial)** （即它是一个真正的逻辑算子，而不是稳定子），**当且仅当它在这个分解下的代表元部分 $\sum x_1 \otimes y_0 + \sum x_0 \otimes y_1 \neq 0$**（这里补上一个$\sum$的理由同前文补${\rm span}$的理由）。


### 定理 4.1.1 [ZP19]：张量积复形的码距下界 (子复形截断证明)

设 $\mathcal{B}, \mathcal{F}$ 为 1-复形，并设 $\mathcal{E} = \mathcal{B} \otimes \mathcal{F}$。设 $B'_1 \subseteq B_1$ 为一个子空间，使得 $B'_1$ 由 $B_1$ 的一组基的子集张成。定义复形 $\mathcal{B}': B'_1 \to B_0$，其中边界映射是 $\partial_1(\mathcal{B})$ 的限制。类似地定义 $\mathcal{F}'$。那么，$\mathcal{E}' = \mathcal{B}' \otimes \mathcal{F}'$ 是良定义的。换言之，我们从映射 $\partial_1^{\mathcal{B}}, \partial_1^{\mathcal{F}}$ 的矩阵中选择一部分列的子集，然后构成张量积。可以直接看出，该复形是 $\mathcal{E}$ 的子复形，即存在自然的包含映射 $E'_i \to E_i$，并且它与边界映射可交换。

**【定理 4.1.1 表述】 [ZP19]**
设 $\mathcal{B}, \mathcal{F}$ 为 1-复形，且设 $\mathcal{E} = \mathcal{B} \otimes \mathcal{F}$。那么，
$$d_1(\mathcal{E}) \ge \min(d_1(\mathcal{B})d_0(\mathcal{F}), d_0(\mathcal{B})d_1(\mathcal{F}))$$

---

**【定理证明】**

设 $v$ 是一个核向量 (kernel vector)，满足 $\vert v\vert < \min(d_1(\mathcal{B})d_0(\mathcal{F}), d_0(\mathcal{B})d_1(\mathcal{F}))$，并设 $v = \sum_b b \otimes g_b + \sum_a a \otimes f_a$。现在假设两个 0-阶同调都不平庸 (trivial)，即 $d_0(\mathcal{B}) = d_0(\mathcal{F}) = 1$。令 $I = \{b \mid g_b \neq 0\}$，$J = \bigcup_a \text{supp}(f_a)$。那么，根据对 $v$ 的权重假设，我们有 $\vert I\vert \le d_1(\mathcal{B})$，$\vert J\vert \le d_1(\mathcal{F})$。令 $B'_1 = \text{span } I$，$F'_1 = \text{span } J$。因此，$H_1(\mathcal{B}') = H_1(\mathcal{F}') = 0$，由 Künneth 定理可知这蕴含了 $H_1(\mathcal{E}') = 0$。

假设映射 $\partial_1^{\mathcal{B}}: B_1 \to B_0$ 为满射，即 $H_0(\mathcal{B}) = 0$。
设 $v \in \ker(\partial_1^{\mathcal{E}})$ 为一个核向量，满足 $\vert v\vert < d_1(\mathcal{B})d_0(\mathcal{F})$。为了使论证保持一般性，假设 $d_0(\mathcal{F}) = 1$，即有界条件简化为 $\vert v\vert < d_1(\mathcal{B})$（如果 $\mathcal{F}$ 那边的边界映射 $\partial_0^{\mathcal{F}}$ 也是满射，那么这里会导致 $H_1(\mathcal{E})=0$，从而证明已完成）。将 $v$ 展开为张量积基底的线性组合：
$$v = \sum_{b \in I} b \otimes g_b + \sum_{a \in B_0} a \otimes f_a$$
其中 $I = \{b \mid g_b \neq 0\}$ 且 $f_a \in F_1$。由重量假设可知，$\vert I\vert \le \vert v\vert < d_1(\mathcal{B})$。

此处对 $\mathcal{F}$ 不作截断限制，即令 $\mathcal{F}' = \mathcal{F}$。令 $B_1^I = \operatorname{span}(I) \subseteq B_1$。由于 $\vert I\vert < d_1(\mathcal{B})$，子空间 $B_1^I$ 无法包含 $\mathcal{B}$ 中的任何非零闭链，故：
$$\ker(\partial_1^{\mathcal{B}}|_{B_1^I}) = \{0\} \implies H_1(B_1^I \xrightarrow{\partial} B_0) = 0$$

定义像空间 $V = \partial_1^{\mathcal{B}}(B_1^I) \subseteq B_0$。由于 $\partial_1^{\mathcal{B}}: B_1 \to B_0$ 是满射，必定存在 $B_1 \setminus I$ 中的一组基底子集 $K$，使得集合 $\{ \partial_1^{\mathcal{B}}(k) \mid k \in K \}$ 构成 $V$ 在 $B_0$ 中的补空间的基底。即有直和分解：
$$B_0 = V \oplus \operatorname{span}(\{ \partial_1^{\mathcal{B}}(k) \mid k \in K \})$$

构造扩张子空间 $B'_1 = B_1^I \oplus \operatorname{span}(K)$，并定义相应的限制子复形 $\mathcal{B}' = (B'_1 \xrightarrow{\partial_1^{\mathcal{B}'}} B_0)$。

以下验证 $\mathcal{B}'$ 的同调群性质：

*   **满射性验证 ($H_0(\mathcal{B}') = 0$)**：
根据直和分解的构造，有 $\operatorname{im}(\partial_1^{\mathcal{B}'}) = \partial_1^{\mathcal{B}}(B_1^I) + \partial_1^{\mathcal{B}}(\operatorname{span}(K)) = V \oplus \text{补空间} = B_0$。映射 $\partial_1^{\mathcal{B}'}$ 为满射，故 $H_0(\mathcal{B}') = 0$。

*   **无核性验证 ($H_1(\mathcal{B}') = 0$)**：
设存在 $x \in B_1^I$ 与 $y \in \operatorname{span}(K)$ 使得 $\partial_1^{\mathcal{B}'}(x + y) = 0$。在 $\mathbb{F}_2$ 上，这等价于 $\partial_1^{\mathcal{B}}(x) = \partial_1^{\mathcal{B}}(y)$。
由于 $\partial_1^{\mathcal{B}}(x) \in V$ 且 $\partial_1^{\mathcal{B}}(y) \in \operatorname{span}(\{ \partial_1^{\mathcal{B}}(k) \mid k \in K \})$，由直和性质可知 $\partial_1^{\mathcal{B}}(x) = \partial_1^{\mathcal{B}}(y) = 0$。
结合 $\ker(\partial_1^{\mathcal{B}}|_{B_1^I}) = \{0\}$ 得到 $x = 0$；由集合 $K$ 的像向量线性无关得到 $y = 0$。因此 $\ker(\partial_1^{\mathcal{B}'}) = \{0\}$，即 $H_1(\mathcal{B}') = 0$。

定义总张量积子复形 $\mathcal{E}' = \mathcal{B}' \otimes \mathcal{F}$。
由 Künneth 定理，$\mathcal{E}'$ 的 1-阶同调群为：
$$H_1(\mathcal{E}') \cong (H_1(\mathcal{B}') \otimes H_0(\mathcal{F})) \oplus (H_0(\mathcal{B}') \otimes H_1(\mathcal{F}))$$
代入上述同调结果，得到：
$$H_1(\mathcal{E}') \cong (0 \otimes H_0(\mathcal{F})) \oplus (0 \otimes H_1(\mathcal{F})) = 0$$

根据构造，两种情况下都有 $v \in \mathcal{E}'$，而且 $v \in \ker(\partial_1^{\mathcal{E}'})$。但由于 $H_1(\mathcal{E}') = 0$，我们得出 $v \in \text{im}(\partial_2^{\mathcal{E}'}) \subseteq \text{im}(\partial_2^{\mathcal{E}})$。这证明了 $v$ 是一个边界元素 (boundary element)。 $\blacksquare$

### 扭曲积的局限性与推论 4.1.2

这个证明不能推广到扭曲积 (twisted product)，因为 $F'_1$ 在 $F_1$ 的自同构群作用下不一定是封闭的。如果我们确实能保证 $F'_1$ 是封闭的，那么该论证就依然成立。如上所述，如果 $H_0(\mathcal{B}) = 0$，那么上述观察可以得出以下推论。

**推论 4.1.2.** 设 $\mathcal{B}, \mathcal{F}$ 为 1-复形，且设 $\mathcal{E} = \mathcal{B} \otimes_\varphi \mathcal{F}$。如果 $H_0(\mathcal{B}) = 0$，那么 $d_1(\mathcal{E}) \ge d_1(\mathcal{B})$。

这种设定很重要，因为正如我们所见，[EKZ20] 和 [HHO21] 中的构造都满足 $H_0(\mathcal{B}) = 0$。这总是可以通过将 $B_0$ 缩小到像空间来实现，即令 $B'_0 := \operatorname{im}(\partial_1)$。

然而，除非 $\dim(F_i)$ 相比于 $\mathcal{B}$ 非常小，否则上述下界是不够的；而且即便如此，该构造也只有在单侧是好的 (good only on one side)，因为另一侧的距离将依赖于 $d_1(\mathcal{F}^*)$。

为了突破 $\sqrt{n}$ 的限制，我们希望得到类似 $d_1(\mathcal{E}) \ge d_1(\mathcal{B})d_1(\mathcal{F})^c$ 的结果（其中 $c$ 为某个常数）。我们将在下文看到，通常的（无扭曲）张量积无法实现这一点，而这正是**扭曲 (twists)** 的重要性所在。

虽然我们目前还没有通用的界限，但对于循环映射 $\mathcal{F} : \mathbb{Z}_l \to \mathbb{Z}_l$ 这一特定情况，通过巧妙地选择扭曲，[HHO21] 证明了大约为 $O(d_1(\mathcal{B})\sqrt{l})$ 的界，而 [PK21] 证明了 $O(\min(d_1(\mathcal{B}), d_0(\mathcal{B})l))$ 的最优线性界。

### 引理 4.2.1 [ZP19]

**引理 4.2.1.** [ZP19] 设 $\mathcal{E} = \mathcal{C} \otimes \mathcal{D}$。那么，$d_k(\mathcal{E}) \le \min_i d_i(\mathcal{C})d_{k-i}(\mathcal{D})$。

书中原文的证明调用了引理 3.0.2，但是引理 3.0.2 的证明过程依赖于 $\mathcal{C}$ 和 $\mathcal{D}$ 是 1-复形，所以根本不能直接调用。

---

**【严密的引理 4.2.1 证明】**

**1. 选取极小代表元**：
设 $j$ 为使 $d_i(\mathcal{C})d_{k-i}(\mathcal{D})$ 取得最小值的索引。根据距离的定义，在同调群 $H_j(\mathcal{C})$ 中必定存在一个非平庸的同调类，我们选取其重量最小的代表元 $x \in C_j$。满足：
*   $x \in \ker(\partial_j^{\mathcal{C}})$ （即 $\partial_j^{\mathcal{C}}(x) = 0$）
*   $[x] \neq 0$
*   $\vert{}x\vert{} = d_j(\mathcal{C})$

同理，在 $D_{k-j}$ 中选取代表元 $y$，满足 $\partial_{k-j}^{\mathcal{D}}(y) = 0$，$\vert{}y\vert{} = d_{k-j}(\mathcal{D})$ 且 $[y] \neq 0$。

**2. 证明 $x \otimes y$ 是总复形 $\mathcal{E}$ 中的闭链（循环）**：
在一般的张量积复形 $\mathcal{E} = \mathcal{C} \otimes \mathcal{D}$ 中，对于任意阶数的 $k$-链，边界算子的通用定义为：
$$\partial_k^{\mathcal{E}}(x \otimes y) = \partial_j^{\mathcal{C}}(x) \otimes y + (-1)^j x \otimes \partial_{k-j}^{\mathcal{D}}(y)$$
（在特征为 2 的域 $\mathbb{F}_2$ 上，符号 $(-1)^j$ 可以直接省略为 $+$）。将 $\partial^{\mathcal{C}}(x) = 0$ 和 $\partial^{\mathcal{D}}(y) = 0$ 代入：
$$\partial_k^{\mathcal{E}}(x \otimes y) = 0 \otimes y + x \otimes 0 = 0$$
这就严格证明了 $x \otimes y \in \ker(\partial_k^{\mathcal{E}})$，即它是一个合法的核向量。

**3. 调用 Künneth 定理（定理 2.1.2）证明其非平庸性**：
对于定义在域（如 $\mathbb{F}_2$）上的任意链复形，定理 2.1.2 指出，同调群的张量积与张量积的同调群是严格同构的（Tor 函子项在域上恒为 0）：
$$H_k(\mathcal{C} \otimes \mathcal{D}) \cong \bigoplus_{i} H_i(\mathcal{C}) \otimes H_{k-i}(\mathcal{D})$$
该同构映射由典范映射 $[x] \otimes [y] \mapsto [x \otimes y]$ 给出。因为我们在第一步保证了 $[x] \neq 0$ 且 $[y] \neq 0$，由张量积的非退化性，必然有：
$$[x \otimes y] \neq 0$$
这就严格证明了 $x \otimes y$ 是 $\mathcal{E}$ 中的一个非平凡核向量 (non-trivial kernel vector)，即它不是纯边界。

**4. 结论**：
既然 $x \otimes y$ 是一个非平凡核向量，那么 $\mathcal{E}_k$ 中最小非平凡核向量的重量 $d_k(\mathcal{E})$ 必然小于等于 $x \otimes y$ 的重量：
$$d_k(\mathcal{E}) \le \vert{}x \otimes y\vert{} = \vert{}x\vert{} \cdot \vert{}y\vert{} = d_j(\mathcal{C})d_{k-j}(\mathcal{D})$$
证毕。 $\blacksquare$

---

现在我们可以明白为什么这个上界在扭曲积（twisted products）的情况下会失效。大部分论证依然成立，广义 Künneth 定理（这里指的是第三章那个引理和推论）也依然成立，但我们不能任意选择 $H_i(\mathcal{F})$ 的代表元，因为它们必须在 $\varphi$ 作用下保持不变（invariant）。在无扭曲的情况下，所有的扭曲映射都是恒等映射，因此这个条件是自然满足的（trivial），但在一般情况下并非如此。例如，当 $\mathcal{F} = \mathbb{Z}_l \to \mathbb{Z}_l$ 时（这里再次出现符号滥用，这里$\mathcal F$是指这样一个1-复形，其中$F_1$和$F_0$都是$\mathcal F_2^l$，并且$\partial_1v_i=a_i+a_{i+1}$，而这里$\varphi$的行为是循环），在置换下唯一保持不变的向量是全 1 特征向量。因此，虽然 $d_0(\mathcal{F}) = 1$（原文写的是$d_i(\mathcal F)=1$，显然不对，按照刚刚的定义，显然$d_1(\mathcal F)=l$），但唯一不变的代表元是 $\mathbb{I}_0$，它具有很大的重量，即 $l$。因此，该界限仅仅给出了一个 $l \cdot \min(d_1(\mathcal{B}), d_0(\mathcal{B}))$ 的上界。这个界限并不弱，因为实际上 [PK21] 已经（在常数因子的意义下）达到了这个界限。让我们更详细地分析一下上述上界在无扭曲情况下的来源，以及如何跨越这个障碍。我们从 $v \in H_1(\mathcal{B})$ 和 $g \in H_0(\mathcal{F})$ 开始。我们可以假设 $\vert{}g\vert{} = 1$，即 $g$ 是一个基向量。在无扭曲的情况下，$v \otimes g \in H_1(\mathcal{E})$，因此我们得到了一个重量为 $d_1(\mathcal{B})$ 的非平凡元素。这在扭曲的情况下不再成立。

$$\partial(v \otimes g) = \sum_{b \in v} \sum_{a \in \partial(b)} a \otimes \varphi(b, a)g \quad (4.1)$$
$$= \sum_a a\otimes\sum_{b \in v \cap a \in \partial(b)} \varphi(b, a)g \quad (4.2)$$
$$= \sum_a a\otimes\partial_1^{\mathcal{F}}(f_a) = \partial_1^{\mathcal{E}} \left( \sum_a a \otimes f_a \right) \quad (4.3)$$

现在说明为什么(4.2)等于(4.3)，首先我们说明对于每个 $a$，向量 $g_a := \sum_{b \in v \cap a \in \partial(b)} \varphi(b, a)g \in F_0$ 具有偶数重量：注意$\partial_1^{\mathcal B}v=0$，而$$\partial_1^{\mathcal B}v=\sum_{b\in v}\partial b=\sum_{b\in v}\sum_{a\in\partial b}a=\sum_aa\sum_{b\in v\land a\in\partial b}1$$

注意$a$是基矢枚举，所以$\sum_{b\in v\land a\in\partial b}1\in2\mathbb Z$，而注意到$g_a$由$\varphi(b,a)g$求和而得，且是偶数项求和，并且$\varphi$的行为不改变重量，故结果重量是偶数。

注意这里的讨论都是基于$\mathcal F=\mathbb Z_l\to\mathbb Z_l$这个设定，$g_a$重量是偶数将直接意味着其是某个$f_a\in F_1$的边界。

以上(4.1)到(4.3)这一串等式意味着这个新向量$w = v \otimes g + \sum_{a \in B_0} a \otimes f_a$ 构成了总复形 $\mathcal{E}$ 中的一个核向量，现在我们说明这个向量非平凡。假定$w$平凡，从而$w=\partial_{2}^{\mathcal E}(u),u=\sum_cc\otimes h_c$，于是$v\otimes g=\sum_cc\otimes\partial h_c$，而注意到$c$是在枚举$B_1$的基矢，这意味着$g=\partial\sum_{c\in v}h_c$，这和$g$的非平凡性矛盾。既然$w$是非平凡核矢，故它直接确立了张量码距离的上界：$d_1(\mathcal{E}) \le d_1(\mathcal{B}) + \sum_a \vert{}f_a\vert{}$。

现在注意到，如果$\varphi$的扭曲程度小于等于$r$（也就是说，$\varphi(?,?)(a_i)$的下标散布不超过范围$r$），意味着$\sum_{b\in v\land a\in\partial b}\varphi(b,a)(g)$向量中的1的散步范围在$r$以内，故$|f_a|\leq r$，于是会有上界$d_1(\mathcal B)+n_0(\mathcal B)r$

### PK 和 HHO 的证明策略

推广上述例子，我们可以说，对于任何满足 $h + v \in \ker(\partial_1^{\mathcal{E}})$ 的 $h \in B_1F_0$，存在唯一一个这样低重量的 $v$，它仅仅依赖于 $\partial h$。这两个证明都有两个主要部分。

*   **（扩展性）** 两者都使用了扩展性（尽管形式不同），以论证如果 $\vert{}h\vert{}$ 太小，那么 $\vert{}\partial h\vert{} = \Omega(\vert{}h\vert{})$。
*   **（结构）** HHO 利用了扭曲上的结构约束来证明 $\vert{}v\vert{} = \tilde{\Omega}(\sqrt{l}\vert{}\partial h\vert{})$，而 PK 利用了 $\mathbb{Z}_l$ 固有的对称性来论证 $\vert{}v\vert{} = \Omega(l\vert{}\partial h\vert{})$。

### 第 5 章解码 (Decoding)

在本章中，我们将考察量子 LDPC 码的最新解码算法。第一个是由 [EKZ20] 提出的一种通用归约（reduction），它将解码两个码的张量积（tensor product）的问题，归约为分别对这两个单独的码进行解码。我们要学习的第二个是 [HHO21] 提出的算法，它为扭曲积（twisted product）提供了解码器。让我们从解码问题的经典定义开始，并看看它的量子推广是什么。

#### 5.1 经典解码

经典设定是这样的：你有一个码字 $x$，你希望通过某种通信信道进行传输。通信过程会引入错误，对于某个错误 $e$，最终输出为 $x + e$。唯一解码（unique decoding）的问题是在给定 $x + e$ 的情况下恢复 $x$，或者等价地说，计算出 $e$。有不同的错误模型试图捕捉这类错误是如何发生的。例如，错误可能是随机的，也可能是对抗性的（adversarial）。在随机情况下，我们要求解码器以高概率工作。一个经典的例子是香农的二元对称信道（binary symmetric channel），其中码字是布尔字符串，每个比特以概率 $p$ 发生翻转。我们将坚持使用汉明模型（Hamming model），在该模型中，错误可能是对抗性的，但我们对这种错误的数量有一个上限。让我们坚持使用我们的二元线性码模型，看看确切的问题是什么。该码是一个线性子空间 $\mathcal{C} \subseteq \mathbb{F}_2^n$，其性质为：对于每一个非零码字 $x \in \mathcal{C}$，都有 $\vert{}x\vert{} \ge d$。（唯一）解码半径（decoding radius）是能够进行唯一解码的最大错误量。因为任意两个码字之间的距离都相隔 $d$，如果错误量 $< d/2$，那么存在唯一一个距离受扰动词（perturbed word）最近的码字。因此，解码可以看作是一个计算过程，即计算 $\arg\min_{y \in \mathcal{C}} \vert{}y + x'\vert{}$，其中 $x' = x + e$ 是我们的输入，并保证 $\vert{}e\vert{} < d/2$，而使距离最小化的就是那个唯一的码字 $x$。虽然这是一个信息论的界限，但在算法上实现这一点具有挑战性，通常只要能解码到该距离的某个常数因子（constant factor）作为半径的错误，我们就心满意足了。

**经典解码**
对于码 $\mathcal{C} \subseteq \mathbb{F}_2^n$，设 $x \in \mathcal{C}$ 且 $e \in \mathbb{F}_2^n$，使得 $\vert{}e\vert{} < d/2$
*   **输入**：$s := x + e$
*   **输出**：$x$ 或者等价地，$y = \arg\min_{y \in \mathcal{C}} \vert{}y + s\vert{}$

### 5.2 量子解码 (Quantum Decoding)

#### 1. 物理架构与 2-复形的对应
作者在一开始给出了一个 2-复形 $\mathcal{C} : C_2 \xrightarrow{\partial_2} C_1 \xrightarrow{\partial_1} C_0$。这实际上是把物理上的 CSS 量子码完全代数化了：
*   **$C_1$（1-链空间）**：这就是你手里的物理量子比特（Qubits）所在的空间。错误 $e$ 就发生在这里。
*   **$\partial_1$（边界算子）**：这代表物理测量（Syndrome Measurement）。当你对系统进行奇偶性校验时，你就相当于对错误 $e$ 施加了 $\partial_1$ 算子，得到的边界 $s = \partial_1(e) \in C_0$ 就是你观测到的“综合式（Syndrome）”。
*   **$\operatorname{im}(\partial_2)$（纯边界空间）**：这是量子纠错独有的稳定子（Stabilizers）。在物理上，这些操作对逻辑量子态没有任何影响，相当于恒等变换。
*   **$W$（逻辑空间）**：也就是一阶同调群 $H_1(\mathcal{C}) = \ker(\partial_1) / \operatorname{im}(\partial_2)$。这才是我们真正用来存储信息的逻辑态。

#### 2. 黑盒与不可克隆定理（物理限制）
在经典通信中，收件人是可以直接拿到“被污染的数据 $x+e$”的。但量子力学里有不可克隆定理和波函数坍缩：如果系统处于逻辑态 $x$ 叠加了错误 $e$，你绝对不能直接去“看”这堆量子比特，一看逻辑态就彻底坍缩没了。这就是作者说的“黑盒”：你只能通过非破坏性的测量（即施加 $\partial_1$）把错误 $e$ 的“影子” $s$ 提取出来，然后在这个黑盒外面纯靠数学去猜 $e$ 是什么。

#### 3. 最核心的差别：量子输出 vs 经典输出
仔细看作者给出的那两个表格。在经典解码里，输出是：“找到那个唯一的错误 $e$”。但在量子解码里，输出变成了：“找到任意一个 $y$，满足 $y = e + \partial_2(w)$”。这正是我们上一轮讨论的简并性（Degeneracy）的严格代数表达！
在经典世界，错了一定位就是错了一定位，你必须精准找到它。在量子世界，因为稳定子 $\operatorname{im}(\partial_2)$ 的存在，错误 $e$ 和错误 $e + \partial_2(w)$ 对逻辑态的破坏是物理上绝对等价的。如果你猜的错误是 $y = e + \partial_2(w)$，那么当你施加修正操作 $y$ 时，系统剩下的总错误是 $y + e = \partial_2(w)$（注意在 $\mathbb{F}_2$ 上加法和减法是一样的）。因为 $\partial_2(w)$ 是一个稳定子，它对系统没有副作用，所以量子态被完美恢复了。因此，量子解码的本质不是找 $e$，而是找 $e$ 所在的同调类 $[e]$。

#### 4. 那个极其突兀的距离条件：$\vert{}y\vert{} < d_1(\mathcal{E}) - \vert{}e\vert{}$
这是全节最反直觉的一句话，作者说“第二个条件利用距离性质强制该差值为一个余边界”。我们用严格的代数来证明为什么这个不等式能保证你找对同调类。
假设你通过某种算法找到了一个猜测值 $y$，它满足 $\partial_1(y) = s$。我们来看真实错误 $e$ 和你的猜测 $y$ 组成的合成向量 $z = y + e$：
*   **它是一个闭链（核向量）**： 因为 $\partial_1(z) = \partial_1(y) + \partial_1(e) = s + s = 0$。所以 $z \in \ker(\partial_1)$。
*   **它的重量上限**： 根据三角不等式，其汉明重量 $\vert{}z\vert{} \le \vert{}y\vert{} + \vert{}e\vert{}$。
*   **触发距离红线**： 因为条件规定了 $\vert{}y\vert{} < d_1(\mathcal{E}) - \vert{}e\vert{}$，代入进去，必然得到：
$$\vert{}z\vert{} < d_1(\mathcal{E})$$
终极逻辑闭环出现了：$d_1(\mathcal{E})$ 是什么？是逻辑空间（同调群）的最小距离！也就是任何能够改变逻辑态的非平凡核向量，其重量至少是 $d_1$。但现在的 $z$ 作为一个核向量，它的重量严格小于 $d_1$！这在代数上只有一个解释：$z$ 绝对不可能是非平凡的，它只能是一个平庸的纯边界！即 $z \in \operatorname{im}(\partial_2)$，所以必然存在某个 $w$，使得 $z = \partial_2(w)$。把 $z = y + e$ 代回去，就得到了 $y = e + \partial_2(w)$。

**总结这一节的人话版本：**
量子解码由于物理限制，只能拿到观测值 $s$。为了恢复量子态，你不需要（也不可能）找回最初的那个确切错误 $e$。你只需要在满足 $\partial_1(y) = s$ 的候选者中，挑一个足够短的 $y$（短到 $\vert{}y\vert{} + \vert{}e\vert{} < d_1$）。只要它足够短，同调代数的性质就会自动向你保证：你找到的 $y$ 和真实的 $e$ 在拓扑上是完全等价的，用它去纠错就万无一失了。
